# Conditional Statements

Conditional statements are used for decision making.

They execute code according to whether a condition is True or False.

Example:

If condition is True:
    execute code

Otherwise:
    skip code

```python
# Simple if statement

age = 20

if age >= 18:
    print("You can vote")
```

# if Statement

if statement checks a condition.

If condition is True, the code inside if block executes.

Syntax:

if condition:
    statement

```python
marks = 80

if marks >= 40:
    print("Pass")
```

# Indentation

Python uses indentation (spaces) to define blocks.

Correct:

if True:
    print("Hello")

Without proper indentation Python gives error.

```python
if True:
    print("Hello Python")
```

# if-else Statement

if-else handles two situations.

If condition is True:
    if block executes

Otherwise:
    else block executes

```python
age = int(input("Enter age: "))

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

# elif Statement

elif is used when we have multiple conditions.

Python checks conditions from top to bottom.

The first True condition executes.

```python
marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")

elif marks >= 75:
    print("Grade B")

elif marks >= 40:
    print("Grade C")

else:
    print("Fail")
```

# Nested if

When one if statement is written inside another if statement.

Example:

if condition:
    if condition:
        statement

```python
age = 20
license = True

if age >= 18:

    if license:
        print("Can drive")

    else:
        print("Need license")

else:
    print("Age not allowed")
```

```python
# ==================================
# Assignment Solutions
# Conditional Statements
# ==================================


# Q1. Positive Negative Zero

num = int(input("Enter number: "))

if num > 0:
    print("Positive")

elif num < 0:
    print("Negative")

else:
    print("Zero")



# Q2. Voting Eligibility

age = int(input("Enter age: "))

if age >= 18:
    print("Eligible for vote")

else:
    print("Not eligible")



# Q3. Greater among two numbers

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("First number is greater")

elif b > a:
    print("Second number is greater")

else:
    print("Both are equal")



# Q4. Even Odd

num = int(input("Enter number: "))

if num % 2 == 0:
    print("Even")

else:
    print("Odd")



# Q5. Pass Fail

marks = int(input("Enter marks: "))

if marks >= 40:
    print("Pass")

else:
    print("Fail")



# Q6. Grade System

marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")

elif marks >= 75:
    print("Grade B")

elif marks >= 40:
    print("Grade C")

else:
    print("Fail")



# Q7. Login System

username = input("Username: ")
password = input("Password: ")

if username == "python" and password == "123":
    print("Login Successful")

else:
    print("Invalid Login")



# Q8. Electricity Bill Category

units = int(input("Enter units: "))

if units <= 100:
    print("Low usage")

elif units <= 300:
    print("Medium usage")

else:
    print("High usage")
```

# Interview Notes

## Difference between if, elif and else

if:
Used for first condition.

elif:
Used for checking additional conditions.

else:
Runs when all conditions are false.


## Importance of indentation

Indentation defines code blocks in Python.

It improves readability and is required by Python syntax.
