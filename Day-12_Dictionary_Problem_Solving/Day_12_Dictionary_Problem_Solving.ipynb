## 📌 Dictionary

A dictionary stores data in the form of key-value pairs.

Example:

student = {
    "name": "Deepika",
    "age": 20
}

Here:
- name and age are keys
- Deepika and 20 are values


Properties:
- Mutable
- Ordered
- Keys must be unique

```python
# Creating a dictionary

student = {
    "name": "Deepika",
    "branch": "AIML",
    "year": 3
}

print(student)
```

```python
student = {
    "name": "Deepika",
    "age": 20
}

# Accessing value using key

print(student["name"])
print(student["age"])
```

## Dictionary Methods

keys()
→ returns all keys

values()
→ returns all values

items()
→ returns key-value pairs

get()
→ safely accesses values

update()
→ modifies dictionary

```python
student = {
    "name": "Deepika",
    "age": 20,
    "course": "AIML"
}

print(student.keys())

print(student.values())

print(student.items())
```

```python
student = {
    "name": "Deepika"
}

# get() avoids error if key is missing

print(student.get("name"))

print(student.get("age"))
```

```python
student = {
    "name": "Deepika",
    "age": 20
}

# Updating existing value and adding new key

student.update({
    "age": 21,
    "city": "Jabalpur"
})

print(student)
```

## 📌 Nested Dictionary

A dictionary inside another dictionary is called a nested dictionary.

It is useful for storing structured data.

```python
students = {

    "student1": {
        "name": "A",
        "marks": 90
    },

    "student2": {
        "name": "B",
        "marks": 85
    }
}

# Accessing nested value

print(students["student1"]["marks"])
```

## 📌 Frequency Counting

Frequency counting means finding how many times each element appears.

Example:

Input:
[1,2,2,3]

Output:
1 → 1 time
2 → 2 times
3 → 1 time

Dictionary is used because it stores mapping:
element → count

```python
numbers = [1,2,2,3,3,3]

frequency = {}

# Traverse every element

for num in numbers:

    # Increase count if exists else start from 1
    frequency[num] = frequency.get(num,0) + 1


print(frequency)
```

## 📌 First Non-Repeating Element

Find the first element whose frequency is 1.

Example:

Input:
[1,2,3,2,1]

Output:
3

```python
numbers = [1,2,3,2,1]

frequency = {}

for num in numbers:
    frequency[num] = frequency.get(num,0)+1


for num in numbers:

    if frequency[num] == 1:
        print(num)
        break
```

## 📌 Character Frequency

Count occurrence of each character in a string.

Example:

Input:
"python"

Output:
p:1
y:1
t:1
...

```python
text = "programming"

frequency = {}

for ch in text:

    frequency[ch] = frequency.get(ch,0)+1


print(frequency)
```

## 📌 Maximum Frequency Element

Find the element which appears maximum number of times.

```python
numbers = [1,2,2,3,3,3]

frequency = {}

for num in numbers:
    frequency[num] = frequency.get(num,0)+1


maximum = 0
answer = None


for key,value in frequency.items():

    if value > maximum:
        maximum = value
        answer = key


print(answer)
```
