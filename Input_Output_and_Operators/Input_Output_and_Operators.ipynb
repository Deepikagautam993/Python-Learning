# 📌 Bitwise Operators in Python

Bitwise operators perform operations on the binary representation of numbers.

Binary uses only two values:

0 and 1

Example:

Decimal:
5

Binary:
101

```python
# Convert decimal number into binary

num = 5

print(bin(num))
```

## 1. Bitwise AND (&)

AND compares each bit.

Rule:

1 & 1 = 1
1 & 0 = 0
0 & 1 = 0
0 & 0 = 0

Both bits must be 1 to get result 1.

```python
# Bitwise AND

a = 5
b = 3

# Binary:
# 5 = 101
# 3 = 011

print(a & b)
```

## 2. Bitwise OR (|)

Rule:

If any one bit is 1, result is 1.

Example:

101
011
---
111

Output = 7

```python
# Bitwise OR

a = 5
b = 3

print(a | b)
```

## 3. Bitwise XOR (^)

Rule:

Same bits → 0

Different bits → 1

Example:

101
011
---
110

Output = 6

```python
# Bitwise XOR

a = 5
b = 3

print(a ^ b)
```

## 4. Bitwise NOT (~)

NOT reverses bits.

Formula:

~n = -(n+1)

Example:

~5 = -(5+1)
   = -6

```python
# Bitwise NOT

num = 5

print(~num)
```

## 5. Left Shift (<<)

Moves bits to left.

Each left shift multiplies number by 2.

Example:

5 << 1

5 becomes 10

```python
# Left Shift

num = 5

print(num << 1)
```

## 6. Right Shift (>>)

Moves bits to right.

Each right shift divides number by 2.

Example:

10 >> 1

10 becomes 5

```python
# Right Shift

num = 10

print(num >> 1)
```

```python
# =====================================
# Input, Output and Operators
# Assignment Solutions
# =====================================


# Q1. Create a login system using username and password

username = input("Enter username: ")
password = input("Enter password: ")

if username == "admin" and password == "123":
    print("Login Successful")
else:
    print("Invalid Login")



# Q2. Create a simple calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)
print("Division:", num1 / num2)



# Q3. Take age input and check Adult or Minor

age = int(input("Enter your age: "))

if age >= 18:
    print("Adult")
else:
    print("Minor")



# Q4. Take marks input and print grade

marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 40:
    print("Grade C")
else:
    print("Fail")



# Q5. Create a program using logical operators

age = int(input("Enter age: "))
has_id = True

if age >= 18 and has_id:
    print("You can enter")
else:
    print("Entry not allowed")
```

```python
# Q6. Check even or odd

number = int(input("Enter number: "))

if number % 2 == 0:
    print("Even")
else:
    print("Odd")



# Q7. Compare two numbers

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("First number is greater")
elif b > a:
    print("Second number is greater")
else:
    print("Both are equal")



# Q8. Student pass fail program

marks = int(input("Enter marks: "))

if marks >= 40:
    print("Pass")
else:
    print("Fail")
```
