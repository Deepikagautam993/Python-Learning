## 📌 What is a Function?

A function is a reusable block of code that runs only when it is called.

### Why we use functions?
- Code reuse
- Better structure
- Less repetition

```python
# defining a simple function

def greet():
    # this line prints message
    print("Hello Deepika")

# calling function
greet()
```

## 📌 Function with Parameters

Parameters allow us to pass data into a function.

```python
def greet(name):
    # name is parameter
    print("Hello", name)

greet("Deepika")
greet("Python Learner")
```

## 📌 Return Statement

return is used to send value back from function.

```python
def add(a, b):
    # returning sum of two numbers
    return a + b

result = add(10, 20)

print(result)
```

## 📌 Default Parameter

If no value is passed, default value is used.

```python
def greet(name="User"):
    # default value is User
    print("Hello", name)

greet()
greet("Deepika")
```

## 📌 Multiple Parameters

We can pass multiple values in a function.

```python
def add(a, b, c):
    # adding three numbers
    return a + b + c

print(add(10, 20, 30))
```

## 📌 *args (Variable Arguments)

Used when number of inputs is unknown.

```python
def add(*numbers):

    total = 0

    # loop through all values
    for num in numbers:
        total += num

    return total

print(add(1, 2, 3, 4, 5))
```

## 📌 **kwargs

Used for key-value type arguments.

```python
def student(**data):

    # printing key-value pairs
    for key, value in data.items():
        print(key, value)

student(name="Deepika", age=20, branch="AIML")
```

## 📌 Lambda Function

A lambda function is a small one-line function.

```python
square = lambda x: x * x

print(square(5))
```

```python
# 1. Multiply function
def multiply(a, b):
    return a * b

print(multiply(2, 3))


# 2. Max using args
def maximum(*nums):
    return max(nums)

print(maximum(1, 5, 3, 9))


# 3. kwargs example
def student(**data):
    print(data)

student(name="Deepika", age=20)


# 4. Lambda
square = lambda x: x*x
print(square(6))


# 5. Default parameter
def greet(name="User"):
    print("Hello", name)

greet()
```

```python
#Q1: Check Even or Odd using Function

def check_even_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(check_even_odd(10))
print(check_even_odd(7))
```

```python
#Q2: Factorial using Function

def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

print(factorial(5))
```

```python
#Q3: Find Maximum using Function

def find_max(numbers):
    max_val = numbers[0]

    for num in numbers:
        if num > max_val:
            max_val = num

    return max_val

print(find_max([10, 5, 20, 8]))
```

```python
#Q4: Palindrome Check Function

def is_palindrome(text):
    if text == text[::-1]:
        return True
    else:
        return False

print(is_palindrome("madam"))
print(is_palindrome("python"))
```

```python
# Q5: Sum of List using Function

def sum_list(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(sum_list([1,2,3,4,5]))
```

```python
# Q6: Count vowels in string

def count_vowels(text):
    vowels = "aeiou"
    count = 0

    for ch in text:
        if ch.lower() in vowels:
            count += 1

    return count

print(count_vowels("python programming"))
```

🧮 Mini Project 1: Simple Calculator (Function-Based)
📌 Features:
Addition
Subtraction
Multiplication
Division

```python
# Simple Calculator using Functions

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b


# User Input
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

print("\nChoose Operation:")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

choice = int(input("Enter choice (1-4): "))


if choice == 1:
    print("Result:", add(num1, num2))

elif choice == 2:
    print("Result:", subtract(num1, num2))

elif choice == 3:
    print("Result:", multiply(num1, num2))

elif choice == 4:
    print("Result:", divide(num1, num2))

else:
    print("Invalid Choice")
```

🧠 Mini Project 2: Quiz App (Function-Based)
📌 Features:
Multiple questions
Score tracking
Final result

```python
# Quiz App using Functions

def ask_question(question, answer):
    user_answer = input(question + " ")

    if user_answer.lower() == answer.lower():
        print("Correct 👍")
        return 1
    else:
        print("Wrong ❌ Correct answer is:", answer)
        return 0


score = 0

print("🎯 Welcome to Quiz App\n")

score += ask_question("Q1: Capital of India?", "Delhi")
score += ask_question("Q2: 2 + 2 = ?", "4")
score += ask_question("Q3: Python is ___ language?", "Programming")

print("\nFinal Score:", score)
```
