## 🐍 What is Python?

**Python is a high-level, interpreted and general-purpose programming language.**

It is used to communicate with computers and create different types of applications.

Python is simple, readable and beginner-friendly.

### Python is used in:

- Web Development
- Data Science
- Artificial Intelligence
- Machine Learning
- Automation

```python
# My first Python program

print("Hello World")

print("I am learning Python")
```

## ⭐ Features of Python

### 1. Easy To Learn

Python has simple syntax which is easy to understand.

Example:

`print("Hello")`

### 2. Readable Language

Python code looks close to English language.

### 3. Interpreted Language

Python code is executed by an interpreter line by line.

### 4. General Purpose Language

Python can be used for different applications.

```python
print("Python")

print(100)

print(10 + 20)
```

## 📝 Comments in Python

Comments are notes written for programmers.

Python ignores comments during execution.

### Single Line Comment:

Use:

#

Example:

# This is a comment

```python
# This is my first comment

print("Learning Python")
```

## ✨ Escape Sequence

Escape sequences are special characters used for formatting output.

### New Line

`\n`

Moves text to the next line.

### Tab

`\t`

Adds space between text.

```python
print("Hello\nPython")

print("Python\tJava")
```

## ⚡ Python Syntax Rules

### 1. Case Sensitive

Python treats uppercase and lowercase differently.

Example:

name ≠ Name


### 2. Indentation

Spaces at the beginning of a line are important in Python.

```python
name = "Deepika"

print(name)
```

## ❌ Common Errors

### 1. Syntax Error

Occurs when Python rules are broken.

### 2. Name Error

Occurs when a variable is not defined.

### 3. Indentation Error

Occurs due to incorrect spacing.
