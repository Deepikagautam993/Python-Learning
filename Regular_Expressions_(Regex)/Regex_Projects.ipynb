# Project 1

## Email Validator

```python
import re

# Regex pattern for validating email addresses
email_pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"

# Take email input from the user
email = input("Enter your email address: ")

# Check whether the email is valid
if re.fullmatch(email_pattern, email):

    print("\n✅ Valid Email Address")

    # -----------------------------
    # Extract Username
    # -----------------------------
    username = email.split("@")[0]

    # -----------------------------
    # Extract Domain
    # -----------------------------
    domain = email.split("@")[1]

    # -----------------------------
    # Extract Domain Name
    # -----------------------------
    domain_name = domain.split(".")[0]

    # -----------------------------
    # Extract Extension
    # -----------------------------
    extension = domain.split(".")[-1]

    print(f"Username      : {username}")
    print(f"Domain        : {domain}")
    print(f"Domain Name   : {domain_name}")
    print(f"Extension     : {extension}")

    # -----------------------------
    # Detect Email Provider
    # -----------------------------
    if domain_name.lower() == "gmail":
        provider = "Google Gmail"

    elif domain_name.lower() == "yahoo":
        provider = "Yahoo Mail"

    elif domain_name.lower() == "outlook":
        provider = "Microsoft Outlook"

    elif domain_name.lower() == "hotmail":
        provider = "Microsoft Hotmail"

    elif domain_name.lower() == "icloud":
        provider = "Apple iCloud"

    else:
        provider = "Other Provider"

    print(f"Provider      : {provider}")

else:
    print("\n❌ Invalid Email Address")
```

# Project 2

## Password Strength Checker

```python
import re

password = input("Enter Password: ")

# Initialize score
score = 0

if len(password) >= 8:
    score += 1
else:
    print("❌ Password should contain at least 8 characters.")

# Check uppercase letter
if re.search(r"[A-Z]", password):
    score += 1
else:
    print("❌ Add at least one uppercase letter.")

# Check lowercase letter
if re.search(r"[a-z]", password):
    score += 1
else:
    print("❌ Add at least one lowercase letter.")

# Check digit
if re.search(r"\d", password):
    score += 1
else:
    print("❌ Add at least one digit.")

# Check special character
if re.search(r"[!@#$%^&*(),.?\":{}|<>]", password):
    score += 1
else:
    print("❌ Add at least one special character.")

print("\n" + "-" * 40)

# Display password strength
if score == 5:
    print("✅ Password Strength : Very Strong")

elif score == 4:
    print("🟢 Password Strength : Strong")

elif score == 3:
    print("🟡 Password Strength : Medium")

elif score == 2:
    print("🟠 Password Strength : Weak")

else:
    print("🔴 Password Strength : Very Weak")

print("-" * 40)

# Display password statistics
print("\nPassword Statistics")
print(f"Length            : {len(password)}")
print(f"Uppercase Letters : {len(re.findall(r'[A-Z]', password))}")
print(f"Lowercase Letters : {len(re.findall(r'[a-z]', password))}")
print(f"Digits            : {len(re.findall(r'\\d', password))}")
print(f"Special Characters: {len(re.findall(r'[!@#$%^&*(),.?\":{}|<>]', password))}")
```

# Project 3

## Phone Number Validator

```python
import re

phone = input("Enter Phone Number: ").strip()

# Remove spaces and hyphens (if any)
phone = phone.replace(" ", "").replace("-", "")

# Regex Patterns
indian_pattern = r"^(?:\+91|91)?[6-9]\d{9}$"
international_pattern = r"^\+\d{10,15}$"

print("\n" + "-" * 45)

# Validate Phone Number
if re.fullmatch(indian_pattern, phone):

    print("✅ Valid Indian Phone Number")

    # Extract country code
    if phone.startswith("+91"):
        country_code = "+91"
        mobile = phone[3:]

    elif phone.startswith("91") and len(phone) == 12:
        country_code = "91"
        mobile = phone[2:]

    else:
        country_code = "Not Provided"
        mobile = phone

    print(f"Country Code : {country_code}")
    print(f"Mobile Number: {mobile}")

    # Detect Telecom Operator (Demo)
    first_digit = mobile[0]

    if first_digit == "9":
        operator = "Jio / Airtel / Vi"

    elif first_digit == "8":
        operator = "Airtel / Vi"

    elif first_digit == "7":
        operator = "Jio"

    else:
        operator = "BSNL / Other"

    print(f"Possible Operator : {operator}")

elif re.fullmatch(international_pattern, phone):

    print("✅ Valid International Phone Number")

else:

    print("❌ Invalid Phone Number")

print("-" * 45)

# Phone Number Statistics
digits = re.findall(r"\d", phone)

print("\nPhone Number Statistics")
print(f"Total Digits : {len(digits)}")
print(f"Starts With  : {phone[0] if phone else 'N/A'}")
```

## Resume Information Extractor

```python
import re

# Sample Resume Text
resume = """
Name: Deepika Gautam

Email: deepika.gautam123@gmail.com
Alternate Email: deepika.work@yahoo.com

Phone: +91-9876543210

Skills:
Python, SQL, Excel, Power BI, Machine Learning, Data Analysis, Git, GitHub

Education:
B.Tech in Computer Science

LinkedIn:
https://www.linkedin.com/in/deepikagautam

GitHub:
https://github.com/deepikagautam
"""

print("=" * 50)
print("Resume Information Extractor")
print("=" * 50)

# -----------------------------
# Extract Emails
# -----------------------------
emails = re.findall(
    r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}",
    resume
)

# -----------------------------
# Extract Phone Numbers
# -----------------------------
phones = re.findall(
    r"(?:\+91[- ]?)?[6-9]\d{9}",
    resume
)

# -----------------------------
# Extract LinkedIn URL
# -----------------------------
linkedin = re.findall(
    r"https://www\.linkedin\.com/in/\S+",
    resume
)

# -----------------------------
# Extract GitHub URL
# -----------------------------
github = re.findall(
    r"https://github\.com/\S+",
    resume
)

# -----------------------------
# Extract Skills
# -----------------------------
skill_list = [
    "Python",
    "SQL",
    "Excel",
    "Power BI",
    "Machine Learning",
    "Data Analysis",
    "Git",
    "GitHub",
    "Java",
    "C++",
    "Pandas",
    "NumPy"
]

skills_found = []

for skill in skill_list:
    if re.search(re.escape(skill), resume, re.IGNORECASE):
        skills_found.append(skill)

# -----------------------------
# Display Results
# -----------------------------
print("\nEmails")
for email in emails:
    print("-", email)

print("\nPhone Numbers")
for phone in phones:
    print("-", phone)

print("\nLinkedIn")
for link in linkedin:
    print("-", link)

print("\nGitHub")
for link in github:
    print("-", link)

print("\nSkills")
for skill in skills_found:
    print("-", skill)

# -----------------------------
# Summary
# -----------------------------
print("\n" + "=" * 50)
print("Summary")
print("=" * 50)

print("Total Emails       :", len(emails))
print("Total Phone Numbers:", len(phones))
print("Total Skills       :", len(skills_found))
print("LinkedIn Found     :", "Yes" if linkedin else "No")
print("GitHub Found       :", "Yes" if github else "No")
```

## Log File Analyzer

```python
import re

# Sample Log File
logs = """
2026-07-21 10:15:23 INFO User login successful
2026-07-21 10:18:45 ERROR Invalid Password
2026-07-21 10:20:10 WARNING Low Disk Space
2026-07-21 10:25:05 INFO File Uploaded
2026-07-21 10:30:18 ERROR Database Connection Failed
2026-07-21 10:32:50 INFO User Logout
2026-07-21 10:35:22 WARNING High Memory Usage
2026-07-21 10:40:10 ERROR Server Timeout
"""

print("=" * 60)
print("Log File Analyzer")
print("=" * 60)

# ------------------------------------
# Extract Dates
# ------------------------------------
dates = re.findall(r"\d{4}-\d{2}-\d{2}", logs)

# ------------------------------------
# Extract Time
# ------------------------------------
times = re.findall(r"\d{2}:\d{2}:\d{2}", logs)

# ------------------------------------
# Count Log Levels
# ------------------------------------
info_logs = re.findall(r"\bINFO\b", logs)
warning_logs = re.findall(r"\bWARNING\b", logs)
error_logs = re.findall(r"\bERROR\b", logs)

# ------------------------------------
# Extract Error Messages
# ------------------------------------
error_messages = re.findall(r"ERROR\s+(.*)", logs)

# ------------------------------------
# Extract Warning Messages
# ------------------------------------
warning_messages = re.findall(r"WARNING\s+(.*)", logs)

# ------------------------------------
# Extract Info Messages
# ------------------------------------
info_messages = re.findall(r"INFO\s+(.*)", logs)

# ------------------------------------
# Display Summary
# ------------------------------------
print("\nSummary")
print("-" * 60)

print("Total Log Entries :", len(logs.strip().split("\n")))
print("INFO Logs         :", len(info_logs))
print("WARNING Logs      :", len(warning_logs))
print("ERROR Logs        :", len(error_logs))

# ------------------------------------
# Display Error Messages
# ------------------------------------
print("\nError Messages")
print("-" * 60)

for message in error_messages:
    print("•", message)

# ------------------------------------
# Display Warning Messages
# ------------------------------------
print("\nWarning Messages")
print("-" * 60)

for message in warning_messages:
    print("•", message)

# ------------------------------------
# Display Info Messages
# ------------------------------------
print("\nInfo Messages")
print("-" * 60)

for message in info_messages:
    print("•", message)

# ------------------------------------
# Unique Dates
# ------------------------------------
print("\nDates Found")
print("-" * 60)

for date in sorted(set(dates)):
    print(date)

# ------------------------------------
# Unique Times
# ------------------------------------
print("\nTimes Found")
print("-" * 60)

for time in times:
    print(time)

# ------------------------------------
# Error Percentage
# ------------------------------------
total_logs = len(logs.strip().split("\n"))

error_percentage = (len(error_logs) / total_logs) * 100

print("\n" + "=" * 60)
print(f"Error Percentage : {error_percentage:.2f}%")
print("=" * 60)
```
