### Question 5

Print all matches with their start and end positions using `finditer()`.

````python
import re

text = "Python Java Python C++ Python"

matches = re.finditer(r"Python", text)

for match in matches:
    print(match.group(), match.start(), match.end())
````

### Question 6

Print all numbers and their positions using `finditer()`.

````python
import re

text = "Order101 Price450 Qty20"

matches = re.finditer(r"\d+", text)

for match in matches:
    print(match.group(), match.span())
````

# Assignment 7

## Topic: Groups

### Question 1

Extract the first name and last name separately using groups.

````python
import re

text = "Deepika Gautam"

result = re.search(r"(\w+)\s(\w+)", text)

if result:
    print("First Name :", result.group(1))
    print("Last Name  :", result.group(2))
````

### Question 2

Extract the day, month and year separately from the given date.

````python
import re

text = "25-12-2026"

result = re.search(r"(\d{2})-(\d{2})-(\d{4})", text)

if result:
    print("Day   :", result.group(1))
    print("Month :", result.group(2))
    print("Year  :", result.group(3))
````

### Question 3

Extract the username and domain from the given email address.

````python
import re

text = "deepika@gmail.com"

result = re.search(r"(\w+)@(\w+\.\w+)", text)

if result:
    print("Username :", result.group(1))
    print("Domain   :", result.group(2))
````

### Question 4

Extract the country code and mobile number separately.

````python
import re

text = "+91-9876543210"

result = re.search(r"\+(\d+)-(\d{10})", text)

if result:
    print("Country Code :", result.group(1))
    print("Mobile No.   :", result.group(2))
````

### Question 5

Extract the product code and quantity separately.

````python
import re

text = "PRD1025 Qty:15"

result = re.search(r"(PRD\d+)\sQty:(\d+)", text)

if result:
    print("Product Code :", result.group(1))
    print("Quantity     :", result.group(2))
````

### Question 6

Extract the protocol and website name from the given URL.

````python
import re

text = "https://www.python.org"

result = re.search(r"(https?)://(www\.\S+)", text)

if result:
    print("Protocol :", result.group(1))
    print("Website  :", result.group(2))
````

### Question 7

Extract the city and PIN code separately.

````python
import re

text = "Bhopal - 462001"

result = re.search(r"([A-Za-z]+)\s-\s(\d{6})", text)

if result:
    print("City :", result.group(1))
    print("PIN  :", result.group(2))
````

### Question 8

Extract the file name and extension separately.

````python
import re

text = "report.pdf"

result = re.search(r"(\w+)\.(\w+)", text)

if result:
    print("File Name :", result.group(1))
    print("Extension :", result.group(2))
````

### Question 9
