# 🚀 Regular Expressions (Regex) in Python

Regular Expressions (Regex) are patterns used to search, match, extract, validate, and replace text.

Regex is widely used in:

- Data Cleaning
- Data Validation
- Web Scraping
- NLP (Natural Language Processing)
- Log File Analysis
- Automation

## Why Regex?

Without Regex, searching or validating text becomes lengthy and repetitive.

Regex allows us to:

- Search text
- Find patterns
- Validate input
- Replace text
- Extract useful information

## Importing the `re` Module

Python provides the built-in `re` module for working with Regular Expressions.

````python
import re
````

## Your First Regex

The `search()` function looks for the first occurrence of a pattern anywhere in the string.

### Syntax

```python
re.search(pattern, string)
```

````python
import re

text = "I am learning Python."

result = re.search("Python", text)

print(result)
````

````python
import re

text = "I am learning Python."

result = re.search("Java", text)

print(result)
````

## Understanding the Output

If the pattern is found:

```python
<re.Match object ...>
```

If the pattern is not found:

```python
None
```

A `Match` object contains information about the matched text.

## Getting the Matched Text

Use the `group()` method to retrieve the matched value.

````python
import re

text = "Python is easy."

result = re.search("Python", text)

print(result.group())
````

## Getting Match Position

Useful methods:

- `start()` → Starting index
- `end()` → Ending index (exclusive)
- `span()` → Returns `(start, end)`

````python
import re

text = "Learning Python is fun."

result = re.search("Python", text)

print("Start :", result.start())
print("End   :", result.end())
print("Span  :", result.span())
````

## Best Practices

- Always import the `re` module once.
- Check if a match exists before using `group()`.
- Use meaningful variable names like `text`, `pattern`, and `result`.

````python
import re

text = "Machine Learning"

result = re.search("Python", text)

if result:
    print(result.group())
else:
    print("Pattern not found")
````

## Common Mistakes

❌ Calling `group()` when no match exists.

```python
result = re.search("Python", "Java")

print(result.group())
```

This raises an `AttributeError` because `result` is `None`.

Always check:

```python
if result:
    print(result.group())
```

# `re.match()`

The `match()` function checks whether the pattern matches **only at the beginning of the string**.

If the pattern is found at index **0**, it returns a Match object.

Otherwise, it returns **None**.

### Syntax

```python
re.match(pattern, string)
```

````python
import re

text = "Python is easy"

result = re.match("Python", text)

print(result)
````

````python
import re

text = "I love Python"

result = re.match("Python", text)

print(result)
````

## Why?

`match()` only checks from the **start of the string**.

Example:

```
Python is easy
^^^^^^
Starts here ✅
```

```
I love Python
       ^^^^^^
Starts later ❌
```

````python
import re

text = "Python Programming"

result = re.match("Python", text)

if result:
    print("Found :", result.group())
    print("Start :", result.start())
    print("End   :", result.end())
    print("Span  :", result.span())
else:
    print("Not Found")
````

# `search()` vs `match()`

| Feature | search() | match() |
|----------|----------|----------|
| Searches entire string | ✅ Yes | ❌ No |
| Checks only beginning | ❌ No | ✅ Yes |
| Most commonly used | ✅ Yes | ❌ Less |

````python
import re

text = "Learning Python is fun"

print("search()")
print(re.search("Python", text))

print()

print("match()")
print(re.match("Python", text))
````

````python
import re

text = "Python Programming"

print(re.search("Python", text))
print(re.match("Python", text))
````

## When to Use `match()`

Use `match()` when you want to check whether a string **starts with** a specific pattern.

Examples:

- Check if a filename starts with `"IMG_"`
- Check if an employee ID starts with `"EMP"`
- Check if a product code starts with `"PRD"`

````python
import re

employee_id = "EMP1025"

result = re.match("EMP", employee_id)

if result:
    print("Valid Employee ID")
else:
    print("Invalid Employee ID")
````

## Common Mistakes

❌ Thinking `match()` searches the whole string.

```python
re.match("Python", "I am learning Python")
```

Output:

```
None
```

Because `"Python"` is **not at the beginning**.

Use `search()` instead:

```python
re.search("Python", "I am learning Python")
```

# `re.fullmatch()`

The `fullmatch()` function checks whether the **entire string** matches the given pattern.

Unlike:

- `search()` → Searches anywhere in the string.
- `match()` → Checks only the beginning.
- `fullmatch()` → Checks the complete string.

### Syntax

```python
re.fullmatch(pattern, string)
```

````python
import re

text = "Python"

result = re.fullmatch("Python", text)

print(result)
````

````python
import re

text = "Python Programming"

result = re.fullmatch("Python", text)

print(result)
````

## Why?

`fullmatch()` compares the **complete string**.

```
Python
^^^^^^
Entire string matches ✅
```

```
Python Programming
^^^^^^^^^^^^^^^^^^
Only first word matches ❌
Entire string does not match.
```

````python
import re

text = "Python"

result = re.fullmatch("Python", text)

if result:
    print("Matched")
else:
    print("Not Matched")
````

# Comparison

| Function | Checks |
|----------|---------|
| `search()` | Anywhere in the string |
| `match()` | Beginning of the string |
| `fullmatch()` | Complete string |

````python
import re

text = "Learning Python"

print("search()")
print(re.search("Python", text))

print()

print("match()")
print(re.match("Python", text))

print()

print("fullmatch()")
print(re.fullmatch("Python", text))
````

````python
import re

text = "Python"

print(re.search("Python", text))
print(re.match("Python", text))
print(re.fullmatch("Python", text))
````

## Real-World Use Cases

`fullmatch()` is commonly used for validation tasks:

- Email Validation
- Mobile Number Validation
- Password Validation
- PIN Code Validation
- OTP Validation
- Username Validation

It ensures that the **entire input** follows the required format.

````python
import re

otp = "123456"

result = re.fullmatch(r"\d{6}", otp)

if result:
    print("Valid OTP")
else:
    print("Invalid OTP")
````

## Common Mistakes

❌ Using `search()` for input validation.

```python
re.search("Python", "I know Python")
```

This returns a match even though the **entire string** is not `"Python"`.

For validation, prefer:

```python
re.fullmatch("Python", "Python")
```

# `re.findall()`

The `findall()` function returns **all matches** of a pattern in a string.

Unlike:

- `search()` → First match only
- `match()` → Beginning only
- `fullmatch()` → Entire string
- `findall()` → **Every match**

### Syntax

```python
re.findall(pattern, string)
```

````python
import re

text = "Python Java Python C++ Python"

result = re.findall("Python", text)

print(result)
````

## Return Type

Unlike `search()` and `match()`, `findall()` returns a **list**.

If no match is found, it returns an **empty list (`[]`)**, not `None`.

````python
import re

text = "Python Java"

result = re.findall("C++", text)

print(result)
````

````python
import re

text = "Apple Mango Apple Orange Apple"

fruits = re.findall("Apple", text)

print(fruits)
print(type(fruits))
````

## Counting Matches

Since `findall()` returns a list, we can use `len()` to count matches.

````python
import re

text = "Python Java Python C++ Python"

matches = re.findall("Python", text)

print("Occurrences :", len(matches))
````

````python
import re

text = "cat dog cat bird cat"

animals = re.findall("cat", text)

for animal in animals:
    print(animal)
````

## Real-World Example

Extract all email IDs from a text.

> We are using a simple pattern for now. Advanced email validation will be covered later when we learn Character Classes and Quantifiers.

````python
import re

text = """
deepika@gmail.com
info@yahoo.com
admin@gmail.com
"""

emails = re.findall(r"\S+@\S+", text)

print(emails)
````

result = re.findall("Python", text)

`findall()` always returns a **list**.

Even if only one match is found:

If no match is found:

```

# `re.finditer()`

The `finditer()` function returns an **iterator** containing all matches of a pattern.

Unlike `findall()`, each match is returned as a **Match object**, allowing access to:

- Matched text
- Start position
- End position
- Span
```
Syntax:

```python
re.finditer(pattern, string)
```

````python
text = "Python Java Python C++ Python"

matches = re.finditer("Python", text)

print(matches)
````

## Why an Iterator?

`finditer()` returns an iterator, so we usually use it inside a `for` loop to access each match.

````python
text = "Python Java Python C++ Python"

matches = re.finditer("Python", text)

for match in matches:
    print(match)
````

````python
text = "Python Java Python C++ Python"

matches = re.finditer("Python", text)

for match in matches:
    print(match.group())
````

## Getting Match Positions

Each Match object provides:

- `group()`
- `start()`
- `end()`
- `span()`

````python
text = "Python Java Python C++ Python"

matches = re.finditer("Python", text)

for match in matches:
    print("Match :", match.group())
    print("Start :", match.start())
    print("End   :", match.end())
    print("Span  :", match.span())
    print("-" * 30)
````

````python
text = "cat dog cat bird cat"

for match in re.finditer("cat", text):
    print(
        f"Found '{match.group()}' at index {match.start()}"
    )
````

## `findall()` vs `finditer()`

| Feature | `findall()` | `finditer()` |
|----------|-------------|--------------|
| Returns | List | Iterator |
| Matched Text | ✅ | ✅ |
| Start Position | ❌ | ✅ |
| End Position | ❌ | ✅ |
| Span | ❌ | ✅ |
| Memory Efficient | ❌ | ✅ |

````python
text = "Python Java Python"

print("findall()")
print(re.findall("Python", text))

print()

print("finditer()")

for match in re.finditer("Python", text):
    print(match.group(), match.span())
````

## When to Use `finditer()`

Use `finditer()` when you need:

- Match text
- Match position
- Large text processing
- Log file analysis
- Text parsing

matches = re.finditer("Python", text)

```

Output:

```python
<callable_iterator object ...>
```

Instead, iterate over it:

```python
for match in matches:

# Character Classes `[ ]`

Character Classes allow us to match **one character** from a given set or range.

### Syntax

```python
[characters]
```

It matches **exactly one character** from the specified set.

## Basic Example

Pattern:

```python
[abc]
```

Matches any **one** character:

- a
- b
- c

It does **not** match "ab" or "abc" together.

````python
text = "apple ball cat dog"

result = re.findall("[abc]", text)

print(result)
````

## Character Range

Instead of writing every character, we can define a range.

Examples:

- `[a-z]` → Lowercase letters
- `[A-Z]` → Uppercase letters
- `[0-9]` → Digits

````python
text = "Python123"

result = re.findall("[a-z]", text)

print(result)
````

````python
text = "Python123"

result = re.findall("[A-Z]", text)

print(result)
````

````python
text = "Python123"

result = re.findall("[0-9]", text)

print(result)
````

## Multiple Ranges

You can combine multiple ranges.

Example:

```python
[A-Za-z]
```

Matches:

- Uppercase letters
- Lowercase letters

````python
text = "Python123AI"

result = re.findall("[A-Za-z]", text)

print(result)
````

````python
text = "Room101"

result = re.findall("[A-Za-z0-9]", text)

print(result)
````

## Negated Character Class

Use `^` as the **first character inside `[]`** to match characters **not** in the set.

Example:

```python
[^0-9]
```

Matches everything **except digits**.

````python
text = "Python123"

result = re.findall("[^0-9]", text)

print(result)
````

````python
text = "Python123AI"

result = re.findall("[^A-Za-z]", text)

print(result)
````

## Common Character Classes

| Pattern | Meaning |
|----------|---------|
| `[abc]` | a, b or c |
| `[a-z]` | Lowercase letters |
| `[A-Z]` | Uppercase letters |
| `[A-Za-z]` | All alphabets |
| `[0-9]` | Digits |
| `[A-Za-z0-9]` | Letters and digits |
| `[^0-9]` | Everything except digits |
| `[^A-Za-z]` | Everything except letters |

````python
text = "Python@123#AI"

print("Letters :", re.findall("[A-Za-z]", text))
print("Digits  :", re.findall("[0-9]", text))
print("Special :", re.findall("[^A-Za-z0-9]", text))
````

## Common Mistakes

❌ Assuming `[abc]` matches the word `"abc"`.

```python
re.findall("[abc]", "abc")
```

Output:

```python
['a', 'b', 'c']
```

Character classes match **one character at a time**, not complete words.

# Special Sequences

Special Sequences are predefined shorthand patterns.

Instead of writing long patterns like:

```python
[0-9]
```

we can simply write:

```python
\d
```

They make Regex shorter, cleaner, and easier to read.

## Common Special Sequences

| Pattern | Meaning |
|----------|---------|
| `\d` | Digit (0–9) |
| `\D` | Not a Digit |
| `\w` | Letter, Digit or `_` |
| `\W` | Not a Letter, Digit or `_` |
| `\s` | Whitespace |
| `\S` | Not Whitespace |

## Why Raw Strings (`r""`)?

Regex patterns often contain backslashes (`\`).

Using a raw string prevents Python from treating them as escape characters.

✅ Recommended

```python
r"\d"
```

❌ Avoid

```python
"\d"
```

````python
text = "Python123AI"

result = re.findall(r"\d", text)

print(result)
````

````python
text = "Python123AI"

result = re.findall(r"\D", text)

print(result)
````

## `\d` vs `[0-9]`

Both produce the same result.

```python
\d
```

is simply a shorter form of

```python
[0-9]
```

````python
text = "Room101"

print(re.findall(r"\d", text))

print(re.findall(r"[0-9]", text))
````

````python
text = "Python_123"

print(re.findall(r"\w", text))
````

## What does `\w` match?

`\w` matches:

- A-Z
- a-z
- 0-9
- underscore (`_`)

Equivalent to:

```python
[A-Za-z0-9_]
```

````python
text = "Python_123@AI"

print(re.findall(r"\W", text))
````

````python
text = "Python is awesome"

print(re.findall(r"\s", text))
````

## What is Whitespace?

Whitespace includes:

- Space
- Tab (`\t`)
- New Line (`\n`)

````python
text = "Python is awesome"

print(re.findall(r"\S", text))
````

````python
text = "Python 123 @ AI"

print("Digits      :", re.findall(r"\d", text))
print("Not Digits  :", re.findall(r"\D", text))
print("Word Chars  :", re.findall(r"\w", text))
print("Non-Word    :", re.findall(r"\W", text))
print("Whitespace  :", re.findall(r"\s", text))
print("Not Space   :", re.findall(r"\S", text))
````

# Quantifiers

Quantifiers specify **how many times** a pattern should occur.

Without quantifiers, Regex matches **only one character** at a time.

Example:

```python
\d
```

Matches only **one digit**.

To match multiple digits, we use quantifiers.

## Types of Quantifiers

| Quantifier | Meaning |
|------------|---------|
| `*` | 0 or more |
| `+` | 1 or more |
| `?` | 0 or 1 |
| `{n}` | Exactly n times |
| `{n,}` | At least n times |
| `{n,m}` | Between n and m times |

## `*` (Zero or More)

Matches the previous pattern **zero or more times**.

````python
text = "aa a aaa b"

print(re.findall(r"a*", text))
````

## `+` (One or More)

Matches the previous pattern **one or more times**.

````python
text = "aa a aaa b"

print(re.findall(r"a+", text))
````

## Difference Between `*` and `+`

- `*` → Can match zero occurrences.
- `+` → Requires at least one occurrence.

````python
text = "bbb"

print("Using * :", re.findall(r"a*", text))
print("Using + :", re.findall(r"a+", text))
````

## `?` (Zero or One)

Matches the previous pattern **zero or one time**.

Useful for optional characters.

````python
text = "color colour"

print(re.findall(r"colou?r", text))
````

## `{n}` (Exactly n Times)

Matches the previous pattern exactly **n** times.

````python
text = "1 12 123 1234 12345"

print(re.findall(r"\d{3}", text))
````

## `{n,}` (At Least n Times)

Matches the previous pattern **n or more times**.

````python
text = "1 12 123 1234 12345"

print(re.findall(r"\d{3,}", text))
````

## `{n,m}` (Between n and m Times)

Matches the previous pattern between **n** and **m** times.

````python
text = "1 12 123 1234 12345"

print(re.findall(r"\d{2,4}", text))
````

## Real-World Examples

### 6-Digit OTP

```python
\d{6}
```

### 10-Digit Phone Number

```python
\d{10}
```

### Password Length (8–16 Characters)

```python
.{8,16}
```

````python
otp = "458921"

if re.fullmatch(r"\d{6}", otp):
    print("Valid OTP")
else:
    print("Invalid OTP")
````

## Quick Summary

| Pattern | Meaning |
|---------|---------|
| `*` | Zero or more |
| `+` | One or more |
| `?` | Zero or one |
| `{3}` | Exactly 3 |
| `{3,}` | 3 or more |
| `{3,5}` | Between 3 and 5 |

## Common Mistakes

❌ Thinking `\d{3}` means "three or more digits".

It actually means:

**Exactly 3 digits**

For three or more digits, use:

```python
\d{3,}
```

# Metacharacters

Metacharacters are special symbols that give **special meaning** to a regex pattern.

Unlike normal characters, they control **how matching is performed**.

## Common Metacharacters

| Symbol | Meaning |
|--------|---------|
| `.` | Any single character (except newline) |
| `^` | Start of the string |
| `$` | End of the string |
| `|` | OR operator |

## `.` (Dot)

The dot (`.`) matches **any single character** except a newline (`\n`).

Examples:

- `c.t` matches:
  - cat
  - cot
  - cut
  - c9t

But it does **not** match:

- ct
- cart

````python
text = "cat cot cut c9t ct"

print(re.findall(r"c.t", text))
````

## `^` (Start of String)

Checks whether the pattern appears **at the beginning** of the string.

````python
text = "Python Programming"

print(re.search(r"^Python", text))
````

````python
text = "I love Python"

print(re.search(r"^Python", text))
````

## `$` (End of String)

Checks whether the pattern appears **at the end** of the string.

````python
text = "I love Python"

print(re.search(r"Python$", text))
````

````python
text = "Python is easy"

print(re.search(r"Python$", text))
````

## `|` (OR Operator)

The pipe (`|`) works like the logical **OR**.

It matches **either** the left pattern **or** the right pattern.

````python
text = "cat dog rabbit"

print(re.findall(r"cat|dog", text))
````

````python
text = "Python Java C++ JavaScript"

print(re.findall(r"Python|Java", text))
````

# Combining Metacharacters

Example:

Meaning:

- Must start with `Python`
- Must end with `Python`

So only the exact string `"Python"` matches.

````python
text = "Python"

print(re.fullmatch(r"^Python$", text))
````

````python
text = "Python Programming"

print(re.fullmatch(r"^Python$", text))
````

# Groups `()`

Groups allow us to combine multiple parts of a pattern into a single unit.

They are mainly used for:

- Extracting specific parts of text
- Organizing complex patterns
- Reusing matched values

## Basic Syntax

```python
(pattern)
```

The text matched inside `()` is called a **capturing group**.

````python
text = "My phone number is 9876543210"

result = re.search(r"(\d+)", text)

print(result.group())
````

## Multiple Groups

Each pair of parentheses creates a separate group.

Groups are numbered starting from **1**.

````python
text = "Date: 21-07-2026"

result = re.search(r"(\d{2})-(\d{2})-(\d{4})", text)

print(result.group(0))
print(result.group(1))
print(result.group(2))
print(result.group(3))
````

## Understanding `group()`

| Method | Returns |
|---------|----------|
| `group(0)` | Entire match |
| `group(1)` | First group |
| `group(2)` | Second group |
| `group(3)` | Third group |

````python
text = "John, Age: 25"

result = re.search(r"([A-Za-z]+), Age: (\d+)", text)

print("Name :", result.group(1))
print("Age  :", result.group(2))
````

## `groups()`

The `groups()` method returns all captured groups as a tuple.

````python
text = "21-07-2026"

result = re.search(r"(\d{2})-(\d{2})-(\d{4})", text)

print(result.groups())
````

## Named Groups

Instead of numbers, groups can have names.

### Syntax

```python
(?P<name>pattern)
```

````python
text = "John 25"

result = re.search(
    r"(?P<name>[A-Za-z]+)\s(?P<age>\d+)",
    text
)

print(result.group("name"))
print(result.group("age"))
````

## Why Named Groups?

Named groups improve readability.

Instead of:

```python
group(2)
```

you can write:

```python
group("age")
```

````python
text = "Alice alice@example.com"

result = re.search(
    r"(?P<name>[A-Za-z]+)\s(?P<email>\S+)",
    text
)

print("Name  :", result.group("name"))
print("Email :", result.group("email"))
````

## Quick Summary

| Method | Description |
|---------|-------------|
| `group(0)` | Complete match |
| `group(1)` | First captured group |
| `groups()` | Tuple of all groups |
| `(?P<name>...)` | Named group |

## Common Mistakes

❌ Using the wrong group number.

```python
result.group(4)
```

If only three groups exist, Python raises an `IndexError`.

Always ensure the requested group exists.

# `re.split()`

The `split()` function splits a string using a regex pattern.

Unlike Python's built-in `split()`, it can split text using complex patterns.

### Syntax

```python
re.split(pattern, string)
```

````python
text = "Python,Java,C++,JavaScript"

result = re.split(r",", text)

print(result)
````

````python
text = "Python, Java; C++ | JavaScript"

result = re.split(r"[,;|]\s*", text)

print(result)
````

## Why use `re.split()`?

Normal `split()` can split using only one separator.

```python
text.split(",")
```

Regex can split using **multiple separators**.

Example:

- comma (,)
- semicolon (;)
- pipe (|)
- spaces

````python
text = "apple   banana    mango"

result = re.split(r"\s+", text)

print(result)
````

## Limiting Splits

Use `maxsplit` to limit the number of splits.

### Syntax

```python
re.split(pattern, string, maxsplit=n)
```

````python
text = "Python Java C++ JavaScript"

result = re.split(r"\s+", text, maxsplit=2)

print(result)
````

# `re.sub()`

The `sub()` function replaces matching text with new text.

### Syntax

```python
re.sub(pattern, replacement, string)
```

````python
text = "I like Java."

result = re.sub(r"Java", "Python", text)

print(result)
````

````python
text = "Python123Java456"

result = re.sub(r"\d", "*", text)

print(result)
````

## Replacing Multiple Spaces

A common data-cleaning task is removing extra spaces.

````python
text = "Python     is      awesome"

result = re.sub(r"\s+", " ", text)

print(result)
````

## Limiting Replacements

Use `count` to limit replacements.

### Syntax

```python
re.sub(pattern, replacement, string, count=n)
```

````python
text = "cat cat cat cat"

result = re.sub(r"cat", "dog", text, count=2)

print(result)
````

````python
text = "Price: 4500"

result = re.sub(r"\d", "X", text)

print(result)
````

## `split()` vs `sub()`

| Function | Purpose |
|----------|---------|
| `split()` | Break text into parts |
| `sub()` | Replace matching text |

## Real-World Examples

### `split()`

- CSV processing
- Log parsing
- Text tokenization

### `sub()`

- Remove extra spaces
- Mask phone numbers
- Hide email IDs
- Clean raw text

## Common Mistakes

❌ Using `str.replace()` when pattern matching is needed.

```python
text.replace(...)
```

works only with exact text.

Use:

```python
re.sub(...)
```

when you need regex patterns.

# `re.compile()`

The `compile()` function creates a reusable regex pattern object.

Instead of writing the same pattern multiple times, you can compile it once and reuse it.

### Syntax

```python
pattern = re.compile(regex_pattern)
```

## Why use `compile()`?

Without `compile()`:

```python
re.search(r"\d+", text)
re.findall(r"\d+", text)
re.fullmatch(r"\d+", text)
```

The same pattern is written repeatedly.

With `compile()`:

```python
pattern = re.compile(r"\d+")
```

Now the same pattern can be reused anywhere.

````python
text = "Python 123 Java 456"

pattern = re.compile(r"\d+")

print(pattern.findall(text))
````

````python
text = "Employee ID: EMP1025"

pattern = re.compile(r"EMP\d+")

result = pattern.search(text)

print(result.group())
````

## Methods Available

A compiled pattern supports the same common methods:

- `search()`
- `match()`
- `fullmatch()`
- `findall()`
- `finditer()`
- `split()`
- `sub()`

````python
text = "100 200 300"

pattern = re.compile(r"\d+")

print(pattern.findall(text))
print(pattern.search(text).group())
````

````python
emails = """
deepika@gmail.com
admin@yahoo.com
test@example.com
"""

email_pattern = re.compile(r"\S+@\S+")

print(email_pattern.findall(emails))
````

## When Should You Use `compile()`?

Use `compile()` when:

- The same pattern is used multiple times.
- Processing large amounts of text.
- Building reusable validation functions.
- Writing production-level applications.

For one-time matching, calling `re.search()` or `re.findall()` directly is perfectly fine.

## Direct Regex vs Compiled Regex

| Direct | Compiled |
|---------|----------|
| Good for one-time use | Good for repeated use |
| Shorter code | Better reusability |
| Best for simple scripts | Best for larger applications |

## Common Mistakes

❌ Compiling a pattern that is used only once.

```python
pattern = re.compile(r"Python")
pattern.search(text)
```

This works, but for a single use:

```python
re.search(r"Python", text)
```

is simpler and easier to read.

# Regex Flags

Regex Flags modify how a regex pattern behaves.

Instead of changing the pattern, flags change the **matching behavior**.

### Syntax

```python
re.function(pattern, string, flag)
```

## Common Flags

| Flag | Description |
|------|-------------|
| `re.IGNORECASE` (`re.I`) | Ignore letter case |
| `re.MULTILINE` (`re.M`) | `^` and `$` work on every line |
| `re.DOTALL` (`re.S`) | `.` matches newline (`\n`) |
| `re.VERBOSE` (`re.X`) | Write readable multi-line regex |

## `re.IGNORECASE` (`re.I`)

Makes matching case-insensitive.

````python
text = "python"

print(re.search(r"Python", text))

print(re.search(r"Python", text, re.IGNORECASE))
````

## `re.MULTILINE` (`re.M`)

Normally, `^` and `$` match only the start and end of the entire string.

With `re.MULTILINE`, they match the start and end of **each line**.

````python
text = """Python
Java
C++"""

print(re.findall(r"^Java", text))

print(re.findall(r"^Java", text, re.MULTILINE))
````

## `re.DOTALL` (`re.S`)

Normally, `.` does **not** match a newline.

With `re.DOTALL`, it matches every character, including `\n`.

````python
text = "Python\nJava"

print(re.findall(r"Python.*Java", text))

print(re.findall(r"Python.*Java", text, re.DOTALL))
````

## `re.VERBOSE` (`re.X`)

Allows writing long regex patterns in a readable format with spaces and comments.

Useful for complex validation patterns.

For beginners, simply remember its purpose. You'll rarely need it in day-to-day coding.

````python
pattern = re.compile(
    r"""
    \d{3}   # Area Code
    -
    \d{3}   # Prefix
    -
    \d{4}   # Number
    """,
    re.VERBOSE
)

print(pattern.fullmatch("123-456-7890"))
````

## Quick Summary

| Flag | Use Case |
|------|----------|
| `re.I` | Ignore uppercase/lowercase |
| `re.M` | Multi-line matching |
| `re.S` | Dot matches newline |
| `re.X` | Readable regex |

## Common Mistakes

❌ Using `re.MULTILINE` expecting `.` to match newlines.

Remember:

- `re.MULTILINE` affects only `^` and `$`.
- `re.DOTALL` affects `.`.

# Escaping Special Characters

Some characters have special meanings in Regex.

To match them literally, use a backslash (`\`).

## Common Escaped Characters

| Character | Escape |
|-----------|--------|
| `.` | `\.` |
| `+` | `\+` |
| `*` | `\*` |
| `?` | `\?` |
| `(` | `\(` |
| `)` | `\)` |
| `[` | `\[` |
| `]` | `\]` |
| `{` | `\{` |
| `}` | `\}` |
| `^` | `\^` |
| `$` | `\$` |
| `|` | `\|` |
| `\` | `\\` |

````python
text = "python.org"

print(re.findall(r"\.", text))
````

````python
text = "Price: $500"

print(re.findall(r"\$", text))
````

````python
text = "C++ Python"

print(re.findall(r"\+\+", text))
````

````python
text = "(Python)"

print(re.findall(r"\(", text))
print(re.findall(r"\)", text))
````

## When Do We Escape?

Escape special characters **only when you want to match them literally**.

Example:

- `.` → Any character
- `\.` → Actual dot

- `+` → One or more
- `\+` → Plus sign

- `$` → End of string
- `\$` → Dollar sign

````python
text = "www.google.com"

print(re.findall(r"\.", text))
````

## Common Mistakes

❌ Searching for a dot using:

```python
r"."
```

This matches **every character**.

✅ Correct:

```python
r"\."
```
