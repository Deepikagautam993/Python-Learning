# What is for Loop?

for loop is used to access each element of a sequence one by one.

Syntax:

for variable in sequence:
    statement

The loop automatically stops when the sequence ends.

```python
# Simple for loop

fruits = ["Apple", "Mango", "Orange"]

for fruit in fruits:
    print(fruit)
```

# Working of for Loop

Example:

for item in sequence:

Step 1:
Take first element

Step 2:
Execute code

Step 3:
Move to next element

Step 4:
Repeat until sequence ends

```python
# for loop with string

name = "Python"

for char in name:
    print(char)
```

```python
# for loop with list

numbers = [10,20,30]

for num in numbers:
    print(num)
```

# range() Function

range() generates a sequence of numbers.

Syntax:

range(start, stop, step)

Important:

start is included

stop is not included

```python
# range(stop)

for i in range(5):
    print(i)
```

```python
# range(start, stop)

for i in range(2,6):
    print(i)
```

```python
# range with step

for i in range(1,10,2):
    print(i)
```

```python
# Reverse range

for i in range(5,0,-1):
    print(i)
```

# Nested for Loop

A loop inside another loop is called nested loop.

Syntax:

for i in sequence:
    for j in sequence:
        statement

Used in:
- Patterns
- Matrix
- 2D data

```python
for i in range(1,4):

    for j in range(1,4):

        print(i,j)
```

```python
# Pattern example

for i in range(1,5):

    for j in range(i):

        print("*", end=" ")

    print()
```

Q6. Find maximum number without max().

Q7. Count vowels in string.

```python
# Q1 Print 1 to 10

for i in range(1,11):
    print(i)



# Q2 Reverse 10 to 1

for i in range(10,0,-1):
    print(i)



# Q3 Multiplication Table

num = int(input("Enter number: "))

for i in range(1,11):
    print(num*i)



# Q4 Sum 1 to n

n = int(input("Enter n: "))

total = 0

for i in range(1,n+1):
    total += i

print(total)



# Q5 String characters

text = input("Enter string: ")

for ch in text:
    print(ch)
```

```python
# Q6 Maximum without max()

numbers = [4,8,2,10]

maximum = numbers[0]

for num in numbers:

    if num > maximum:
        maximum = num

print(maximum)



# Q7 Count vowels

text = "python"

count = 0

for ch in text:

    if ch in "aeiou":
        count += 1

print(count)



# Q8 Sum of list

numbers = [10,20,30]

total = 0

for num in numbers:

    total += num

print(total)
```

```python
# Q9 Pattern

for i in range(1,5):

    for j in range(i):

        print("*", end="")

    print()



# Q10 Difference Notes

# for loop:
# Used when number of iterations is known
#
# while loop:
# Used when condition decides repetition
```

# Interview Notes

## for loop vs while loop

for loop:
- Sequence based
- Used when iterations are fixed


while loop:
- Condition based
- Used when iterations are unknown


## range()

range creates a sequence of numbers.
stop value is always excluded.
