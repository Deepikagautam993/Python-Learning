# Question 1

Create a lambda function to calculate the square of a number.

Test it using:

- 6
- 11

````python
square = lambda number: number ** 2

print(square(6))
print(square(11))
````

# Question 3

Create a lambda function that returns the smaller of two numbers.

````python
smaller = lambda a, b: a if a < b else b

print(smaller(45, 18))
print(smaller(12, 67))
````

# Question 4

Create a lambda function that checks whether a number is positive or negative.

````python
check = lambda number: "Positive" if number >= 0 else "Negative"

print(check(25))
print(check(-9))
````

# Question 5

Create a lambda function that calculates the area of a rectangle.

Test it using:

- Length = 14, Width = 9
- Length = 25, Width = 12

````python
area = lambda length, width: length * width

print(area(14, 9))
print(area(25, 12))
````

# Question 6

Create a lambda function that returns the largest among three numbers.

Test it using:

- 18, 42, 31

````python
largest = lambda a, b, c: max(a, b, c)

print(largest(18, 42, 31))
print(largest(95, 67, 88))
````

# Question 7

Create an Immediately Invoked Lambda Function (IIFE) that calculates the simple interest.

Formula:

Simple Interest = (P × R × T) / 100

Use:

- Principal = 5000
- Rate = 8
- Time = 2

````python
interest = (lambda p, r, t: (p * r * t) / 100)(5000, 8, 2)

print("Simple Interest :", interest)
````

# Question 8

Create a function that returns a lambda function.

The returned lambda should divide a number by a fixed value.

Example:

```python
divide_by_5 = divider(5)
```

Calling:

```python
divide_by_5(50)
```

should return:

```
10.0
```

````python
def divider(value):

    return lambda number: number / value


divide_by_5 = divider(5)
divide_by_10 = divider(10)

print(divide_by_5(50))
print(divide_by_10(80))
````

# Question 9

Sort the following list of tuples based on **age** in ascending order using a lambda function.

```python
employees = [

    ("Rohan", 29),
    ("Sneha", 24),
    ("Amit", 31),
    ("Pooja", 27)

]
```

````python
employees = [

    ("Rohan", 29),
    ("Sneha", 24),
    ("Amit", 31),
    ("Pooja", 27)

]

result = sorted(
    employees,
    key=lambda employee: employee[1]
)

print(result)
````

# Question 10

Sort the following list of dictionaries based on **price** in descending order using `list.sort()`.

```python
products = [

    {"name": "Tablet", "price": 28000},
    {"name": "Keyboard", "price": 3500},
    {"name": "Camera", "price": 54000},
    {"name": "Speaker", "price": 6500}

]
```

````python
products = [

    {"name": "Tablet", "price": 28000},
    {"name": "Keyboard", "price": 3500},
    {"name": "Camera", "price": 54000},
    {"name": "Speaker", "price": 6500}

]

products.sort(
    key=lambda product: product["price"],
    reverse=True
)

print(products)
````

# Question 11

# Question 12

Find the product with the lowest quantity using `min()` and a lambda function.

```python
inventory = [
```

# Question 14

Sort the following list of dictionaries first by **department** and then by **salary** using a lambda function.

```python
employees = [

    {"name": "Aman", "department": "HR", "salary": 45000},
    {"name": "Riya", "department": "IT", "salary": 65000},
    {"name": "Karan", "department": "HR", "salary": 52000},
```

````python
employees = [

    {"name": "Aman", "department": "HR", "salary": 45000},
    {"name": "Riya", "department": "IT", "salary": 65000},
    {"name": "Karan", "department": "HR", "salary": 52000},
    {"name": "Neha", "department": "IT", "salary": 60000}

]

employees.sort(
    key=lambda employee: (employee["department"], employee["salary"])
)

print(employees)
````
