# 🚀 Lambda Functions in Python

## Overview

A **Lambda Function** is a small anonymous function written using the `lambda` keyword.

It is mainly used for short operations where writing a complete `def` function is unnecessary.

> A lambda function contains **only one expression**, and its result is returned automatically.

## Syntax

```python
lambda parameters: expression
```

### Syntax Breakdown

- `lambda` → Keyword to create a lambda function.
- `parameters` → Input values.
- `expression` → Single expression evaluated and returned.
```

````python
# Normal Function

def multiply(a, b):

    return a * b


print(multiply(12, 8))
````

````python
# Lambda Function

multiply = lambda a, b: a * b

print(multiply(12, 8))
````

## Why Use Lambda?

Use lambda when:

- A function is small.
- It is used only once.
- Short and readable code is preferred.
- A function is required as an argument to another function.

````python
square = lambda number: number ** 2

print(square(9))
print(square(15))
````

````python
area = lambda length, width: length * width

print(area(18, 12))
````

## Important Notes

- Lambda functions have no explicit name.
- They automatically return the result.
- Only one expression is allowed.
- Complex logic should be written using `def`.

## Lambda with Conditional Expression

A conditional expression can be written directly inside a lambda function.

### Syntax

```python
lambda parameters: value_if_true if condition else value_if_false
```

````python
status = lambda marks: "Pass" if marks >= 33 else "Fail"

print(status(78))
print(status(25))
````

````python
discount = lambda amount: amount * 0.10 if amount >= 5000 else amount * 0.05

print(discount(7000))
print(discount(2500))
````

## Immediately Invoked Lambda Function (IIFE)

A lambda function can be created and executed immediately without storing it in a variable.

### Syntax

```python
(lambda parameters: expression)(arguments)
```

````python
result = (lambda x, y: x ** y)(2, 5)

print(result)
````

````python
print((lambda text: text[::-1])("Python"))
````

## Lambda Returning from a Function

A function can return a lambda function.

This is useful when creating customized functions dynamically.

````python
def multiplier(value):

    return lambda number: number * value


double = multiplier(2)
triple = multiplier(3)
quadruple = multiplier(4)

print(double(15))
print(triple(15))
print(quadruple(15))
````

````python
def greeting(message):

    return lambda name: f"{message}, {name}!"


welcome = greeting("Welcome")

print(welcome("Deepika"))
print(welcome("Rahul"))
````

## Lambda vs Normal Function

| Normal Function | Lambda Function |
|-----------------|-----------------|
| Uses `def` keyword | Uses `lambda` keyword |
| Can contain multiple statements | Only one expression |
| Explicit `return` statement | Returns automatically |
| Better for complex logic | Better for short operations |
| Reusable | Mostly used for short-lived functions |

## Best Practices

- Keep lambda expressions short.
- Prefer `def` for complex logic.
- Use meaningful variable names.
- Avoid nested or overly complex lambda expressions.

````python
# Good Practice

percentage = lambda total, obtained: (obtained / total) * 100

print(f"{percentage(500, 438):.2f}%")
````

## Key Takeaways

- Lambda functions are anonymous functions.
- They contain only one expression.
- They automatically return the result.
- They are commonly used with functions like `map()`, `filter()`, `reduce()`, and `sorted()`.

## Lambda with `sorted()`

The `sorted()` function returns a new sorted list.

The `key` parameter allows us to define custom sorting logic using a lambda function.

````python
numbers = [42, 15, 73, 8, 29]

ascending = sorted(numbers, key=lambda number: number)

descending = sorted(numbers, key=lambda number: -number)

print("Ascending :", ascending)
print("Descending :", descending)
````

````python
students = [

    ("Riya", 82),
    ("Aman", 95),
    ("Karan", 76),
    ("Neha", 88)

]

result = sorted(students, key=lambda student: student[1])

print(result)
````

````python
employees = [

    {"name": "Ankit", "salary": 52000},
    {"name": "Priya", "salary": 68000},
    {"name": "Rahul", "salary": 47000}

]

sorted_salary = sorted(
    employees,
    key=lambda employee: employee["salary"]
)

print(sorted_salary)
````

## Lambda with `list.sort()`

`list.sort()` sorts the original list.

Unlike `sorted()`, it does not create a new list.

````python
cities = [

    ("Delhi", 22000000),
    ("Indore", 3500000),
    ("Mumbai", 20500000),
    ("Bhopal", 2400000)

]

cities.sort(key=lambda city: city[1])

print(cities)
````

## Lambda with `max()`

The `key` parameter is used to decide how the maximum value should be selected.

````python
products = [

    {"name": "Laptop", "price": 72000},
    {"name": "Phone", "price": 28000},
    {"name": "Monitor", "price": 18000}

]

expensive = max(
    products,
    key=lambda product: product["price"]
)

print(expensive)
````

## Lambda with `min()`

The `key` parameter is also supported by `min()`.

````python
books = [

    ("Python", 450),
    ("C++", 650),
    ("Java", 520),
    ("SQL", 350)

]

cheapest = min(
    books,
    key=lambda book: book[1]
)

print(cheapest)
````

````python
players = [

    {"name": "Virat", "score": 102},
    {"name": "Rohit", "score": 85},
    {"name": "Gill", "score": 128}

]

highest = max(
    players,
    key=lambda player: player["score"]
)

print(highest)
````

## Common Use Cases

Lambda functions are commonly used with:

- `sorted()`
- `list.sort()`
- `max()`
- `min()`
- `map()`
- `filter()`
- `reduce()`

## Important Notes

- `sorted()` → Returns a new sorted list.
- `list.sort()` → Modifies the original list.
- `key=lambda ...` defines the sorting or comparison rule.
- Lambda is commonly used when the comparison logic is short.
