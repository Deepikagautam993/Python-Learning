## ❓What is Comprehension?

A **Comprehension** is a short and Pythonic way to create a new collection (List, Dictionary, or Set) from an existing iterable.

Instead of writing multiple lines using loops, Python allows us to write the same logic in a single line using comprehensions.

### Example

Traditional Way

```python
square = []

for i in range(1,6):
    square.append(i**2)

print(square)
```

Comprehension

```python
square = [i**2 for i in range(1,6)]

print(square)
```

Both produce the same output, but the second approach is shorter and easier to read.

---

4. Generator Expression *(Python does not have Tuple Comprehension.)*

````python
# List Comprehension

square = [i ** 2 for i in range(1, 6)]

print(square)
````

# 📌 List Comprehension with if Condition

We can filter elements using an `if` condition.

### Syntax

```python
new_list = [expression for item in iterable if condition]
```

````python
# Traditional Method

even_numbers = []

for i in range(1, 11):
    if i % 2 == 0:
        even_numbers.append(i)

print(even_numbers)
````

````python
# List Comprehension

even_numbers = [i for i in range(1, 11) if i % 2 == 0]

print(even_numbers)
````

````python
# Filter Positive Numbers

numbers = [-10, 20, -5, 40, 0, 60]

positive_numbers = [num for num in numbers if num > 0]

print(positive_numbers)
````

````python
# Filter Long Names

names = ["Ram", "Deepika", "Amit", "Shivani", "AI"]

long_names = [name for name in names if len(name) > 4]

print(long_names)
````

# List Comprehension with if-else

Sometimes, we don't want to filter elements. Instead, we want to create a new value for every element based on a condition.

### Syntax

```python
new_list = [value_if_true if condition else value_if_false for item in iterable]
```

Unlike `if` filtering, every element is included in the final list.

````python
# Traditional Method

result = []

for i in range(1, 11):
    if i % 2 == 0:
        result.append("Even")
    else:
        result.append("Odd")

print(result)
````

````python
# List Comprehension

result = ["Even" if i % 2 == 0 else "Odd" for i in range(1, 11)]

print(result)
````

````python
# Positive and Negative Numbers

numbers = [-5, 8, -10, 25, 0]

result = ["Positive" if num > 0 else "Non-Positive" for num in numbers]

print(result)
````

````python
# Pass and Fail

marks = [35, 82, 45, 18, 91, 39]

status = ["Pass" if mark >= 40 else "Fail" for mark in marks]

print(status)
````

# Multiple for in List Comprehension

Multiple `for` loops in List Comprehension work like nested loops.

### Syntax

```python
[
    expression
    for item1 in iterable1
    for item2 in iterable2
]
```

````python
# Traditional Nested Loop

colors = ["Red", "Blue"]
fruits = ["Apple", "Mango"]

result = []

for color in colors:
    for fruit in fruits:
        result.append(color + " " + fruit)

print(result)
````

````python
# List Comprehension

colors = ["Red", "Blue"]
fruits = ["Apple", "Mango"]

result = [
    color + " " + fruit
    for color in colors
    for fruit in fruits
]

print(result)
````

````python
# Create Pairs

letters = ["A", "B"]
numbers = [1, 2, 3]

pairs = [
    (letter, number)
    for letter in letters
    for number in numbers
]

print(pairs)
````

````python
# Add Elements from Two Lists

list1 = [1, 2]
list2 = [10, 20]

result = [
    x + y
    for x in list1
    for y in list2
]

print(result)
````

# Important Points

- List Comprehension creates a new list.
- It is shorter and more readable than a traditional loop.
- Use `if` when you want to filter elements.
- Use `if-else` when every element should produce an output.
- Multiple `for` loops work like nested loops.
- List Comprehension should be kept simple for better readability.

# Dictionary Comprehension

## What is Dictionary Comprehension?

Dictionary Comprehension is a concise and Pythonic way to create dictionaries using a single line of code.

It is an alternative to writing a traditional `for` loop for creating dictionaries.

---

## Syntax

```python
new_dict = {key: value for item in iterable}
```

### Components

- **key** → Key of the dictionary
- **value** → Value associated with the key
- **item** → Current element
- **iterable** → List, Tuple, Set, Range, etc.
```

````python
# Traditional Method

square = {}

for i in range(1, 6):
    square[i] = i ** 2

print(square)
````

````python
# Dictionary Comprehension

square = {i: i ** 2 for i in range(1, 6)}

print(square)
````

# Example 1 : Store Cube of Numbers

````python
cube = {i: i ** 3 for i in range(1, 6)}

print(cube)
````

# Example 2 : Store Length of Names

````python
names = ["Deepika", "Rahul", "Amit", "Shivani"]

name_length = {name: len(name) for name in names}

print(name_length)
````

# Example 3 : Convert Numbers into Strings

````python
numbers = [1, 2, 3, 4, 5]

number_string = {num: str(num) for num in numbers}

print(number_string)
````

# Dictionary Comprehension with if Condition

We can filter dictionary elements using an `if` condition.

### Syntax

```python
{
    key: value
    for item in iterable
    if condition
}
```

````python
# Traditional Method

even_square = {}

for i in range(1, 11):
    if i % 2 == 0:
        even_square[i] = i ** 2

print(even_square)
````

````python
# Dictionary Comprehension

even_square = {
    i: i ** 2
    for i in range(1, 11)
    if i % 2 == 0
}

print(even_square)
````

# Example 4 : Store Only Long Names

````python
names = ["Ram", "Deepika", "AI", "Shivani", "Amit"]

long_names = {
    name: len(name)
    for name in names
    if len(name) > 4
}

print(long_names)
````

# Example 5 : Store Positive Numbers

````python
numbers = [-5, 10, -2, 15, 0, 20]

positive = {
    num: num ** 2
    for num in numbers
    if num > 0
}

print(positive)
````

# Important Points

- Dictionary Comprehension always creates a dictionary.
- Keys must be unique.
- Duplicate keys overwrite previous values.
- It can also be used with `if` conditions.
- The syntax is very similar to List Comprehension.

# Set Comprehension

## What is Set Comprehension?

Set Comprehension is used to create a new set in a single line.

It automatically removes duplicate values because sets only store unique elements.

---

## Syntax

```python
new_set = {expression for item in iterable}
```

### Components

- **expression** → Value to store in the set.
- **item** → Current element.
- **iterable** → List, Tuple, Set, Range, etc.
```

````python
# Traditional Method

square = set()

for i in range(1, 6):
    square.add(i ** 2)

print(square)
````

````python
# Set Comprehension

square = {i ** 2 for i in range(1, 6)}

print(square)
````

# Example 1 : Create a Set

````python
numbers = {i for i in range(1, 11)}

print(numbers)
````

# Example 2 : Remove Duplicate Values

````python
numbers = [10, 20, 30, 10, 20, 40, 50]

unique_numbers = {num for num in numbers}

print(unique_numbers)
````

# Example 3 : Square of Even Numbers

````python
square = {
    i ** 2
    for i in range(1, 11)
    if i % 2 == 0
}

print(square)
````

# Example 4 : Convert Names into Uppercase

````python
names = ["deepika", "rahul", "amit", "rahul"]

upper_names = {
    name.upper()
    for name in names
}

print(upper_names)
````

# Important Points

- Set stores only unique values.
- Duplicate values are automatically removed.
- Order is not guaranteed.
- Syntax is similar to List Comprehension.
- No key-value pair is used.

# Generator Expression

## What is Generator Expression?

A Generator Expression is used to generate values one by one instead of storing all values in memory at once.

It is memory efficient and works similarly to List Comprehension.

---

## Syntax

```python
generator = (expression for item in iterable)
```

> **Note:** Parentheses `()` create a Generator Expression, **not** a Tuple.

Python does **not** support Tuple Comprehension.
```

````python
# Generator Expression

numbers = (i for i in range(1, 6))

print(numbers)
````

````python
# Access Generator Values

numbers = (i for i in range(1, 6))

for value in numbers:
    print(value)
````

````python
# Convert Generator into List

numbers = (i ** 2 for i in range(1, 6))

result = list(numbers)

print(result)
````

````python
# Convert Generator into Tuple

numbers = (i ** 2 for i in range(1, 6))

result = tuple(numbers)

print(result)
````

````python
# Generator with if Condition

even_numbers = (
    i
    for i in range(1, 11)
    if i % 2 == 0
)

print(list(even_numbers))
````

# List Comprehension vs Generator Expression

| List Comprehension | Generator Expression |
|--------------------|----------------------|
| Uses [] | Uses () |
| Stores all values immediately | Generates values one by one |
| Faster for small data | Better for large data |
| More memory usage | Less memory usage |
| Returns List | Returns Generator |
