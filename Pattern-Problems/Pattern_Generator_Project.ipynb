## 📌 Problem Statement

Create a Python program that allows user to select a pattern type and print it using loops.

Patterns included:
1. Square
2. Right Triangle
3. Pyramid
4. Inverted Pyramid

```python
# ===== MENU DISPLAY =====
# This shows available pattern options to user

print("===================================")
print("      PATTERN GENERATOR APP        ")
print("===================================")

print("1. Square Pattern")
print("2. Right Triangle Pattern")
print("3. Pyramid Pattern")
print("4. Inverted Pyramid Pattern")

print("===================================")
```

```python
# ===== USER INPUT =====
# Taking input from user for pattern size and choice

n = int(input("Enter size (n): "))      # size of pattern
choice = int(input("Enter your choice (1-4): "))  # pattern selection
```

```python
# ===== SQUARE PATTERN =====
# Logic: rows = n, columns = n

if choice == 1:

    print("\nSquare Pattern:\n")

    for i in range(n):          # outer loop = rows

        for j in range(n):      # inner loop = columns
            print("*", end=" ")

        print()  # new line after each row
```

```python
# ===== RIGHT TRIANGLE PATTERN =====
# Logic: row 1 → 1 star, row 2 → 2 stars ...

elif choice == 2:

    print("\nRight Triangle Pattern:\n")

    for i in range(1, n+1):     # rows

        for j in range(i):      # columns depend on row number
            print("*", end=" ")

        print()
```

```python
# ===== PYRAMID PATTERN =====
# Logic: spaces + stars

elif choice == 3:

    print("\nPyramid Pattern:\n")

    for i in range(1, n+1):

        # printing spaces before stars
        for j in range(n - i):
            print(" ", end="")

        # printing stars
        for j in range(i):
            print("* ", end="")

        print()
```

```python
# ===== INVERTED PYRAMID PATTERN =====
# Logic: stars decrease, spaces increase

elif choice == 4:

    print("\nInverted Pyramid Pattern:\n")

    for i in range(n, 0, -1):

        # spaces
        for j in range(n - i):
            print(" ", end="")

        # stars
        for j in range(i):
            print("* ", end="")

        print()
```

```python
# ===== INVALID INPUT HANDLING =====
# If user enters wrong choice

else:
    print("Invalid Choice! Please select between 1 to 4.")
```

```python
# ================================
# 🚀 PATTERN GENERATOR PROJECT
# ================================

n = int(input("Enter size (n): "))
choice = int(input("Enter your choice (1-4): "))

print("\nOUTPUT:\n")

if choice == 1:

    # rows
    for i in range(n):

        # columns
        for j in range(n):
            print("*", end=" ")

        print()


elif choice == 2:

    # rows
    for i in range(1, n+1):

        # columns depend on row number
        for j in range(i):
            print("*", end=" ")

        print()


elif choice == 3:

    # rows
    for i in range(1, n+1):

        # spaces
        for j in range(n - i):
            print(" ", end="")

        # stars
        for j in range(i):
            print("* ", end="")

        print()

elif choice == 4:

    # rows
    for i in range(n, 0, -1):

        # spaces
        for j in range(n - i):
            print(" ", end="")

        # stars
        for j in range(i):
            print("* ", end="")

        print()

else:
    print("Invalid Choice! Please select 1 to 4.")
```
