# Iterable

Before learning **Iterator**, we first need to understand **Iterable** because every Iterator starts from an Iterable.

---

## What is an Iterable?

An **Iterable** is any object whose elements can be accessed one by one.

In simple words,

> **If an object can be used inside a `for` loop, it is an Iterable.**

Python already provides many built-in Iterables.

Some common Iterables are:

- List
- Tuple
- String
- Dictionary
- Set
- Range

Whenever we write a `for` loop, Python automatically reads one element at a time from the Iterable.

Example:

```python
numbers = [10, 20, 30]

for num in numbers:
    print(num)
```

Here, **numbers** is an Iterable.

# How Does a `for` Loop Work?

Most beginners think that a `for` loop directly reads elements from a list.

Actually, that is **not** what happens.

Internally, Python first converts the Iterable into an **Iterator**, and then repeatedly asks the Iterator for the next element until no elements are left.

The process is:

```
Iterable
     │
     ▼
iter()
     │
     ▼
Iterator
     │
next()
     │
     ▼
One element at a time
```

This is exactly what happens behind every `for` loop.

````python
# List is an Iterable

numbers = [10, 20, 30, 40]

for num in numbers:
    print(num)
````

````python
# Tuple is an Iterable

numbers = (1, 2, 3, 4)

for num in numbers:
    print(num)
````

````python
# String is an Iterable

text = "Python"

for ch in text:
    print(ch)
````

````python
# Dictionary is an Iterable

student = {
    "name": "Deepika",
    "age": 21,
    "course": "Python"
}

for key in student:
    print(key)
````

````python
# Set is an Iterable

numbers = {10, 20, 30, 40}

for num in numbers:
    print(num)
````

````python
# Range is also an Iterable

for i in range(1, 6):
    print(i)
````

# Why Do We Need Iterables?

Imagine you have a list containing **10 lakh numbers**.

Without Iterables, Python would need a separate way to access each element.

Instead, Python follows a common rule:

> **Any object that can provide its elements one by one is called an Iterable.**

This allows the same `for` loop to work with:

- Lists
- Tuples
- Strings
- Dictionaries
- Sets
- Files
- Generators

The `for` loop doesn't need to know what type of object it is working with. It simply asks for the next value until the data is finished.

# Important Points

- An Iterable is an object that can be traversed one element at a time.
- Every List is an Iterable.
- Every Tuple is an Iterable.
- Every String is an Iterable.
- Every Dictionary is an Iterable.
- Every Set is an Iterable.
- Every Range object is an Iterable.
- A `for` loop works only with Iterables.
- Internally, Python converts an Iterable into an Iterator before iteration starts.

# Generator

## What is a Generator?

A **Generator** is a special type of **Iterator** that generates values **one at a time** instead of storing all values in memory.

Unlike a normal function, a Generator uses the **`yield`** keyword instead of `return`.

Generators are memory-efficient because they produce values only when they are needed.

---

## Why Do We Need Generators?

Suppose you want numbers from **1 to 1,000,000**.

### Normal Function

- Creates and stores all values in memory.
- Uses more memory.

### Generator

- Creates one value at a time.
- Uses very little memory.
- Suitable for working with large datasets.

Generators are especially useful when dealing with large files, huge datasets, or continuous streams of data.

````python
# Normal Function

def get_numbers():
    return [1, 2, 3, 4, 5]

numbers = get_numbers()

print(numbers)
````

````python
# Generator Function

def get_numbers():
    yield 1
    yield 2
    yield 3
    yield 4
    yield 5

numbers = get_numbers()

print(numbers)
````

# return vs yield

A normal function uses **`return`** to send back a value and then immediately stops execution.

A Generator uses **`yield`**, which returns one value and **pauses** the function.

When the next value is requested, execution continues from where it stopped.

This is why Generators can produce values one by one without storing everything in memory.

````python
# return Example

def greet():
    return "Hello"
    return "Python"   # This line will never execute

print(greet())
````

````python
# yield Example

def greet():
    yield "Hello"
    yield "Python"

gen = greet()

print(next(gen))
print(next(gen))
````

# How `yield` Works

Every time `next()` is called:

1. The Generator starts executing.
2. Execution stops at the first `yield`.
3. The yielded value is returned.
4. Calling `next()` again resumes execution from the same position.
5. This continues until there are no more `yield` statements.

````python
def numbers():
    print("Start")

    yield 10

    print("Middle")

    yield 20

    print("End")

    yield 30


gen = numbers()

print(next(gen))
print(next(gen))
print(next(gen))
````

# Generator Flow

```
Generator Function
        │
        ▼
 Generator Object
        │
     next()
        │
        ▼
 First yield
        │
     next()
        │
        ▼
 Second yield
        │
     next()
        │
        ▼
 Third yield
        │
     next()
        │
        ▼
 StopIteration
```

````python
# Using a Generator in a for Loop

def numbers():
    yield 1
    yield 2
    yield 3
    yield 4
    yield 5

for num in numbers():
    print(num)
````

# Important Points

- A Generator is a special type of Iterator.
- A Generator uses the `yield` keyword.
- `yield` pauses the function instead of stopping it.
- Values are generated one at a time.
- Generators are memory-efficient.
- A Generator object can be iterated only once.
