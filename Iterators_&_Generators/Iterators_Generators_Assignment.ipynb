## Question 6

Convert the following list into an Iterator using `iter()` and print the first **three** elements using `next()`.

```python
fruits = ["Apple", "Banana", "Mango", "Orange", "Kiwi"]
```

````python
fruits = ["Apple", "Banana", "Mango", "Orange", "Kiwi"]

fruit_iterator = iter(fruits)

print(next(fruit_iterator))
print(next(fruit_iterator))
print(next(fruit_iterator))
````

## Question 7

Create a Generator Function that yields the first **five multiples of 3**.

````python
def multiples_of_three():
    for i in range(1, 6):
        yield i * 3


gen = multiples_of_three()

for num in gen:
    print(num)
````

## Question 10

Create a Generator that yields numbers from **1 to 10**.
Print them using a `for` loop.

````python
def generate_numbers():
    for i in range(1, 11):
        yield i


numbers = generate_numbers()

for num in numbers:
    print(num)
````

## Question 11

Create a Generator that yields the square of numbers from **1 to 8**.

````python
def square_generator():
    for i in range(1, 9):
        yield i ** 2


result = square_generator()

for value in result:
    print(value)
````

## Question 12

Create a Generator that yields only even numbers between **1 and 20**.

````python
def even_generator():
    for i in range(1, 21):
        if i % 2 == 0:
            yield i


numbers = even_generator()

for num in numbers:
    print(num)
````

# 🔴 Level 3 - Interview Level

## Question 13

Create a Generator that yields the characters of a given string one by one.

```python
text = "Generator"
```

````python
def character_generator(text):
    for ch in text:
        yield ch


result = character_generator("Generator")

for ch in result:
    print(ch)
````

## Question 14

Create a Generator that yields numbers from **100 to 110**.

Print the first **five** numbers using `next()`.

````python
def number_generator():
    for i in range(100, 111):
        yield i


numbers = number_generator()

print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
````

## Question 15

Create a Generator that yields multiples of **5** up to **50**.

````python
def multiples_of_five():
    for i in range(5, 51, 5):
        yield i


result = multiples_of_five()

for value in result:
    print(value)
````
