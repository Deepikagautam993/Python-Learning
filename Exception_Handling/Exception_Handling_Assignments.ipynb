## Q1. Handle Division by Zero

Create a program that takes two numbers
and handles division by zero error.

```python
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    result = num1 / num2

    print("Result:", result)

except ZeroDivisionError:
    print("Cannot divide by zero")
```

## Q2. Handle Invalid Input

Take integer input from user.

If user enters a string,
handle ValueError.

```python
try:
    num = int(input("Enter number: "))

    print("Number is:", num)

except ValueError:
    print("Please enter valid integer")
```

## Q3. Age Validation

Take age as input.

If age is less than 18,
raise an exception.

```python
try:
    age = int(input("Enter age: "))

    if age < 18:
        raise Exception("Not eligible")

    print("Eligible")

except Exception as e:
    print(e)
```

## Q4. Use else and finally

Create a program that shows:

- Success message if no error
- Finally message always

```python
try:
    num = 10 / 2

except:
    print("Error")

else:
    print("Operation successful")

finally:
    print("Program completed")
```

## Q5. Create Custom Exception

Create a custom exception
for negative numbers.

```python
class NegativeNumberError(Exception):
    pass


try:
    num = int(input("Enter number: "))

    if num < 0:
        raise NegativeNumberError("Negative number not allowed")

    print(num)

except NegativeNumberError as e:
    print(e)
```

## Q6. Safe Calculator

Create calculator that handles:

- Wrong input
- Division by zero
- Invalid operator

```python
try:

    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))

    op = input("Enter operator: ")

    if op == "+":
        print(a+b)

    elif op == "-":
        print(a-b)

    elif op == "*":
        print(a*b)

    elif op == "/":
        print(a/b)

    else:
        print("Invalid operator")


except ValueError:
    print("Enter numbers only")

except ZeroDivisionError:
    print("Cannot divide by zero")
```

#Solution 1 — Password Checker with raise

```python
def check_password(password):

    if len(password) < 8:
        raise Exception("Password must be at least 8 characters")

    return "Password accepted"


try:
    password = input("Enter password: ")

    print(check_password(password))

except Exception as e:
    print(e)
```

#Solution 2 — ATM Withdrawal System

```python
balance = 5000

try:
    amount = int(input("Enter withdrawal amount: "))

    if amount > balance:
        raise Exception("Insufficient balance")

    if amount <= 0:
        raise Exception("Invalid amount")

    balance -= amount

    print("Withdrawal successful")
    print("Remaining balance:", balance)

except ValueError:
    print("Enter valid amount")

except Exception as e:
    print(e)

    print("Withdrawal failed")
```

# Solution 3 — Login System using Exception

```python
correct_password = "python123"


try:

    password = input("Enter password: ")

    if password != correct_password:
        raise Exception("Incorrect password")

    print("Login successful")


except Exception as e:
    print(e)
```

# Solution 4 — File Opening Error Handling

```python
try:

    file = open("data.txt", "r")

    content = file.read()

    print(content)

    file.close()


except FileNotFoundError:

    print("File does not exist")
```
