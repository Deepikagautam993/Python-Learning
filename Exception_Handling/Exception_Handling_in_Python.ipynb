# 🚀 Exception Handling in Python

Exception Handling is used to handle runtime errors
and prevent the program from crashing.

Topics Covered:

- Error vs Exception
- try-except
- Specific Exceptions
- Multiple Exceptions
- else block
- finally block
- raise keyword
- Custom Exception
- Mini Project

# 📌 What is an Error?

An error is a problem in a program that stops
the normal execution of code.

Example:

- Syntax Error
- Runtime Error
- Logical Error

```python
# Example of runtime error

print(10/0)
```

# 📌 What is an Exception?

An exception is a runtime problem that can be handled
using exception handling.

Examples:

- Division by zero
- Wrong input
- Invalid operation

# Error vs Exception

Error:
- Stops program execution
- Usually cannot be handled

Exception:
- Occurs during execution
- Can be handled using try-except

# 📌 try-except

try block contains risky code.

except block handles the error.

```python
# Basic try-except example

try:
    num = int(input("Enter number: "))
    print(num)

except:
    print("Invalid input")
```

# 📌 Specific Exception Handling

Handling specific exceptions is better practice.

```python
try:

    a = 10
    b = 0

    print(a/b)

except ZeroDivisionError:

    print("Cannot divide by zero")
```

# 📌 Multiple Exceptions

Different errors can be handled separately.

```python
try:

    num = int(input("Enter number: "))

    print(10/num)


except ValueError:

    print("Enter valid number")


except ZeroDivisionError:

    print("Zero is not allowed")
```

# 📌 else Block

else executes only when no exception occurs.

```python
try:

    result = 10/2


except:

    print("Error")


else:

    print("Result:", result)
```

# 📌 finally Block

finally always executes whether exception occurs or not.

Used for cleanup work like:
- Closing files
- Releasing resources

```python
try:

    print(10/0)


except:

    print("Error occurred")


finally:

    print("Program finished")
```

# 📌 raise Keyword

raise is used to manually create an exception.

```python
age = 15


if age < 18:

    raise Exception("Age should be 18+")
```

# 📌 Custom Exception

We can create our own exception class
by inheriting Exception class.

```python
class AgeError(Exception):
    pass


age = 16


try:

    if age < 18:
        raise AgeError("Not eligible")


except AgeError as e:

    print(e)
```

# 🧮 Mini Project: Safe Calculator

Features:

- Handles wrong input
- Handles division by zero
- Handles invalid operator

```python
try:

    num1 = float(input("Enter first number: "))

    num2 = float(input("Enter second number: "))


    operator = input("Enter operator (+,-,*,/): ")


    if operator == "+":

        print(num1 + num2)


    elif operator == "-":

        print(num1 - num2)


    elif operator == "*":

        print(num1 * num2)


    elif operator == "/":

        print(num1 / num2)


    else:

        print("Invalid operator")


except ValueError:

    print("Please enter numbers only")


except ZeroDivisionError:

    print("Cannot divide by zero")
```
