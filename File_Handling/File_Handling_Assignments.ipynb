## Q1. Create a file and write your name and course.

Create a file named student.txt and store your details.

```python
# Create and write data into file

with open("student.txt","w") as file:

    file.write("Name: Deepika\n")
    file.write("Course: AIML")

print("Data saved")
```

## Q2. Read complete file content.

Read student.txt file and display data.

```python
# Reading complete file

with open("student.txt","r") as file:

    data = file.read()

    print(data)
```

## Q3. Append new information.

Add college name without removing old data.

```python
# Append data

with open("student.txt","a") as file:

    file.write("\nCollege: BGIEM")

print("Updated")
```

## Q4. Count number of words in a file.

```python
with open("student.txt","r") as file:

    data = file.read()

    words = data.split()

    print("Total words:", len(words))
```

## Q5. Count number of lines in a file.

```python
with open("student.txt","r") as file:

    lines = file.readlines()

    print("Total lines:", len(lines))
```

## Q6. Copy data from one file to another.

```python
# Read source file

with open("student.txt","r") as file:

    data = file.read()


# Write into new file

with open("copy.txt","w") as file:

    file.write(data)


print("Copied successfully")
```

## Q7. Search a word in a file.

Check whether a word exists or not.

```python
word = input("Enter word: ")

with open("student.txt","r") as file:

    data = file.read()


if word in data:

    print("Word found")

else:

    print("Word not found")
```

## Q8. Replace a word in a file.

```python
with open("student.txt","r") as file:

    data = file.read()


data = data.replace("AIML","Python")


with open("student.txt","w") as file:

    file.write(data)


print("Updated")
```

## Q9. Find longest word in a file.

```python
with open("student.txt","r") as file:

    data = file.read()


words = data.split()

longest = max(words, key=len)

print("Longest word:", longest)
```

## Q10. Create Student Record System.

Store student details permanently using file.

```python
def add_student():

    name = input("Enter name: ")
    marks = input("Enter marks: ")


    with open("students.txt","a") as file:

        file.write(
            f"Name: {name}, Marks: {marks}\n"
        )


def view_students():

    try:

        with open("students.txt","r") as file:

            print(file.read())

    except FileNotFoundError:

        print("No records found")


while True:

    print("\n1. Add Student")
    print("2. View Students")
    print("3. Exit")


    choice = input("Choice: ")


    if choice == "1":

        add_student()


    elif choice == "2":

        view_students()


    elif choice == "3":

        break

    else:

        print("Invalid choice")
```

## Q11. Create Log File System

Store program activities with time.

```python
from datetime import datetime


with open("log.txt","a") as file:

    file.write(
        str(datetime.now()) + " Program started\n"
    )

print("Log saved")
```
