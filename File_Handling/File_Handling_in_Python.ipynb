# 📂 File Handling in Python

## 📌 Overview

File Handling allows Python programs to store and retrieve data from files.

Unlike variables, which store data temporarily in memory (RAM), files store data permanently on the storage device.

Using File Handling, we can:

- Create files
- Read data
- Write data
- Append new data
- Update existing data
- Delete files
- Organize application data permanently

# 📌 Why Do We Need File Handling?

Variables store data only while the program is running.

Example:

```python
name = "Deepika"
```

After the program ends, the value stored in the variable is lost.

If we want to store information permanently, we need files.

Examples:

- Student Records
- Banking Data
- Expense Tracker
- Login Information
- Notes Application
- Inventory System

# 📌 Memory vs File Storage

| Memory (RAM) | File Storage |
|--------------|--------------|
| Temporary | Permanent |
| Fast | Slightly Slower |
| Data is lost after program ends | Data remains saved |
| Used while program is running | Used for long-term storage |

# 📌 Opening a File

Python provides the built-in `open()` function.

### Syntax

```python
file_object = open(file_name, mode)
```

### Parameters

- `file_name` → Name or path of the file
- `mode` → Operation to perform on the file

# 📌 Example

```python
file = open("student.txt", "r")

print(file)

file.close()
```

Output

```
<_io.TextIOWrapper
name='student.txt'
mode='r'
encoding='utf-8'>
```

`open()` returns a **file object**, which is used to perform operations on the file.

# 📌 File Modes

A file mode specifies the operation to perform on a file.

| Mode | Description |
|------|-------------|
| `r` | Read |
| `w` | Write |
| `a` | Append |
| `x` | Create New File |
| `rb` | Read Binary |
| `wb` | Write Binary |
| `r+` | Read and Write |
| `w+` | Write and Read |
| `a+` | Append and Read |

# 📌 Difference Between File Modes

| Mode | File Exists | File Doesn't Exist | Old Data |
|------|-------------|--------------------|----------|
| `r` | Opens | Error | Safe |
| `w` | Opens | Creates | Deleted |
| `a` | Opens | Creates | Preserved |
| `x` | Error | Creates | New File |

# 📖 Read Mode (`r`)

Read mode is used to read data from an existing file.

### Syntax

```python
file = open("student.txt", "r")
```

### Important Points

- File must already exist.
- Cannot modify the file.
- If the file does not exist, Python raises a `FileNotFoundError`.

````python
file = open("student.txt", "r")

content = file.read()

print(content)

file.close()
````

# ✍️ Write Mode (`w`)

Write mode is used to write data into a file.

### Syntax

```python
file = open("student.txt", "w")
```

### Important Points

- Creates the file if it does not exist.
- If the file already exists, all previous data is deleted.
- Always starts writing from the beginning of the file.

````python
file = open("student.txt", "w")

file.write("Welcome to Python\n")
file.write("File Handling")

file.close()
````

# ➕ Append Mode (`a`)

Append mode is used to add new data at the end of a file.

### Syntax

```python
file = open("student.txt", "a")
```

### Important Points

- Creates the file if it does not exist.
- Existing data remains unchanged.
- New data is always added at the end of the file.

````python
file = open("student.txt", "a")

file.write("\nLearning Python Modules")

file.close()
````

# 🆕 Create Mode (`x`)

Create mode is used to create a new file.

### Syntax

```python
file = open("student.txt", "x")
```

### Important Points

- Creates a new file.
- Raises a `FileExistsError` if the file already exists.

````python
file = open("new_file.txt", "x")

file.close()
````

# 📖 Reading Methods

Python provides three methods to read data from a file.

- `read()`
- `readline()`
- `readlines()`

## 1. `read()`

Reads the complete file.

### Syntax

```python
file.read()
```

````python
file = open("student.txt", "r")

data = file.read()

print(data)

file.close()
````

## `read(size)`

You can also specify the number of characters to read.

Example:

```python
file.read(8)
```

Only the first 8 characters will be read.

````python
file = open("student.txt", "r")

print(file.read(8))

file.close()
````

## 2. `readline()`

Reads only one line at a time.

Each call to `readline()` reads the next line.

````python
file = open("student.txt", "r")

print(file.readline())
print(file.readline())
print(file.readline())

file.close()
````

## 3. `readlines()`

Reads all lines and returns them as a list.

````python
file = open("student.txt", "r")

lines = file.readlines()

print(lines)

file.close()
````

# 📌 Difference Between Reading Methods

| Method | Returns | Best Used For |
|---------|----------|---------------|
| `read()` | String | Read complete file |
| `readline()` | String | Read one line at a time |
| `readlines()` | List | Read all lines as a list |

# ✍️ Writing Methods

Python provides two methods to write data into a file.

- `write()`
- `writelines()`

## 1. `write()`

The `write()` method is used to write a single string into a file.

### Syntax

```python
file.write(data)
```

### Important Points

- Accepts only a string.
- Returns the number of characters written.
- Does not automatically move to the next line.

````python
file = open("student.txt", "w")

characters = file.write("Python File Handling")

print(characters)

file.close()
````

## Writing Multiple Lines using `write()`

To write multiple lines, use the newline character (`\n`).

````python
file = open("student.txt", "w")

file.write("Python\n")
file.write("Modules\n")
file.write("Packages")

file.close()
````

## 2. `writelines()`

The `writelines()` method writes multiple strings from a list into a file.

### Syntax

```python
file.writelines(iterable)
```

### Important Points

- Accepts a list (or any iterable) of strings.
- Does **not** add newline characters automatically.

````python
file = open("student.txt", "w")

lines = [
    "Apple\n",
    "Banana\n",
    "Mango\n",
    "Orange"
]

file.writelines(lines)

file.close()
````

# 📌 Difference Between `write()` and `writelines()`

| `write()` | `writelines()` |
|------------|----------------|
| Writes one string | Writes multiple strings |
| Accepts a string | Accepts a list (or iterable) of strings |
| Used for small text | Used for multiple lines |

# 📍 File Pointer

Whenever a file is opened, Python maintains a **file pointer**.

The file pointer keeps track of the current position inside the file.

After every read or write operation, the pointer automatically moves forward.

## `tell()`

The `tell()` method returns the current position of the file pointer.

### Syntax

```python
file.tell()
```

````python
file = open("student.txt", "r")

print(file.tell())

file.read(6)

print(file.tell())

file.close()
````

## `seek()`

The `seek()` method moves the file pointer to a specified position.

### Syntax

```python
file.seek(position)
```

````python
file = open("student.txt", "r")

print(file.read(6))

file.seek(0)

print(file.read())

file.close()
````

# 📌 Useful File Object Methods

| Method | Purpose |
|---------|---------|
| `close()` | Closes the file |
| `readable()` | Checks whether the file can be read |
| `writable()` | Checks whether the file can be written |
| `tell()` | Returns current pointer position |
| `seek()` | Moves the pointer |
| `flush()` | Immediately writes buffered data to the file |

## `readable()`

Returns `True` if the file is opened in a readable mode.

````python
file = open("student.txt", "r")

print(file.readable())

file.close()
````

## `writable()`

Returns `True` if the file is opened in a writable mode.

````python
file = open("student.txt", "a")

print(file.writable())

file.close()
````

## `flush()`

The `flush()` method forces buffered data to be written to the file immediately.

Normally, Python writes buffered data when the file is closed.

### Syntax

```python
file.flush()
```

````python
file = open("student.txt", "w")

file.write("Learning Python")

file.flush()      # Save data immediately

file.close()
````

# 📌 Important Notes

- The file pointer moves automatically after reading or writing.
- `tell()` shows the current pointer position.
- `seek()` changes the pointer position.
- `write()` accepts a single string.
- `writelines()` accepts multiple strings.
- `flush()` is useful when data should be saved immediately.

# 📌 The `with` Statement

The `with` statement is the recommended way to work with files.

It automatically closes the file after the block of code is executed, even if an error occurs.

### Syntax

```python
with open(file_name, mode) as file:
    # Perform file operations
```

## Why Use `with`?

Without `with`:

```python
file = open("student.txt", "r")

data = file.read()

file.close()
```

With `with`:

```python
with open("student.txt", "r") as file:
    data = file.read()
```

### Advantages

- No need to call `close()`
- Cleaner code
- Automatically closes the file
- Reduces the chance of resource leaks

````python
with open("student.txt", "r") as file:

    content = file.read()

    print(content)
````

# 📌 Mini Project - Notes Manager

## Features

- Add Notes
- View Notes
- Store Notes Permanently
- Uses `with` Statement

````python
def add_note():

    note = input("Enter Note: ")

    with open("notes.txt", "a") as file:
        file.write(note + "\n")

    print("Note Saved Successfully.")


def view_notes():

    try:
        with open("notes.txt", "r") as file:
            print("\nYour Notes:\n")
            print(file.read())

    except FileNotFoundError:
        print("No Notes Found.")


while True:

    print("\n===== Notes Manager =====")
    print("1. Add Note")
    print("2. View Notes")
    print("3. Exit")

    choice = input("Enter Choice: ")

    if choice == "1":
        add_note()

    elif choice == "2":
        view_notes()

    elif choice == "3":
        print("Thank You!")
        break

    else:
        print("Invalid Choice.")
````
