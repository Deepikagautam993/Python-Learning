# 📌 Why File Handling?

Variables store data temporarily.

Example:

name = "Deepika"

When program ends, data is lost.

Files allow us to store data permanently.

# 📌 Opening a File

Python provides open() function.

Syntax:

file_object = open(filename, mode)

Example:

open("data.txt","r")

```python
# Opening a file

file = open("data.txt", "r")

print(file)

file.close()
```

# 📌 File Modes

Modes decide what operation we want to perform.

r  → Read file
w  → Write file
a  → Append data
x  → Create new file

Binary modes:

rb → Read binary
wb → Write binary

# 1. Read Mode (r)

Read mode is used to read existing files.

Features:

- File must exist
- Cannot modify data

```python
# Reading complete file

file = open("data.txt", "r")

data = file.read()

print(data)

file.close()
```

# 2. Write Mode (w)

Write mode is used to add new data.

Important:

If file already contains data,
it will overwrite the old data.

```python
# Writing data into file

file = open("data.txt", "w")

file.write("Python File Handling")

file.close()
```

# 3. Append Mode (a)

Append adds new data at the end.

Existing data is not removed.

```python
# Appending data

file = open("data.txt", "a")

file.write("\nNew line added")

file.close()
```

# 4. Create Mode (x)

Creates a new file.

If file already exists,
it gives an error.

```python
# Creating new file

file = open("newfile.txt", "x")

file.close()
```

# 📌 Reading Methods

Python provides:

1. read()
2. readline()
3. readlines()

```python
# read()

file = open("data.txt","r")

print(file.read())

file.close()
```

```python
# readline()

file = open("data.txt","r")

print(file.readline())

file.close()
```

```python
# readlines()

file = open("data.txt","r")

print(file.readlines())

file.close()
```

# 📌 Writing Methods

write()

Used to write a single string.

writelines()

Used to write multiple lines.

```python
# write()

file = open("data.txt","w")

file.write("Hello Python")

file.close()
```

```python
# writelines()

file = open("data.txt","w")

lines = [
    "Python\n",
    "File\n",
    "Handling"
]

file.writelines(lines)

file.close()
```

# writelines()

file = open("data.txt","w")

lines = [
    "Python\n",
    "File\n",
    "Handling"
]

file.writelines(lines)

file.close()

```python
file = open("data.txt","r")

print(file.tell())

file.read(5)

print(file.tell())

file.seek(0)

print(file.tell())

file.close()
```

# 📌 with Statement

with automatically closes the file.

Benefits:

- No need to call close()
- Safer code
- Cleaner syntax

```python
with open("data.txt","r") as file:

    data = file.read()

    print(data)
```

# 📌 Exception Handling with Files

Files may cause errors:

- File not found
- Permission error

```python
try:

    file = open("abc.txt","r")

    print(file.read())

except FileNotFoundError:

    print("File does not exist")
```

# 📌 Check File Exists

Using os module we can check file existence.

```python
import os

if os.path.exists("data.txt"):

    print("File exists")

else:

    print("File not found")
```

# 📌 Delete a File

os.remove() deletes a file.

```python
import os

if os.path.exists("data.txt"):

    os.remove("data.txt")

    print("Deleted")
```

# 📝 Mini Project: Notes Manager

Features:

- Add notes
- View notes
- Save permanently using file

```python
def add_note():

    note = input("Enter note: ")

    with open("notes.txt","a") as file:

        file.write(note + "\n")

    print("Note saved")


def view_notes():

    try:

        with open("notes.txt","r") as file:

            print(file.read())

    except FileNotFoundError:

        print("No notes found")


while True:

    print("\n1. Add Note")
    print("2. View Notes")
    print("3. Exit")


    choice = input("Choose: ")


    if choice == "1":

        add_note()


    elif choice == "2":

        view_notes()


    elif choice == "3":

        break


    else:

        print("Invalid choice")
```
