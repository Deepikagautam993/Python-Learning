# 🚀 Pattern Generator Project

This project creates a menu-driven pattern generator using Python.

Concepts used:
- Functions
- Loops
- Conditional statements
- User input

## 🎯 Objective

Create a program where users can:

1. Select a pattern
2. Enter pattern size
3. Generate the selected pattern

This project combines all previous pattern concepts into one application.

## 🔄 Program Flow

User
 ↓
Select Pattern
 ↓

 ↓
Function Call
 ↓
Pattern Output

## Functions Used

Each pattern has its own function:

square_pattern()
triangle_pattern()
pyramid_pattern()
diamond_pattern()

Benefits:
- Code reuse
- Clean structure
- Easy modification

```python
def square_pattern(n):

    for i in range(n):

        for j in range(n):
            print("*", end=" ")

        print()
```

```python
def triangle_pattern(n):

    for i in range(1, n+1):

        for j in range(i):
            print("*", end=" ")

        print()
```

```python
def pyramid_pattern(n):

    for i in range(1, n+1):

        for j in range(n-i):
            print(" ", end="")

        for j in range(i):
            print("* ", end="")

        print()
```

```python
def diamond_pattern(n):

    for i in range(1, n+1):

        for j in range(n-i):
            print(" ", end="")

        for j in range(i):
            print("* ", end="")

        print()


    for i in range(n-1,0,-1):

        for j in range(n-i):
            print(" ", end="")

        for j in range(i):
            print("* ", end="")

        print()
```

```python
def square_pattern(n):
    for i in range(n):
        for j in range(n):
            print("*", end=" ")
        print()


def triangle_pattern(n):
    for i in range(1, n+1):
        for j in range(i):
            print("*", end=" ")
        print()


def pyramid_pattern(n):
    for i in range(1, n+1):

        for j in range(n-i):
            print(" ", end="")

        for j in range(i):
            print("* ", end="")

        print()


def diamond_pattern(n):

    for i in range(1,n+1):

        for j in range(n-i):
            print(" ", end="")

        for j in range(i):
            print("* ", end="")

        print()


    for i in range(n-1,0,-1):

        for j in range(n-i):
            print(" ", end="")

        for j in range(i):
            print("* ", end="")

        print()


print("===== PATTERN GENERATOR =====")

print("1. Square Pattern")
print("2. Triangle Pattern")
print("3. Pyramid Pattern")
print("4. Diamond Pattern")

choice = int(input("Enter choice: "))
n = int(input("Enter size: "))


if choice == 1:
    square_pattern(n)

elif choice == 2:
    triangle_pattern(n)

elif choice == 3:
    pyramid_pattern(n)

elif choice == 4:
    diamond_pattern(n)

else:
    print("Invalid Choice")
```
