# OS Module in Python

## OS Module

The `os` module provides functions to interact with the operating system. It is commonly used for working with files, folders, and directories.

```python
import os
```

## Current Working Directory

Use `os.getcwd()` to get the current working directory.

```python
import os

print(os.getcwd())
```

## List Files and Folders

Use `os.listdir()` to display all files and folders in the current directory.

```python
import os

print(os.listdir())
```

## Create a Folder

Use `os.mkdir()` to create a new folder.

```python
import os

os.mkdir("Python_Files")

print("Folder created successfully.")
```

## Rename a Folder

Use `os.rename()` to rename an existing file or folder.

```python
import os

os.rename("Python_Files", "My_Folder")

print("Folder renamed successfully.")
```

## Remove an Empty Folder

Use `os.rmdir()` to delete an empty folder.

```python
import os

os.rmdir("My_Folder")

print("Folder removed successfully.")
```

## Check File or Folder Exists

Use `os.path.exists()` to check whether a file or folder exists.

```python
import os

print(os.path.exists("students.csv"))
print(os.path.exists("student.json"))
```

## Check File or Directory

Use `os.path.isfile()` and `os.path.isdir()` to identify files and folders.

```python
import os

print(os.path.isfile("students.csv"))
print(os.path.isdir("My_Folder"))
```

## Absolute Path

Use `os.path.abspath()` to get the absolute path of a file or folder.

```python
import os

print(os.path.abspath("students.csv"))
```

## File Name

Use `os.path.basename()` to get the file or folder name from a path.

```python
import os

path = os.path.abspath("students.csv")

print(os.path.basename(path))
```

## Directory Name

Use `os.path.dirname()` to get the directory path.

```python
import os

path = os.path.abspath("students.csv")

print(os.path.dirname(path))
```

## Join Paths

Use `os.path.join()` to create platform-independent file paths.

```python
import os

path = os.path.join("Python_Files", "students.csv")

print(path)
```

## Remove a File

Use `os.remove()` to delete a file.

```python
import os

# Remove the file only if it exists
if os.path.exists("sample.txt"):
    os.remove("sample.txt")
    print("File deleted successfully.")
else:
    print("File not found.")
```

## Traverse a Directory

Use `os.walk()` to iterate through folders and files.

```python
import os

for folder, subfolders, files in os.walk("."):
    print("Folder :", folder)
    print("Subfolders :", subfolders)
    print("Files :", files)
    print("-" * 40)
```
