Print today's day, month, and year separately.

# JSON Assignments

### Question 1

Convert the following Python dictionary into a JSON string.

```python
import json

student = {
    "name": "Deepika",
    "age": 21,
    "course": "AI & ML"
}

json_data = json.dumps(student)

print(json_data)
```

### Question 2

Convert the following JSON string into a Python dictionary.

```python
import json

json_data = '{"name":"Deepika","age":21,"course":"AI & ML"}'

student = json.loads(json_data)

print(student)
```

### Question 3

Write the following dictionary to a JSON file named `student.json`.

```python
import json

student = {
    "name": "Deepika",
    "age": 21,
    "course": "AI & ML"
}

with open("student.json", "w") as file:
    json.dump(student, file, indent=4)

print("Data written successfully.")
```

### Question 4

Read data from the `student.json` file and print it.

```python
import json

with open("student.json", "r") as file:
    student = json.load(file)

print(student)
```

### Question 5

Print the following dictionary as formatted JSON using an indentation of 4 spaces.

```python
import json

employee = {
    "id": 101,
    "name": "Rahul",
    "department": "Data Science"
}

print(json.dumps(employee, indent=4))
```

# CSV Assignments

### Question 1

Create a CSV file named `students.csv` and write the following student records into it.

```python
import csv

with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Age", "Course"])
    writer.writerow(["Deepika", 21, "AI & ML"])
    writer.writerow(["Rahul", 22, "Data Science"])

print("CSV file created successfully.")
```

### Question 2

Read all records from `students.csv` and print them.

```python
import csv

with open("students.csv", "r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

### Question 3

Write the following dictionary data into a CSV file named `employees.csv`.

```python
import csv

employees = [
    {"Name": "Amit", "Department": "HR"},
    {"Name": "Priya", "Department": "Finance"}
]

with open("employees.csv", "w", newline="") as file:
    fieldnames = ["Name", "Department"]

    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(employees)

print("Employee data written successfully.")
```

### Question 4

Read all records from `employees.csv` using `DictReader`.

```python
import csv

with open("employees.csv", "r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        print(row)
```

### Question 5

Print only the names of all students from `students.csv`.

```python
import csv

with open("students.csv", "r") as file:
    reader = csv.reader(file)

    next(reader)   # Skip header

    for row in reader:
        print(row[0])
```

### Question 3

Create a folder named `Practice`.

```python
import os

folder_name = "Practice"

if not os.path.exists(folder_name):
    os.mkdir(folder_name)
    print("Folder created successfully.")
else:
    print("Folder already exists.")
```

### Question 4

Rename the folder `Practice` to `Python_Practice`.

```python
import os

old_name = "Practice"
new_name = "Python_Practice"

if os.path.exists(old_name):
    os.rename(old_name, new_name)
    print("Folder renamed successfully.")
else:
    print("Source folder not found.")
```
