# Date & Time in Python

## datetime Module

The `datetime` module is used to work with dates and times in Python. It helps create, format, compare, and perform calculations on date and time values.

```python
import datetime
```

## Current Date

Use `date.today()` to get the current system date.

```python
import datetime

today = datetime.date.today()

print(today)
```

## Current Time

Use `datetime.now().time()` to get the current system time.

```python
import datetime

current_time = datetime.datetime.now().time()

print(current_time)
```

## Current Date & Time

Use `datetime.now()` to get both the current date and time together.

```python
import datetime

current_datetime = datetime.datetime.now()

print(current_datetime)
```

## Create a Date

Use the `date()` constructor to create a custom date.

```python
import datetime

my_date = datetime.date(2026, 7, 22)

print(my_date)
```

## Create a Time

Use the `time()` constructor to create a custom time.

```python
import datetime

my_time = datetime.time(10, 30, 45)

print(my_time)
```

## Create Date & Time

Use the `datetime()` constructor to create both date and time together.

```python
import datetime

my_datetime = datetime.datetime(2026, 7, 22, 10, 30, 45)

print(my_datetime)
```

## Date Components

Access individual parts of a date using attributes like `year`, `month`, and `day`.

```python
import datetime

today = datetime.datetime.now()

print("Year  :", today.year)
print("Month :", today.month)
print("Day   :", today.day)
```

## Time Components

Access individual parts of time using attributes like `hour`, `minute`, `second`, and `microsecond`.

```python
import datetime

now = datetime.datetime.now()

print("Hour        :", now.hour)
print("Minute      :", now.minute)
print("Second      :", now.second)
print("Microsecond :", now.microsecond)
```

## Weekday

Use `weekday()` to get the day of the week as a number.

- Monday = 0
- Tuesday = 1
- Wednesday = 2
- Thursday = 3
- Friday = 4
- Saturday = 5
- Sunday = 6

```python
import datetime

today = datetime.datetime.now()

print(today.weekday())
```

## ISO Format

Use `isoformat()` to convert a date and time into the standard ISO 8601 format.

```python
import datetime

print(datetime.datetime.now().isoformat())
```

## Timestamp

A timestamp represents the number of seconds since January 1, 1970 (Unix Epoch).

```python
import datetime

print(datetime.datetime.now().timestamp())
```

## Date Formatting (strftime)

Use `strftime()` to convert a datetime object into a formatted string.

```python
import datetime

now = datetime.datetime.now()

print(now.strftime("%d-%m-%Y"))
print(now.strftime("%d/%m/%Y"))
print(now.strftime("%A"))
print(now.strftime("%B"))
print(now.strftime("%I:%M %p"))
```

## Common Format Codes

Commonly used format codes for displaying dates and times.

```python
import datetime

now = datetime.datetime.now()

print("%d :", now.strftime("%d"))   # Day
print("%m :", now.strftime("%m"))   # Month
print("%Y :", now.strftime("%Y"))   # Year
print("%y :", now.strftime("%y"))   # Short Year

print("%A :", now.strftime("%A"))   # Full Weekday
print("%a :", now.strftime("%a"))   # Short Weekday

print("%B :", now.strftime("%B"))   # Full Month
print("%b :", now.strftime("%b"))   # Short Month

print("%H :", now.strftime("%H"))   # 24-Hour
print("%I :", now.strftime("%I"))   # 12-Hour
print("%M :", now.strftime("%M"))   # Minute
print("%S :", now.strftime("%S"))   # Second
print("%p :", now.strftime("%p"))   # AM / PM
```

## Common Date Formats

Examples of commonly used date and time formats.

```python
import datetime

now = datetime.datetime.now()

print(now.strftime("%d-%m-%Y"))
print(now.strftime("%Y-%m-%d"))
print(now.strftime("%d/%m/%Y"))
print(now.strftime("%A, %d %B %Y"))
print(now.strftime("%d %b %Y"))
print(now.strftime("%I:%M:%S %p"))
print(now.strftime("%H:%M:%S"))
```

## String to Date (strptime)

Use `strptime()` to convert a formatted string into a datetime object.

```python
import datetime

date = datetime.datetime.strptime(
    "22-07-2026",
    "%d-%m-%Y"
)

print(date)
```

## timedelta

Use `timedelta` to add or subtract days, weeks, hours, minutes, or seconds.

```python
import datetime

today = datetime.datetime.now()

print("Today          :", today)
print("After 10 Days  :", today + datetime.timedelta(days=10))
print("Before 7 Days  :", today - datetime.timedelta(days=7))
print("After 2 Weeks  :", today + datetime.timedelta(weeks=2))
print("After 5 Hours  :", today + datetime.timedelta(hours=5))
print("After 30 Minutes:", today + datetime.timedelta(minutes=30))
```

## Difference Between Two Dates

Subtract one date from another to calculate the difference.

```python
import datetime

date1 = datetime.date(2026, 7, 22)
date2 = datetime.date(2026, 8, 10)

difference = date2 - date1

print("Difference :", difference)
print("Days :", difference.days)
```

## Compare Dates

Date objects can be compared using comparison operators.

```python
import datetime

date1 = datetime.date(2026, 7, 22)
date2 = datetime.date(2026, 8, 1)

print(date1 == date2)
print(date1 < date2)
print(date1 > date2)
```

## Replace Date or Time Values

Use `replace()` to modify specific parts of a date or datetime object.

```python
import datetime

now = datetime.datetime.now()

updated = now.replace(year=2030, month=12, day=31)

print("Current :", now)
print("Updated :", updated)
```

## Practice Example

The following example combines multiple datetime operations together.

```python
import datetime

now = datetime.datetime.now()

print("Current Date & Time :", now)
print("Formatted Date      :", now.strftime("%d-%m-%Y"))
print("Current Year        :", now.year)
print("Current Month       :", now.month)
print("Current Day         :", now.day)
print("Next Week           :", now + datetime.timedelta(days=7))
```
