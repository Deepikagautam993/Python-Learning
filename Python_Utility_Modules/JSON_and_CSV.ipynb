# JSON & CSV in Python

## JSON Module

JSON (JavaScript Object Notation) is a lightweight format used to store and exchange structured data between applications and APIs.

```python
import json
```

## Python Dictionary to JSON

Use `json.dumps()` to convert a Python dictionary into a JSON string.

```python
import json

student = {
    "name": "Deepika",
    "age": 21,
    "course": "AI & ML"
}

json_data = json.dumps(student)

print(json_data)
print(type(json_data))
```

## JSON to Python Dictionary

Use `json.loads()` to convert a JSON string into a Python dictionary.

```python
import json

json_data = '{"name":"Deepika","age":21,"course":"AI & ML"}'

student = json.loads(json_data)

print(student)
print(type(student))
```

## Write JSON to a File

Use `json.dump()` to store Python data in a JSON file.

```python
import json

student = {
    "name": "Deepika",
    "age": 21,
    "course": "AI & ML"
}

with open("student.json", "w") as file:
    json.dump(student, file, indent=4)

print("JSON file created successfully.")
```

## Read JSON from a File

Use `json.load()` to read JSON data from a file.

```python
import json

with open("student.json", "r") as file:
    student = json.load(file)

print(student)
```

## Pretty Print JSON

Use the `indent` parameter to make JSON output more readable.

```python
import json

student = {
    "name": "Deepika",
    "age": 21,
    "skills": ["Python", "SQL", "Pandas"]
}

print(json.dumps(student, indent=4))
```

## JSON Summary

### Functions

- `dumps()` → Python Object → JSON String
- `loads()` → JSON String → Python Object
- `dump()` → Python Object → JSON File
- `load()` → JSON File → Python Object

## CSV Module

CSV (Comma-Separated Values) is a simple file format used to store tabular data. It is widely used in Excel, databases, and data analysis.

```python
import csv
```

## Write Data to a CSV File

Use `csv.writer()` to write rows into a CSV file.

```python
import csv

with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Age", "Course"])
    writer.writerow(["Deepika", 21, "AI & ML"])
    writer.writerow(["Rahul", 22, "Data Science"])

print("CSV file created successfully.")
```

## Read Data from a CSV File

Use `csv.reader()` to read data row by row.

```python
import csv

with open("students.csv", "r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

## Write Dictionary Data to a CSV File

Use `csv.DictWriter()` to write dictionaries into a CSV file.

```python
import csv

students = [
    {"Name": "Deepika", "Age": 21, "Course": "AI & ML"},
    {"Name": "Rahul", "Age": 22, "Course": "Data Science"}
]

with open("students_dict.csv", "w", newline="") as file:
    fieldnames = ["Name", "Age", "Course"]

    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(students)

print("Dictionary data written successfully.")
```

## Read Dictionary Data from a CSV File

Use `csv.DictReader()` to read CSV data as dictionaries.

```python
import csv

with open("students_dict.csv", "r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        print(row)
```

## JSON vs CSV

Both formats are used to store data, but they are designed for different purposes.

| Feature | JSON | CSV |
|---------|------|-----|
| Data Format | Structured | Tabular |
| Supports Nested Data | ✅ Yes | ❌ No |
| Human Readable | ✅ Yes | ✅ Yes |
| Common Use | APIs | Excel, Data Analysis |
