# 🧮 Project 1 : Scientific Calculator

## Overview

This project demonstrates how to organize Python code using **Modules**.

The calculator supports different mathematical operations, where each operation is stored in a separate module.

````python
## calculator.py

import math

# Addition
def add(a, b):
    return a + b


# Subtraction
def subtract(a, b):
    return a - b


# Multiplication
def multiply(a, b):
    return a * b


# Division
def divide(a, b):
    if b == 0:
        return "Division by zero is not allowed."
    return a / b


# Power
def power(a, b):
    return a ** b


# Square Root
def square_root(number):
    if number < 0:
        return "Square root of a negative number is not possible."
    return math.sqrt(number)


# Factorial
def factorial(number):
    if number < 0:
        return "Factorial of a negative number is not possible."
    return math.factorial(number)
````

````python
## main.py


import calculator

print("===== Scientific Calculator =====")

print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
print("5. Power")
print("6. Square Root")
print("7. Factorial")

choice = int(input("\nEnter your choice: "))

if choice == 1:
    num1 = float(input("Enter First Number: "))
    num2 = float(input("Enter Second Number: "))
    print("Result:", calculator.add(num1, num2))

elif choice == 2:
    num1 = float(input("Enter First Number: "))
    num2 = float(input("Enter Second Number: "))
    print("Result:", calculator.subtract(num1, num2))

elif choice == 3:
    num1 = float(input("Enter First Number: "))
    num2 = float(input("Enter Second Number: "))
    print("Result:", calculator.multiply(num1, num2))

elif choice == 4:
    num1 = float(input("Enter First Number: "))
    num2 = float(input("Enter Second Number: "))
    print("Result:", calculator.divide(num1, num2))

elif choice == 5:
    base = float(input("Enter Base: "))
    exponent = float(input("Enter Exponent: "))
    print("Result:", calculator.power(base, exponent))

elif choice == 6:
    number = float(input("Enter Number: "))
    print("Result:", calculator.square_root(number))

elif choice == 7:
    number = int(input("Enter Number: "))
    print("Result:", calculator.factorial(number))

else:
    print("Invalid Choice")
````

# bmi_package/__init__.py

## This file tells Python that bmi_package is a package.

````python
## bmi_package/calculate.py


def calculate_bmi(weight, height):
    """
    Calculate BMI using:
    BMI = Weight / (Height × Height)

    weight -> kilograms
    height -> meters
    """

    bmi = weight / (height ** 2)

    return round(bmi, 2)
````

````python
## bmi_package/category.py


def get_bmi_category(bmi):

    if bmi < 18.5:
        return "Underweight"

    elif bmi < 25:
        return "Normal Weight"

    elif bmi < 30:
        return "Overweight"

    return "Obese"
````

````python
## main.py


from bmi_package.calculate import calculate_bmi
from bmi_package.category import get_bmi_category

print("===== BMI Calculator =====")

weight = float(input("Enter Weight (kg): "))
height = float(input("Enter Height (m): "))

bmi = calculate_bmi(weight, height)

category = get_bmi_category(bmi)

print("\nBMI :", bmi)
print("Category :", category)
````

# student_package/__init__.py

## Marks student_package as a Python package.

````python
## student_package/input_data.py



def get_student_details():

    student = {
        "name": input("Enter Student Name : "),
        "roll": input("Enter Roll Number : ")
    }

    return student


def get_marks():

    marks = {
        "Python": int(input("Python Marks : ")),
        "Math": int(input("Math Marks : ")),
        "English": int(input("English Marks : ")),
        "Science": int(input("Science Marks : "))
    }

    return marks
````

````python
## student_package/calculation.py


def calculate_result(marks):

    total = sum(marks.values())

    percentage = total / len(marks)

    return total, round(percentage, 2)
````

````python
## student_package/grade.py


def calculate_grade(percentage):

    if percentage >= 90:
        return "A+"

    elif percentage >= 80:
        return "A"

    elif percentage >= 70:
        return "B"

    elif percentage >= 60:
        return "C"

    elif percentage >= 40:
        return "D"

    return "Fail"
````

````python
## student_package/report.py


def generate_report(student, marks, total, percentage, grade):

    print("\n" + "=" * 35)
    print("       STUDENT REPORT CARD")
    print("=" * 35)

    print(f"Name        : {student['name']}")
    print(f"Roll Number : {student['roll']}")

    print("\nSubject Marks")

    for subject, mark in marks.items():
        print(f"{subject:<10}: {mark}")

    print("\n-----------------------------")

    print(f"Total       : {total}")
    print(f"Percentage  : {percentage}%")
    print(f"Grade       : {grade}")

    print("=" * 35)
````

````python
## main.py




from student_package.input_data import get_student_details
from student_package.input_data import get_marks

from student_package.calculation import calculate_result
from student_package.grade import calculate_grade
from student_package.report import generate_report


print("===== Student Report Card =====")

student = get_student_details()

marks = get_marks()

total, percentage = calculate_result(marks)

grade = calculate_grade(percentage)

generate_report(student, marks, total, percentage, grade)
````

```
    │── view_expense.py
    │── summary.py
```

# expense_package/__init__.py


## Marks expense_package as a Python package.

````python
## expense_package/add_expense.py


def add_expense(expenses):

    category = input("Enter Category : ")
    amount = float(input("Enter Amount : "))

    expense = {
        "Category": category,
        "Amount": amount
    }

    expenses.append(expense)

    print("\nExpense Added Successfully.")
````

````python
## expense_package/view_expense.py



def view_expenses(expenses):

    if not expenses:
        print("\nNo Expenses Found.")
        return

    print("\n========== Expenses ==========")

    for index, expense in enumerate(expenses, start=1):

        print(f"{index}. {expense['Category']} - ₹{expense['Amount']}")
````

````python
## expense_package/summary.py


def show_summary(expenses):

    if not expenses:
        print("\nNo Expense Data Available.")
        return

    total = 0

    highest = expenses[0]

    for expense in expenses:

        total += expense["Amount"]

        if expense["Amount"] > highest["Amount"]:
            highest = expense

    print("\n========== Summary ==========")

    print("Total Expenses :", len(expenses))
    print("Total Amount   : ₹", total)
    print("Highest Expense:", highest["Category"], f"(₹{highest['Amount']})")
````

````python
## main.py


from expense_package.add_expense import add_expense
from expense_package.view_expense import view_expenses
from expense_package.summary import show_summary

expenses = []

while True:

    print("\n===== Expense Tracker =====")

    print("1. Add Expense")
    print("2. View Expenses")
    print("3. Expense Summary")
    print("4. Exit")

    choice = input("Enter Choice : ")

    if choice == "1":
        add_expense(expenses)

    elif choice == "2":
        view_expenses(expenses)

    elif choice == "3":
        show_summary(expenses)

    elif choice == "4":
        print("\nThank You!")
        break

    else:
        print("\nInvalid Choice.")
````

# inventory_package/__init__.py

## Marks inventory_package as a Python package.

````python
## inventory_package/add_product.py


def add_product(products):

    product = {
        "id": input("Product ID : "),
        "name": input("Product Name : "),
        "price": float(input("Price : ")),
        "quantity": int(input("Quantity : "))
    }

    products.append(product)

    print("\nProduct Added Successfully.")
````

````python
## inventory_package/view_products.py


def view_products(products):

    if not products:
        print("\nInventory is Empty.")
        return

    print("\n========== Inventory ==========")

    for product in products:

        print(f"""
ID       : {product["id"]}
Name     : {product["name"]}
Price    : ₹{product["price"]}
Quantity : {product["quantity"]}
-----------------------------
""")
````

````python
## inventory_package/search_product.py


def search_product(products):

    product_id = input("Enter Product ID : ")

    for product in products:

        if product["id"] == product_id:

            print("\nProduct Found")

            print(product)

            return

    print("\nProduct Not Found.")
````

````python
## inventory_package/update_product.py


def update_product(products):

    product_id = input("Enter Product ID : ")

    for product in products:

        if product["id"] == product_id:

            new_quantity = int(input("New Quantity : "))

            product["quantity"] = new_quantity

            print("\nQuantity Updated Successfully.")

            return

    print("\nProduct Not Found.")
````

````python
## inventory_package/delete_product.py


def delete_product(products):

    product_id = input("Enter Product ID : ")

    for product in products:

        if product["id"] == product_id:

            products.remove(product)

            print("\nProduct Deleted Successfully.")

            return

    print("\nProduct Not Found.")
````

````python
## inventory_package/summary.py


def inventory_summary(products):

    if not products:
        print("\nInventory is Empty.")
        return

    total_products = len(products)

    total_quantity = 0

    total_value = 0

    for product in products:

        total_quantity += product["quantity"]

        total_value += product["price"] * product["quantity"]

    print("\n========== Inventory Summary ==========")

    print("Total Products :", total_products)
    print("Total Quantity :", total_quantity)
    print("Inventory Value: ₹", total_value)

````

````python
## main.py


from inventory_package.add_product import add_product
from inventory_package.view_products import view_products
from inventory_package.search_product import search_product
from inventory_package.update_product import update_product
from inventory_package.delete_product import delete_product
from inventory_package.summary import inventory_summary

products = []

while True:

    print("\n===== Inventory Management =====")

    print("1. Add Product")
    print("2. View Products")
    print("3. Search Product")
    print("4. Update Quantity")
    print("5. Delete Product")
    print("6. Inventory Summary")
    print("7. Exit")

    choice = input("Enter Choice : ")

    if choice == "1":
        add_product(products)

    elif choice == "2":
        view_products(products)

    elif choice == "3":
        search_product(products)

    elif choice == "4":
        update_product(products)

    elif choice == "5":
        delete_product(products)

    elif choice == "6":
        inventory_summary(products)

    elif choice == "7":
        print("\nThank You!")
        break

    else:
        print("\nInvalid Choice.")
````

# movie_package/__init__.py

## Marks movie_package as a Python package.

````python
## movie_package/movies.py


movies = {
    "1": {"name": "Avengers", "price": 250, "available_seats": 50},
    "2": {"name": "Interstellar", "price": 300, "available_seats": 40},
    "3": {"name": "Pushpa 2", "price": 220, "available_seats": 60},
    "4": {"name": "Kalki 2898 AD", "price": 280, "available_seats": 45}
}

def view_movies():
    print("\n========== Movies ==========")

    for movie_id, details in movies.items():
        print(
            f"{movie_id}. {details['name']} | "
            f"Ticket: ₹{details['price']} | "
            f"Seats: {details['available_seats']}"
        )
````

````python
## movie_package/booking.py


from movie_package.movies import movies

def book_ticket():

    movie_id = input("Enter Movie ID: ")

    if movie_id not in movies:
        print("\nInvalid Movie ID.")
        return None

    seats = int(input("Number of Tickets: "))

    if seats > movies[movie_id]["available_seats"]:
        print("\nSeats Not Available.")
        return None

    movies[movie_id]["available_seats"] -= seats

    total = seats * movies[movie_id]["price"]

    booking = {
        "movie": movies[movie_id]["name"],
        "tickets": seats,
        "total": total
    }

    print("\nBooking Successful.")

    return booking
````

````python
## movie_package/seats.py

from movie_package.movies import movies

def check_seats():

    print("\nAvailable Seats")

    for details in movies.values():
        print(f"{details['name']} : {details['available_seats']}")
````

````python
## movie_package/bill.py


def generate_bill(booking):

    if booking is None:
        return

    print("\n========== Movie Ticket ==========")
    print(f"Movie          : {booking['movie']}")
    print(f"Tickets        : {booking['tickets']}")
    print(f"Total Amount   : ₹{booking['total']}")
    print("==================================")
````

````python
## main.py


from movie_package.movies import view_movies
from movie_package.booking import book_ticket
from movie_package.seats import check_seats
from movie_package.bill import generate_bill

while True:

    print("\n===== Movie Ticket Booking =====")
    print("1. View Movies")
    print("2. Book Ticket")
    print("3. Check Available Seats")
    print("4. Exit")

    choice = input("Enter Choice: ")

    if choice == "1":
        view_movies()

    elif choice == "2":
        booking = book_ticket()
        generate_bill(booking)

    elif choice == "3":
        check_seats()

    elif choice == "4":
        print("\nThank You!")
        break

    else:
        print("\nInvalid Choice.")
````
