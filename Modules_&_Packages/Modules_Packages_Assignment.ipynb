## Question 1

The `math` module provides many mathematical functions.

Write a program to calculate the **Greatest Common Divisor (GCD)** of two numbers using the `math` module.

Input Example

```

```

````python
import math

num1 = 48
num2 = 18

gcd = math.gcd(num1, num2)

print("GCD =", gcd)
````

## Question 6

Generate a **6-digit OTP** using the `random` module.

Example Output

```
483917
```

````python
import random

otp = ""

for _ in range(6):
    otp += str(random.randint(0, 9))

print("Generated OTP :", otp)
````

## Question 7

Generate a random password of **8 characters**.

Use:

- Uppercase letters
- Lowercase letters
- Digits

````python
import random
import string

characters = string.ascii_letters + string.digits

password = ""

for _ in range(8):
    password += random.choice(characters)

print("Password :", password)
````

## Question 9

Simulate tossing a coin **15 times**.

Print the total number of:

- Heads
- Tails

````python
import random

heads = 0
tails = 0

for _ in range(15):

    toss = random.choice(["Head", "Tail"])

    if toss == "Head":
        heads += 1
    else:
        tails += 1

print("Heads :", heads)
print("Tails :", tails)
````

## Question 13

### Number Guessing Game

Generate a random number between **1 and 20**.

Ask the user to guess the number until the correct answer is entered.

After every incorrect attempt, display an appropriate message.

Requirements

- Use the `random` module.
- Count the total number of attempts.
- Display a success message when the user guesses correctly.

````python
import random

secret_number = random.randint(1, 20)

attempts = 0

while True:

    guess = int(input("Enter your guess: "))

    attempts += 1

    if guess == secret_number:
        print("Correct!")
        print("Attempts:", attempts)
        break

    elif guess < secret_number:
        print("Try a Bigger Number")

    else:
        print("Try a Smaller Number")
````

## Question 15

### Random Team Generator

Create **2 teams** from the following players.

Each team should contain **4 unique players**.

```python
players = [
    "Virat",
    "Rohit",
    "Gill",
    "Hardik",
    "Rahul",
    "Surya",
    "Jadeja",
    "Bumrah"
]
```

````python
import random

players = [
    "Virat",
    "Rohit",
    "Gill",
    "Hardik",
    "Rahul",
    "Surya",
    "Jadeja",
    "Bumrah"
]

random.shuffle(players)

team_1 = players[:4]
team_2 = players[4:]

print("Team 1")
print(team_1)

print()

print("Team 2")
print(team_2)
````
