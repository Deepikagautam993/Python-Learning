# What is a Module?

A **Module** is simply a Python file (`.py`) that contains Python code.

A module can contain:

- Variables
- Functions
- Classes
- Objects

Instead of writing everything in one file, we divide our program into multiple files called **Modules**.

This makes the code:

- Easier to read
- Easier to manage
- Easier to reuse

---

## Real-Life Example

Imagine you are writing a College Management System.

Instead of writing everything in one file, you create separate files.

```
student.py
teacher.py
fees.py
attendance.py
result.py
```

Each file has a specific responsibility.

Each of these files is called a **Module**.

# Why Do We Need Modules?

Without Modules:

- One large file
- Difficult to find code
- Difficult to debug
- Difficult to reuse

With Modules:

- Code is divided into small files.
- Each file has a specific purpose.
- Code becomes reusable.
- Projects become easier to maintain.

For small programs, Modules may not seem necessary.

For large projects containing hundreds or thousands of lines of code, Modules become essential.

# import Statement

The `import` keyword is used to use a Module in our program.

## Syntax

```python
import module_name
```

After importing a module, we can access its functions and variables using the dot (`.`) operator.

````python
import math

print(math.sqrt(25))
print(math.factorial(5))
print(math.pi)
````

````python
import math

number = 64

square_root = math.sqrt(number)

print("Square Root :", square_root)
````

##  Using Multiple Functions

The same module can provide multiple functions.

````python
import math

number = 5

print("Factorial :", math.factorial(number))
print("Power :", math.pow(2, 5))
print("Square Root :", math.sqrt(81))
print("Value of PI :", math.pi)
````

````python
import random

print(random.randint(1, 10))
print(random.choice(["Python", "Java", "C++"]))
````

````python
import random

numbers = [10, 20, 30, 40, 50]

random_number = random.choice(numbers)

print(random_number)

print(random.randint(1, 100))
````

# How Does import Work?

When Python executes:

```python
import math
```

Python performs these steps:

1. Searches for the `math` module.
2. Loads the module into memory.
3. Makes all its functions and variables available.
4. The module can now be used in the program.

A module is loaded only once during program execution.

# import Statement

The `import` statement is used to load an entire module into a Python program.

After importing, we can access its functions, variables, and constants using the **dot (`.`) operator**.

## Syntax

```python
import module_name

module_name.function_name()
```

## Important Notes

- Imports the complete module.
- Access members using the dot (`.`) operator.
- Easy to understand and commonly used.
- Best choice when multiple functions from the same module are required.

## Example 1 : Using the math Module

````python
import math

print(math.sqrt(100))
print(math.factorial(6))
print(math.ceil(4.2))
print(math.floor(4.8))
````

## Example 2 : Using the random Module

````python
import random

print(random.randint(1, 50))

colors = ["Red", "Green", "Blue", "Black"]

print(random.choice(colors))
````

# from ... import

Sometimes we don't need the entire module.

We can import only the required function.

## Syntax

```python
from module_name import function_name
```

## Important Notes

- Imports only the required function.
- No need to write the module name.
- Makes the code shorter and cleaner.

## Example 1 : Importing sqrt()

````python
from math import sqrt

number = 144

print(sqrt(number))
````

## Example 2 : Importing Multiple Functions

````python
from math import sqrt, factorial

print(sqrt(49))
print(factorial(5))
````

# import vs from ... import

| import | from ... import |
|---------|-----------------|
| Imports complete module | Imports selected function(s) |
| Uses dot (`.`) operator | No dot operator required |
| Better when many functions are used | Better when only a few functions are needed |


# import ... as

Sometimes module names are long.

We can give a short name (alias) using the `as` keyword.

## Syntax

```python
import module_name as alias
```

## Important Notes

- Creates a shorter name.
- Makes code easier to write.
- Very common in Python libraries.

## Example : Using Alias

````python
import math as m

print(m.sqrt(81))
print(m.factorial(4))
print(m.pi)
````

# __name__

Every Python file has a special built-in variable called `__name__`.

Python automatically assigns a value to it.

- If the file is executed directly, `__name__` becomes `"__main__"`.
- If the file is imported into another file, `__name__` becomes the module name.

This helps us control which code should run when a file is executed directly.

## Example : Checking __name__

````python
print(__name__)
````

## Example : Using __name__ == "__main__"

This example is mainly used in Python files (`.py`) and is commonly executed in VS Code.

````python
def welcome():
    print("Welcome to Python Modules!")

if __name__ == "__main__":
    welcome()
````

# User-Defined Module

A User-Defined Module is a Python file created by the programmer.

Instead of using only built-in modules, we can create our own modules and reuse them in different programs.

## Example Folder Structure

```
Project/
│
├── calculator.py
└── main.py
```

````python
## calculator.py

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b
````

````python
## main.py

import calculator

print(calculator.add(15, 7))
print(calculator.subtract(20, 8))
````

# Important Notes

- `dir()` helps explore a module.
- `__name__` is automatically created by Python.
- `"__main__"` means the file is executed directly.
- User-defined modules are normal Python files (`.py`).
- Modules can be imported into other Python files.

# Package

A **Package** is a collection of related Python modules stored inside a folder.

A module is a single `.py` file, whereas a package is a folder containing multiple modules.

Packages help organize large Python projects into different sections.

## Syntax

```
Package/
│
├── module1.py
├── module2.py
└── __init__.py
```

## Important Notes

- Package = Folder
- Module = Python File
- One package can contain multiple modules.
- Packages improve project organization.

# Module vs Package

| Module | Package |
|---------|----------|
| Single Python file (.py) | Folder containing modules |
| Stores related code | Stores related modules |
| Easy to reuse | Better organization for large projects |

# __init__.py

`__init__.py` is a special Python file placed inside a package.

Its purpose is to tell Python that the folder should be treated as a package.

In modern Python versions, it is optional in many cases, but it is still commonly used in projects.

## Example

```
my_package/
│
├── __init__.py
├── math_operations.py
└── string_operations.py
```

````python
## math_operations.py

def square(number):
    return number * number


def cube(number):
    return number * number * number
````

````python
## string_operations.py

def to_upper(text):
    return text.upper()


def count_characters(text):
    return len(text)
````

````python
## main.py

from my_package.math_operations import square, cube
from my_package.string_operations import to_upper, count_characters

print(square(9))
print(cube(4))

print(to_upper("deepika"))
print(count_characters("Artificial Intelligence"))
````

# Package Structure

```
Project/
│
├── main.py
│
└── my_package/
    │
    ├── __init__.py
    ├── math_operations.py
    └── string_operations.py
```

This structure is commonly used in Python projects.
