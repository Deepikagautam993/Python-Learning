# 🚀 Project 1 : Login Authentication System

## Objective

Create a simple authentication system using decorators.

## Concepts Used

- Decorators
- @wraps
- Functions
- Global Variables
- Menu Driven Program

## Features

- Login
- Logout
- Dashboard
- Profile
- Authentication using Decorator

```python
from functools import wraps

# Login Status
is_logged_in = False


# Decorator
def login_required(function):

    @wraps(function)
    def wrapper(*args, **kwargs):

        global is_logged_in

        if is_logged_in:
            return function(*args, **kwargs)

        print("\n❌ Please login first.\n")

    return wrapper


# Login Function
def login():

    global is_logged_in

    username = input("Username : ")
    password = input("Password : ")

    if username == "admin" and password == "1234":

        is_logged_in = True

        print("\n✅ Login Successful\n")

    else:

        print("\n❌ Invalid Credentials\n")


# Logout Function
def logout():

    global is_logged_in

    is_logged_in = False

    print("\n👋 Logged Out Successfully\n")


# Protected Functions
@login_required
def dashboard():

    print("\n📊 Dashboard Opened\n")


@login_required
def profile():

    print("\n👤 Admin Profile")
    print("Name : Deepika")
    print("Role : Admin\n")


# Main Menu
while True:

    print("=" * 40)
    print("      LOGIN AUTHENTICATION")
    print("=" * 40)

    print("1. Login")
    print("2. Dashboard")
    print("3. Profile")
    print("4. Logout")
    print("5. Exit")

    choice = input("\nEnter Choice : ")

    if choice == "1":

        login()

    elif choice == "2":

        dashboard()

    elif choice == "3":

        profile()

    elif choice == "4":

        logout()

    elif choice == "5":

        print("\nThank You 😊")
        break

    else:

        print("\n❌ Invalid Choice\n")
```

# 🚀 Project 2 : Student Result Logger

## Objective

Create a student result system that automatically logs every result calculation.

## Concepts Used

- Decorators
- functools.wraps
- File Handling
- datetime Module

## Features

- Enter Student Name
- Enter Marks
- Calculate Percentage
- Display Grade
- Automatically Save Log in `result_log.txt`

```python
from functools import wraps
from datetime import datetime


# Decorator to save every result into a log file
def result_logger(function):

    @wraps(function)
    def wrapper(*args, **kwargs):

        result = function(*args, **kwargs)

        student_name = result["name"]
        percentage = result["percentage"]
        grade = result["grade"]

        with open("result_log.txt", "a") as file:

            file.write(
                f"{datetime.now()} | "
                f"Name: {student_name} | "
                f"Percentage: {percentage:.2f}% | "
                f"Grade: {grade}\n"
            )

        print("\n✅ Result saved successfully.\n")

        return result

    return wrapper


@result_logger
def calculate_result():

    name = input("Enter Student Name : ")

    marks = []

    for subject in range(1, 6):

        mark = float(input(f"Enter Marks of Subject {subject} : "))
        marks.append(mark)

    total = sum(marks)

    percentage = total / 5

    if percentage >= 90:
        grade = "A+"

    elif percentage >= 80:
        grade = "A"

    elif percentage >= 70:
        grade = "B"

    elif percentage >= 60:
        grade = "C"

    elif percentage >= 50:
        grade = "D"

    else:
        grade = "Fail"

    print("\n========== RESULT ==========")
    print("Student :", name)
    print("Total Marks :", total)
    print(f"Percentage : {percentage:.2f}%")
    print("Grade :", grade)

    return {
        "name": name,
        "percentage": percentage,
        "grade": grade
    }


while True:

    print("\n========== STUDENT RESULT LOGGER ==========")
    print("1. Calculate Result")
    print("2. Exit")

    choice = input("Enter Choice : ")

    if choice == "1":

        calculate_result()

    elif choice == "2":

        print("\nThank You 😊")
        break

    else:

        print("\n❌ Invalid Choice")
```

# 🚀 Project 3 : Execution Time Analyzer

## Objective

Create a decorator that measures the execution time of different functions.

## Concepts Used

- Decorators
- functools.wraps
- time Module

## Features

- Measure Function Execution Time
- Compare Different Functions
- Automatic Time Display

```python
import time
from functools import wraps


# Decorator
def execution_timer(function):

    @wraps(function)
    def wrapper(*args, **kwargs):

        start_time = time.perf_counter()

        result = function(*args, **kwargs)

        end_time = time.perf_counter()

        execution_time = end_time - start_time

        print(f"\nExecution Time : {execution_time:.6f} seconds\n")

        return result

    return wrapper


@execution_timer
def calculate_sum():

    total = 0

    for number in range(1, 1000001):

        total += number

    print("Sum :", total)


@execution_timer
def calculate_squares():

    squares = []

    for number in range(1, 50001):

        squares.append(number ** 2)

    print("Total Squares :", len(squares))


@execution_timer
def search_number():

    numbers = list(range(1, 1000001))

    target = 987654

    if target in numbers:
        print("Number Found")

    else:
        print("Number Not Found")


while True:

    print("=" * 45)
    print("      EXECUTION TIME ANALYZER")
    print("=" * 45)

    print("1. Calculate Sum")
    print("2. Generate Squares")
    print("3. Search Number")
    print("4. Exit")

    choice = input("\nEnter Choice : ")

    if choice == "1":

        calculate_sum()

    elif choice == "2":

        calculate_squares()

    elif choice == "3":

        search_number()

    elif choice == "4":

        print("\nThank You ")
        break

    else:

        print("\n Invalid Choice\n")
```

# 🚀 Project 4 : Input Validation System

## Objective

Create a system that validates user input before executing functions.

## Concepts Used

- Decorators
- functools.wraps
- Input Validation
- Functions
- Menu Driven Program

## Features

- Validate Age
- Validate Salary
- Validate Marks
- Prevent Invalid Input

```python
from functools import wraps


# Decorator
def validate_positive(function):

    @wraps(function)
    def wrapper(value):

        if value < 0:

            print("\n Value cannot be negative.\n")
            return

        return function(value)

    return wrapper


@validate_positive
def check_age(age):

    if age < 18:
        print("Minor")

    else:
        print("Adult")


@validate_positive
def check_salary(salary):

    print(f"Monthly Salary : ₹{salary:.2f}")


@validate_positive
def check_marks(marks):

    if marks > 100:

        print(" Marks cannot be greater than 100.")

    else:

        print(f"Marks : {marks}")


while True:

    print("=" * 45)
    print("      INPUT VALIDATION SYSTEM")
    print("=" * 45)

    print("1. Check Age")
    print("2. Check Salary")
    print("3. Check Marks")
    print("4. Exit")

    choice = input("\nEnter Choice : ")

    if choice == "1":

        age = int(input("Enter Age : "))
        check_age(age)

    elif choice == "2":

        salary = float(input("Enter Salary : "))
        check_salary(salary)

    elif choice == "3":

        marks = float(input("Enter Marks : "))
        check_marks(marks)

    elif choice == "4":

        print("\nThank You ")
        break

    else:

        print("\n Invalid Choice\n")
```

# 🚀 Project 5 : Role-Based Access System

## Objective

Create a system where different users can access different features based on their role.

## Concepts Used

- Decorator Factory
- Decorators
- functools.wraps
- Global Variables
- Menu Driven Program

## Features

- Login as Admin / Teacher / Student
- Role-Based Access
- Protected Functions
- Logout

```python
from functools import wraps

# Current logged-in role
current_role = None


# Decorator Factory
def role_required(role):

    def decorator(function):

        @wraps(function)
        def wrapper(*args, **kwargs):

            if current_role == role:
                return function(*args, **kwargs)

            print(f"\n Access Denied! '{role}' role is required.\n")

        return wrapper

    return decorator


# Login
def login():

    global current_role

    print("\nLogin As")
    print("1. Admin")
    print("2. Teacher")
    print("3. Student")

    choice = input("Enter Choice : ")

    if choice == "1":
        current_role = "Admin"

    elif choice == "2":
        current_role = "Teacher"

    elif choice == "3":
        current_role = "Student"

    else:
        print("Invalid Choice")
        return

    print(f"\n Logged in as {current_role}\n")


# Logout
def logout():

    global current_role

    current_role = None

    print("\n Logged Out Successfully\n")


@role_required("Admin")
def manage_users():

    print("Managing Users...")


@role_required("Teacher")
def upload_marks():

    print("Uploading Student Marks...")


@role_required("Student")
def view_result():

    print("Displaying Student Result...")


while True:

    print("=" * 45)
    print("       ROLE BASED ACCESS SYSTEM")
    print("=" * 45)

    print("1. Login")
    print("2. Manage Users (Admin)")
    print("3. Upload Marks (Teacher)")
    print("4. View Result (Student)")
    print("5. Logout")
    print("6. Exit")

    choice = input("\nEnter Choice : ")

    if choice == "1":

        login()

    elif choice == "2":

        manage_users()

    elif choice == "3":

        upload_marks()

    elif choice == "4":

        view_result()

    elif choice == "5":

        logout()

    elif choice == "6":

        print("\nThank You ")
        break

    else:

        print("\n Invalid Choice\n")
```

# 🚀 Project 6 : Function Call Logger

## Objective

Create a system that automatically logs every function call into a file.

## Concepts Used

- Decorators
- functools.wraps
- File Handling
- datetime Module
- Menu Driven Program

## Features

- Automatic Function Logging
- Store Function Name
- Store Date & Time
- Store Execution Status
- View Log File

```python
from functools import wraps
from datetime import datetime


# Decorator
def logger(function):

    @wraps(function)
    def wrapper(*args, **kwargs):

        try:
            result = function(*args, **kwargs)

            status = "Success"

        except Exception:

            status = "Failed"

            print("❌ Function Execution Failed")

            return

        with open("function_logs.txt", "a") as file:

            file.write(
                f"{datetime.now()} | "
                f"{function.__name__} | "
                f"{status}\n"
            )

        print("\n✅ Function Logged Successfully\n")

        return result

    return wrapper


@logger
def add_numbers():

    num1 = float(input("Enter First Number : "))
    num2 = float(input("Enter Second Number : "))

    print("Sum :", num1 + num2)


@logger
def find_square():

    number = int(input("Enter Number : "))

    print("Square :", number ** 2)


def view_logs():

    try:

        with open("function_logs.txt", "r") as file:

            print("\n========== FUNCTION LOGS ==========\n")

            print(file.read())

    except FileNotFoundError:

        print("\nNo Logs Found.\n")


while True:

    print("=" * 45)
    print("         FUNCTION CALL LOGGER")
    print("=" * 45)

    print("1. Add Numbers")
    print("2. Find Square")
    print("3. View Logs")
    print("4. Exit")

    choice = input("\nEnter Choice : ")

    if choice == "1":

        add_numbers()

    elif choice == "2":

        find_square()

    elif choice == "3":

        view_logs()

    elif choice == "4":

        print("\nThank You 😊")

        break

    else:

        print("\n❌ Invalid Choice\n")
```
