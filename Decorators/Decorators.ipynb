# 🎯 Decorators in Python

## 📌 Overview

A **Decorator** is a function that adds extra functionality to another function **without modifying its original code**.

Decorators follow the **DRY (Don't Repeat Yourself)** principle by allowing common functionality to be reused across multiple functions.

---

## Uses of Decorators

- Logging
- Authentication
- Authorization
- Execution Time Measurement
- Input Validation
- Caching
- Flask & Django Routes

# 📌 Decorator Flow

```
Original Function
        │
        ▼
Decorator
        │
        ▼
Modified Function
```

The original function remains unchanged.

# 📌 Basic Structure of a Decorator

```python
def decorator(function):

    def wrapper():

        # Extra Work

        function()

        # Extra Work

    return wrapper
```

### Components

- `decorator()` → Receives the original function.
- `wrapper()` → Adds extra functionality.
- `function()` → Calls the original function.
- `return wrapper` → Returns the modified function.

````python
def decorator(function):

    def wrapper():

        print("Function Started")

        function()

        print("Function Finished")

    return wrapper


def greet():

    print("Hello Python")


greet = decorator(greet)

greet()
````

# 📌 Output Flow

```
greet()

↓

wrapper()

↓

Function Started

↓

Original Function

↓

Function Finished
```

````python
def decorator(function):

    def wrapper():

        print("-" * 30)

        function()

        print("-" * 30)

    return wrapper


def display():

    print("Learning Decorators")


display = decorator(display)

display()
````

# 📌 Using `@` Syntax

Instead of:

```python
function = decorator(function)
```

Python provides a cleaner syntax.

````python
def decorator(function):

    def wrapper():

        print("Start")

        function()

        print("End")

    return wrapper


@decorator
def message():

    print("Welcome")


message()
````

# 📌 Why Use `@`?

Both approaches are exactly the same.

```python
@decorator
def greet():
    ...
```

is equivalent to:

```python
greet = decorator(greet)
```

The `@` syntax is shorter and easier to read.

````python
def border(function):

    def wrapper():

        print("=" * 40)

        function()

        print("=" * 40)

    return wrapper


@border
def student():

    print("Student Record")


student()
````

# 📌 Decorators with Parameters

The previous decorator worked only for functions without arguments.

To support any number of positional and keyword arguments, use:

- `*args`
- `**kwargs`

This makes the decorator reusable for different functions.

````python
def decorator(function):

    def wrapper(*args, **kwargs):

        print("Function Started")

        function(*args, **kwargs)

        print("Function Finished")

    return wrapper


@decorator
def greet(name):

    print(f"Hello {name}")


greet("Deepika")
````

````python
def decorator(function):

    def wrapper(*args, **kwargs):

        print("Checking...")

        function(*args, **kwargs)

        print("Done")

    return wrapper


@decorator
def add(a, b):

    print("Sum =", a + b)


add(25, 15)
````

# 📌 Why `*args` and `**kwargs`?

Without them:

```python
def wrapper():
```

Only functions without parameters can be decorated.

With them:

```python
def wrapper(*args, **kwargs):
```

The same decorator can work with any function, regardless of the number of arguments.

# 📌 Returning Values from Decorators

If the original function returns a value, the decorator should also return it.

Otherwise, the return value will be lost.

````python
def decorator(function):

    def wrapper(*args, **kwargs):

        print("Calculating...")

        result = function(*args, **kwargs)

        print("Calculation Completed")

        return result

    return wrapper


@decorator
def multiply(a, b):

    return a * b


answer = multiply(6, 8)

print(answer)
````

# 📌 Decorator Execution Flow

```
Function Call

↓

Decorator

↓

Wrapper

↓

Original Function

↓

Return Result

↓

Wrapper Returns Result
```

# 📌 Multiple Decorators

More than one decorator can be applied to the same function.

Execution starts from the decorator nearest to the function and then moves outward.

````python
def stars(function):

    def wrapper():

        print("*" * 30)

        function()

        print("*" * 30)

    return wrapper


def hash_line(function):

    def wrapper():

        print("#" * 30)

        function()

        print("#" * 30)

    return wrapper


@stars
@hash_line
def welcome():

    print("Welcome to Python")


welcome()
````

# 📌 Order of Execution

```python
@A
@B
def demo():
    ...
```

Equivalent to:

```python
demo = A(B(demo))
```

Execution Order:

```
A

↓

B

↓

Original Function

↓

B

↓

A
```

````python
def uppercase(function):

    def wrapper():

        text = function()

        return text.upper()

    return wrapper


@uppercase
def language():

    return "python"


print(language())
````

# 📌 Important Notes

- Always use `*args` and `**kwargs` in reusable decorators.
- Return the original function's result if it returns a value.
- Multiple decorators execute from the bottom upward during wrapping.
- The `@` syntax improves readability.

# 📌 Preserving Function Metadata (`functools.wraps`)

When a function is decorated, the wrapper function replaces the original function.

As a result, information like the function name and documentation is lost.

Python provides `functools.wraps` to preserve the original function's metadata.

````python
from functools import wraps


def decorator(function):

    @wraps(function)
    def wrapper(*args, **kwargs):

        print("Executing Function")

        return function(*args, **kwargs)

    return wrapper


@decorator
def greet():

    """Greeting Function"""

    print("Hello")


print(greet.__name__)

print(greet.__doc__)
````

# 📌 Why Use `wraps`?

Without `wraps`

```python
print(greet.__name__)
```

Output

```
wrapper
```

With `wraps`

```
greet
```

Always use `@wraps` while creating reusable decorators.

# 📌 Built-in Decorators

Python already provides several decorators.

Most commonly used:

- `@staticmethod`
- `@classmethod`
- `@property`

## `@staticmethod`

A static method belongs to the class but does not use `self` or `cls`.

It behaves like a normal function placed inside a class.

````python
class Calculator:

    @staticmethod
    def add(a, b):

        return a + b


print(Calculator.add(15, 30))
````

## `@classmethod`

A class method receives the class (`cls`) as its first parameter.

It is commonly used to work with class variables or create alternative constructors.

````python
class Student:

    school = "ABC Public School"

    @classmethod
    def show_school(cls):

        print(cls.school)


Student.show_school()
````

## `@property`

The `@property` decorator allows a method to be accessed like an attribute.

This provides a clean interface while keeping the implementation inside a method.

````python
class Rectangle:

    def __init__(self, length, width):

        self.length = length
        self.width = width

    @property
    def area(self):

        return self.length * self.width


box = Rectangle(8, 5)

print(box.area)
````

# 📌 Real-World Use Cases of Decorators

Decorators are widely used for:

- User Authentication
- Permission Checking
- Logging
- Performance Measurement
- Input Validation
- Caching
- API Rate Limiting

````python
from functools import wraps


def login_required(function):

    @wraps(function)
    def wrapper(*args, **kwargs):

        is_logged_in = True

        if is_logged_in:

            return function(*args, **kwargs)

        print("Login Required")

    return wrapper


@login_required
def dashboard():

    print("Welcome to Dashboard")


dashboard()
````

````python
import time

from functools import wraps


def execution_time(function):

    @wraps(function)
    def wrapper(*args, **kwargs):

        start = time.time()

        result = function(*args, **kwargs)

        end = time.time()

        print(f"Execution Time: {end - start:.6f} seconds")

        return result

    return wrapper


@execution_time
def calculate():

    total = 0

    for i in range(1000000):

        total += i

    return total


calculate()
````

# 📌 Decorator Factory (Decorator with Arguments)

Until now, we have passed only the function to a decorator.

Sometimes we also want to pass our own values to the decorator.

Examples:

- Repeat a function 3 times
- Set a custom execution limit
- Pass a custom message
- Set user roles

In such cases, we use a **Decorator Factory**.

A Decorator Factory is simply a function that returns a decorator.

Structure:

Decorator Factory

↓

Decorator

↓

Wrapper

↓

Original Function

````python
def repeat(times):

    def decorator(function):

        def wrapper(*args, **kwargs):

            for _ in range(times):

                function(*args, **kwargs)

        return wrapper

    return decorator


@repeat(3)
def hello():

    print("Hello Python")


hello()
````

## Execution Flow

```
repeat(3)

↓

decorator()

↓

wrapper()

↓

hello()
```

The value `3` is stored inside the decorator and used every time the function is called.

````python
def show_message(message):

    def decorator(function):

        def wrapper(*args, **kwargs):

            print(message)

            function(*args, **kwargs)

        return wrapper

    return decorator


@show_message("Welcome to Student Portal")
def dashboard():

    print("Dashboard Opened")


dashboard()
````

````python
def multiply_output(number):

    def decorator(function):

        def wrapper(*args, **kwargs):

            result = function(*args, **kwargs)

            return result * number

        return wrapper

    return decorator


@multiply_output(5)
def get_value():

    return 10


print(get_value())
````

# 📌 Real-World Example

Authentication based on user role.

```
@role_required("Admin")

↓

Check User Role

↓

Access Granted / Access Denied

↓

Original Function
```

Many web frameworks use this pattern.

````python
from functools import wraps


def role_required(role):

    def decorator(function):

        @wraps(function)
        def wrapper(*args, **kwargs):

            current_role = "Admin"

            if current_role == role:

                return function(*args, **kwargs)

            print("Access Denied")

        return wrapper

    return decorator


@role_required("Admin")
def delete_record():

    print("Record Deleted")


delete_record()
````
