### Question 1

Create a decorator that prints:

```
Program Started
```

before executing a function.

Decorate a function that prints:

```
Learning Python Decorators
```

````python
def start_message(function):

    def wrapper():

        print("Program Started")

        function()

    return wrapper


@start_message
def learn():

    print("Learning Python Decorators")


learn()
````

### Question 3

Create a decorator that prints:

```
Executing...
```

before calling a function that accepts two numbers.

Call the decorated function with different values.

````python
def execute(function):

    def wrapper(*args, **kwargs):

        print("Executing...")

        function(*args, **kwargs)

    return wrapper


@execute
def subtract(a, b):

    print("Difference =", a - b)


subtract(45, 18)
subtract(100, 25)
````

### Question 6

Create a decorator that calculates and prints the execution time of a function.

Decorate a function that prints numbers from 1 to 100000.

````python
import time

def timer(function):

    def wrapper(*args, **kwargs):

        start = time.time()

        function(*args, **kwargs)

        end = time.time()

        print(f"\nExecution Time: {end - start:.6f} seconds")

    return wrapper


@timer
def print_numbers():

    for number in range(1, 100001):
        pass


print_numbers()
````

### Question 9

Create a decorator factory named `repeat(times)`.

The decorated function should execute according to the value of `times`.

Example:

```python
@repeat(4)
```

The function should run 4 times.

````python
def repeat(times):

    def decorator(function):

        def wrapper(*args, **kwargs):

            for _ in range(times):
                function(*args, **kwargs)

        return wrapper

    return decorator


@repeat(4)
def welcome():

    print("Welcome Student")


welcome()
````

### Question 10

Create a decorator that allows a function to execute only if the user is logged in.

If the user is not logged in, print:

```
Please Login First
```

````python
from functools import wraps


def login_required(function):

    @wraps(function)
    def wrapper(*args, **kwargs):

        logged_in = False

        if logged_in:
            return function(*args, **kwargs)

        print("Please Login First")

    return wrapper


@login_required
def profile():

    print("Welcome to Your Profile")


profile()
````

### Question 12

Create a decorator that counts how many times a function has been called.

Print the count before executing the function.

````python
from functools import wraps


def call_counter(function):

    count = 0

    @wraps(function)
    def wrapper(*args, **kwargs):

        nonlocal count

        count += 1

        print(f"Function Called: {count} time(s)")

        return function(*args, **kwargs)

    return wrapper


@call_counter
def greet():

    print("Hello!")


greet()
greet()
greet()
````

### Question 13

Create a decorator factory named `role_required(role)`.

Only users with the required role should be allowed to execute the function.

If the role does not match, print:

```
Access Denied
```

Otherwise, execute the function.

````python
from functools import wraps


def role_required(role):

    def decorator(function):

        @wraps(function)
        def wrapper(*args, **kwargs):

            current_role = "Admin"

            if current_role == role:
                return function(*args, **kwargs)

            print("Access Denied")

        return wrapper

    return decorator


@role_required("Admin")
def delete_user():

    print("User Deleted Successfully")


delete_user()
````

### Question 14

Create a decorator named `retry(max_attempts)`.

If the decorated function raises an exception, retry it until the maximum number of attempts is reached.

If all attempts fail, print:

```
Operation Failed
```

````python
from functools import wraps


def retry(max_attempts):

    def decorator(function):

        @wraps(function)
        def wrapper(*args, **kwargs):

            for attempt in range(1, max_attempts + 1):

                try:
                    return function(*args, **kwargs)

                except Exception:

                    print(f"Attempt {attempt} Failed")

            print("Operation Failed")

        return wrapper

    return decorator


counter = 0


@retry(3)
def connect():

    global counter

    counter += 1

    if counter < 3:
        raise ConnectionError("Connection Error")

    print("Connected Successfully")


connect()
````

### Bonus Question

Create a decorator that logs every function call into a file named `log.txt`.

Each log entry should contain:

- Function Name
- Date & Time
- Status (Success)

Use the `with` statement to write the log.

````python
from functools import wraps
from datetime import datetime


def logger(function):

    @wraps(function)
    def wrapper(*args, **kwargs):

        result = function(*args, **kwargs)

        with open("log.txt", "a") as file:

            file.write(
                f"{datetime.now()} | {function.__name__} | Success\n"
            )

        return result

    return wrapper


@logger
def calculate():

    print("Calculation Completed")


calculate()
````
