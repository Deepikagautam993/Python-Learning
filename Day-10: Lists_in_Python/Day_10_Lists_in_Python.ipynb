## 📌 What is a List?

A list is a linear data structure in Python used to store multiple values in a single variable.

Lists are:
- Ordered
- Mutable
- Allow duplicate values
- Can store different data types

```python
# Creating a list

numbers = [10, 20, 30, 40]

print(numbers)
```

```python
# List can store different data types

data = [10, "Python", 5.5, True]

print(data)
```

## Indexing

List index starts from 0.

Example:

0 → first element  
1 → second element  

```python
#Positive Indexing

numbers = [10,20,30,40]

print(numbers[0])
print(numbers[2])
```

```python
#Negative Indexing

numbers = [10,20,30,40]

print(numbers[-1])
```

## Slicing

Syntax:

list[start:end]

Start index included
End index excluded

```python
numbers = [10,20,30,40,50]

print(numbers[1:4])
print(numbers[:3])
print(numbers[2:])
print(numbers[::-1])
```

## Adding Elements

append()
→ add at end

insert()
→ add at specific index

extend()
→ add multiple elements

```python
#append()

numbers = [10,20,30]

numbers.append(40)

print(numbers)
```

```python
#insert()
numbers = [10,20,30]

numbers.insert(1,15)

print(numbers)
```

```python
#extend()

numbers = [10,20]

numbers.extend([30,40])

print(numbers)
```

## Removing Elements

remove()
→ delete by value

pop()
→ delete by index

del
→ delete index

clear()
→ empty list

```python
numbers = [10,20,30]

numbers.remove(20)

print(numbers)
```

```python
numbers = [10,20,30]

value = numbers.pop()

print(numbers)
print(value)
```

```python
numbers = [10,20,30]

del numbers[0]

print(numbers)
```

```python
numbers = [10,20,30]

numbers.clear()

print(numbers)
```

# 🎯 Practice Questions

1. Create a list of 5 numbers and print it.

2. Find sum of all elements.

3. Find largest element.

4. Reverse list without reverse().

5. Count frequency of a number.

```python
# 1. Create list

numbers = [5,10,15,20,25]

print(numbers)


# 2. Sum

total = 0

for i in numbers:
    total += i

print(total)


# 3. Largest element

largest = numbers[0]

for i in numbers:
    if i > largest:
        largest = i

print(largest)


# 4. Reverse without reverse()

print(numbers[::-1])


# 5. Count frequency

nums = [1,2,3,2,2,4]

count = 0

for i in nums:
    if i == 2:
        count += 1

print(count)
```
