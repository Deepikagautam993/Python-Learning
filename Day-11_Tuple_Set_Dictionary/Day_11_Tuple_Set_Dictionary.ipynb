## 📌 Tuple

Tuple is an ordered collection that stores multiple values.

Properties:

- Ordered
- Immutable
- Allows duplicates
- Supports indexing

```python
# Creating tuple

numbers = (10,20,30,40)

print(numbers)
```

```python
data = (10,"Python",5.5,True)

print(data)
```

```python
a = (10,)

print(type(a))
```

```python
t = (10,20,30,40)

print(t[0])
print(t[-1])
```

```python
t = (10,20,30,40,50)

print(t[1:4])
```

## 📌 Set

Set is a collection that stores only unique values.

Properties:

- Mutable
- Unordered
- No duplicate values

```python
numbers = {10,20,30,10}

print(numbers)
```

```python
a = set()

print(type(a))
```

```python
s = {10,20}

s.add(30)

print(s)
```

```python
s = {10,20,30}

s.remove(20)

print(s)
```

```python
a = {1,2,3}
b = {3,4,5}

print(a | b)
print(a & b)
```

## 📌 Dictionary

Dictionary stores data in key-value pairs.

Example:

name → Deepika
age → 20

```python
student = {
    "name":"Deepika",
    "course":"AIML",
    "age":20
}

print(student)
```

```python
student = {
    "name":"Deepika",
    "age":20
}

print(student["name"])
```

```python
student = {
    "name":"Deepika",
    "age":20
}


student["city"]="Jabalpur"

student["age"]=21

print(student)
```

```python
student = {
    "name":"Deepika",
    "age":20
}

student.pop("age")

print(student)
```

# List vs Tuple vs Set vs Dictionary

| Structure | Mutable | Ordered | Duplicate |
|---|---|---|---|
| List | Yes | Yes | Yes |
| Tuple | No | Yes | Yes |
| Set | Yes | No | No |
| Dictionary | Yes | Yes | Keys No |

## Practice Questions

1. Create a tuple and access elements.

2. Remove duplicates from a list using set.

3. Create dictionary of student details.

4. Count frequency of elements using dictionary.

5. Find common elements between two sets.

```python
# 1 Tuple

t = (10,20,30)

print(t[1])


# 2 Remove duplicates

nums = [1,2,2,3,3,4]

unique = set(nums)

print(unique)


# 3 Dictionary

student = {
    "name":"Deepika",
    "branch":"AIML"
}

print(student)


# 4 Frequency count

nums = [1,2,2,3,3,3]

freq = {}

for i in nums:
    freq[i] = freq.get(i,0)+1

print(freq)


# 5 Common elements

a = {1,2,3}
b = {3,4,5}

print(a & b)
```
