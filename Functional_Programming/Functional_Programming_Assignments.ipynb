### Question 1

Use `map()` to convert the following temperatures from Celsius to Kelvin.

```python
temperatures = [18, 25, 32, 40]
```

Formula:

```
Kelvin = Celsius + 273.15
```

````python
temperatures = [18, 25, 32, 40]

kelvin = list(
    map(
        lambda temp: temp + 273.15,
        temperatures
    )
)

print(kelvin)
````


### Question 6

Use `reduce()` to find the minimum number.

```python
numbers = [48, 17, 62, 9, 31]
```


````python
from functools import reduce

numbers = [48, 17, 62, 9, 31]

minimum = reduce(
    lambda x, y: x if x < y else y,
    numbers
)

print(minimum)
````

### Question 13

Calculate the total amount for each product using `map()`.

```python
prices = [120, 350, 80, 150]

quantities = [3, 2, 5, 4]
```

Formula:

```
Total = Price × Quantity
```

````python
prices = [120, 350, 80, 150]

quantities = [3, 2, 5, 4]

total_amount = list(
    map(
        lambda price, quantity: price * quantity,
        prices,
        quantities
    )
)

print(total_amount)
````

### Question 15

### Question 16

From the following marks:

```python
marks = [28, 45, 67, 31, 82, 19]
```

- Select only passing marks (`>= 33`) using `filter()`.
- Add **5 grace marks** to each passing mark using `map()`.

````python
marks = [28, 45, 67, 31, 82, 19]

passing_marks = filter(
    lambda mark: mark >= 33,
    marks
)

updated_marks = list(
    map(
        lambda mark: mark + 5,
        passing_marks
    )
)

print(updated_marks)
````

### Question 18

From the following list of numbers:

```python
numbers = [5, 12, 7, 18, 25, 30]
```

1. Select only even numbers using `filter()`.
2. Multiply each selected number by **10** using `map()`.
3. Find the total sum using `reduce()`.

````python
from functools import reduce

numbers = [5, 12, 7, 18, 25, 30]

even_numbers = filter(
    lambda number: number % 2 == 0,
    numbers
)

updated_numbers = map(
    lambda number: number * 10,
    even_numbers
)

result = reduce(
    lambda x, y: x + y,
    updated_numbers
)

print(result)
````
