# 🚀 Project 1 : Student Result Analyzer

## 🎯 Objective

Create a program that analyzes student marks using Functional Programming concepts.

## 📚 Concepts Used

- map()
- filter()
- max()
- min()
- any()
- all()
- lambda

```python
# ================================
# Project 1 : Student Result Analyzer
# ================================

students = [
    {"name": "Aarav", "marks": 92},
    {"name": "Diya", "marks": 48},
    {"name": "Kabir", "marks": 76},
    {"name": "Meera", "marks": 31},
    {"name": "Rohan", "marks": 85}
]


# ================================
# Calculate Percentage
# ================================

percentages = list(
    map(
        lambda student: {
            "name": student["name"],
            "percentage": student["marks"]
        },
        students
    )
)

print("\nStudent Percentages")

for student in percentages:
    print(student)


# ================================
# Passed Students
# ================================

passed_students = list(
    filter(
        lambda student: student["marks"] >= 33,
        students
    )
)

print("\nPassed Students")

for student in passed_students:
    print(student)


# ================================
# Failed Students
# ================================

failed_students = list(
    filter(
        lambda student: student["marks"] < 33,
        students
    )
)

print("\nFailed Students")

for student in failed_students:
    print(student)


# ================================
# Highest Marks
# ================================

highest = max(
    students,
    key=lambda student: student["marks"]
)

print("\nHighest Marks")

print(highest)


# ================================
# Lowest Marks
# ================================

lowest = min(
    students,
    key=lambda student: student["marks"]
)

print("\nLowest Marks")

print(lowest)


# ================================
# Check if Everyone Passed
# ================================

everyone_passed = all(
    student["marks"] >= 33
    for student in students
)

print("\nEveryone Passed :", everyone_passed)


# ================================
# Check if Anyone Scored Above 90
# ================================

topper_exists = any(
    student["marks"] > 90
    for student in students
)

print("Any Student Above 90 :", topper_exists)



# ================================
# Class Statistics
# ================================

total_students = len(students)

average_marks = sum(
    student["marks"] for student in students
) / total_students

passed_count = len(passed_students)

failed_count = len(failed_students)

pass_percentage = (passed_count / total_students) * 100



print("\n===============================")
print("Class Statistics")
print("===============================")

print("Total Students :", total_students)

print("Average Marks :", round(average_marks, 2))

print("Passed Students :", passed_count)

print("Failed Students :", failed_count)

print("Pass Percentage :", round(pass_percentage, 2), "%")
```

# 🚀 Project 2 : Employee Salary Processor

## 🎯 Objective

Process employee salary data using Functional Programming concepts.

## 📚 Concepts Used

- map()
- filter()
- reduce()
- max()
- min()
- lambda

```python
from functools import reduce

# ================================
# Project 2 : Employee Salary Processor
# ================================

employees = [

    {"name": "Aman", "salary": 42000},
    {"name": "Riya", "salary": 68000},
    {"name": "Neha", "salary": 52000},
    {"name": "Karan", "salary": 39000},
    {"name": "Ishita", "salary": 76000}

]


# ================================
# 10% Salary Increment
# ================================

updated_salary = list(
    map(
        lambda employee: {
            "name": employee["name"],
            "salary": round(employee["salary"] * 1.10)
        },
        employees
    )
)

print("\nUpdated Salaries")

for employee in updated_salary:
    print(employee)


# ================================
# Employees with Salary >= 50000
# ================================

high_salary = list(
    filter(
        lambda employee: employee["salary"] >= 50000,
        employees
    )
)

print("\nHigh Salary Employees")

for employee in high_salary:
    print(employee)


# ================================
# Highest Salary
# ================================

highest_salary = max(
    employees,
    key=lambda employee: employee["salary"]
)

print("\nHighest Salary")

print(highest_salary)


# ================================
# Lowest Salary
# ================================

lowest_salary = min(
    employees,
    key=lambda employee: employee["salary"]
)

print("\nLowest Salary")

print(lowest_salary)


# ================================
# Total Salary Expense
# ================================

total_salary = reduce(
    lambda total, employee: total + employee["salary"],
    employees,
    0
)

print("\nTotal Salary Expense :", total_salary)


# ================================
# Average Salary
# ================================

average_salary = total_salary / len(employees)

print("Average Salary :", round(average_salary, 2))
```

# 🚀 Project 3 : Sales Report Generator

## 🎯 Objective

Generate a sales report using Functional Programming concepts.

## 📚 Concepts Used

- zip()
- map()
- reduce()
- max()
- min()
- lambda

```python
from functools import reduce

# ================================
# Project 3 : Sales Report Generator
# ================================

products = [
    "Laptop",
    "Mouse",
    "Keyboard",
    "Monitor",
    "Printer"
]

prices = [55000, 800, 2500, 12000, 9500]

quantities = [3, 15, 8, 4, 2]


# ================================
# Combine Product Data
# ================================

sales_data = list(
    zip(products, prices, quantities)
)

print("\nSales Data")

for record in sales_data:
    print(record)


# ================================
# Calculate Revenue
# ================================

revenue = list(
    map(
        lambda item: {
            "product": item[0],
            "revenue": item[1] * item[2]
        },
        sales_data
    )
)

print("\nRevenue Report")

for item in revenue:
    print(item)


# ================================
# Total Revenue
# ================================

total_revenue = reduce(
    lambda total, item: total + item["revenue"],
    revenue,
    0
)

print("\nTotal Revenue :", total_revenue)


# ================================
# Highest Revenue Product
# ================================

highest = max(
    revenue,
    key=lambda item: item["revenue"]
)

print("\nHighest Revenue Product")

print(highest)


# ================================
# Lowest Revenue Product
# ================================

lowest = min(
    revenue,
    key=lambda item: item["revenue"]
)

print("\nLowest Revenue Product")

print(lowest)


# ================================
# Sales Statistics
# ================================

average_revenue = total_revenue / len(revenue)

print("\n===============================")
print("Sales Statistics")
print("===============================")

print("Total Products :", len(products))
print("Average Revenue :", round(average_revenue, 2))
```

# 🚀 Project 4 : Attendance Analyzer

## 🎯 Objective

Analyze student attendance using Functional Programming concepts.

## 📚 Concepts Used

- enumerate()
- any()
- all()
- filter()
- lambda

```python
# ================================
# Project 4 : Attendance Analyzer
# ================================

students = [
    {"name": "Aarav", "present": True},
    {"name": "Diya", "present": False},
    {"name": "Kabir", "present": True},
    {"name": "Meera", "present": True},
    {"name": "Rohan", "present": False}
]

# ================================
# Attendance List
# ================================

print("\nAttendance Register")

for roll_no, student in enumerate(students, start=1):
    status = "Present" if student["present"] else "Absent"
    print(f"{roll_no}. {student['name']} - {status}")


# ================================
# Present Students
# ================================

present_students = list(
    filter(
        lambda student: student["present"],
        students
    )
)

print("\nPresent Students")

for student in present_students:
    print(student["name"])


# ================================
# Absent Students
# ================================

absent_students = list(
    filter(
        lambda student: not student["present"],
        students
    )
)

print("\nAbsent Students")

for student in absent_students:
    print(student["name"])


# ================================
# Attendance Statistics
# ================================

total_students = len(students)

present_count = len(present_students)

absent_count = len(absent_students)

attendance_percentage = (
    present_count / total_students
) * 100

print("\n===============================")
print("Attendance Statistics")
print("===============================")

print("Total Students :", total_students)
print("Present :", present_count)
print("Absent :", absent_count)
print("Attendance Percentage :", round(attendance_percentage, 2), "%")


# ================================
# Check Attendance Status
# ================================

everyone_present = all(
    student["present"] for student in students
)

anyone_absent = any(
    not student["present"] for student in students
)

print("\nEveryone Present :", everyone_present)
print("Anyone Absent :", anyone_absent)
```

# 🚀 Project 5 : Product Inventory Analyzer

## 🎯 Objective

Analyze product inventory and generate stock reports using Functional Programming.

## 📚 Concepts Used

- zip()
- map()
- filter()
- max()
- min()
- lambda

```python
# ================================
# Project 5 : Product Inventory Analyzer
# ================================

products = [
    "Laptop",
    "Mouse",
    "Keyboard",
    "Monitor",
    "Printer"
]

prices = [55000, 800, 2500, 12000, 9500]

stock = [12, 4, 25, 7, 2]


# ================================
# Combine Product Data
# ================================

inventory = list(
    zip(products, prices, stock)
)

print("\nInventory Report")

for item in inventory:
    print(item)


# ================================
# Update Prices (10% Discount)
# ================================

updated_prices = list(
    map(
        lambda item: {
            "product": item[0],
            "new_price": round(item[1] * 0.90, 2),
            "stock": item[2]
        },
        inventory
    )
)

print("\nUpdated Prices")

for item in updated_prices:
    print(item)


# ================================
# Low Stock Products (< 5)
# ================================

low_stock = list(
    filter(
        lambda item: item[2] < 5,
        inventory
    )
)

print("\nLow Stock Products")

for item in low_stock:
    print(item)


# ================================
# Most Expensive Product
# ================================

highest_price = max(
    inventory,
    key=lambda item: item[1]
)

print("\nMost Expensive Product")

print(highest_price)


# ================================
# Least Expensive Product
# ================================

lowest_price = min(
    inventory,
    key=lambda item: item[1]
)

print("\nLeast Expensive Product")

print(lowest_price)


# ================================
# Inventory Statistics
# ================================

total_products = len(products)

total_stock = sum(stock)

average_price = sum(prices) / total_products

print("\n===============================")
print("Inventory Statistics")
print("===============================")

print("Total Products :", total_products)
print("Total Stock :", total_stock)
print("Average Price :", round(average_price, 2))
```

# 🚀 Project 6 : Student Grade Generator

## 🎯 Objective

Generate grades and analyze student performance using Functional Programming.

## 📚 Concepts Used

- map()
- filter()
- enumerate()
- any()
- all()
- lambda

```python
# ================================
# Project 6 : Student Grade Generator
# ================================

students = [
    {"name": "Aarav", "marks": 95},
    {"name": "Diya", "marks": 82},
    {"name": "Kabir", "marks": 68},
    {"name": "Meera", "marks": 54},
    {"name": "Rohan", "marks": 27}
]


# ================================
# Generate Grades
# ================================

graded_students = list(
    map(
        lambda student: {
            "name": student["name"],
            "marks": student["marks"],
            "grade":
                "A+" if student["marks"] >= 90 else
                "A" if student["marks"] >= 80 else
                "B" if student["marks"] >= 70 else
                "C" if student["marks"] >= 60 else
                "D" if student["marks"] >= 33 else
                "F"
        },
        students
    )
)


# ================================
# Grade Report
# ================================

print("\nStudent Grade Report\n")

for roll_no, student in enumerate(graded_students, start=1):

    print(
        f"{roll_no}. "
        f"{student['name']} | "
        f"Marks: {student['marks']} | "
        f"Grade: {student['grade']}"
    )


# ================================
# Passed Students
# ================================

passed_students = list(
    filter(
        lambda student: student["marks"] >= 33,
        graded_students
    )
)

print("\nPassed Students")

for student in passed_students:
    print(student["name"])


# ================================
# Failed Students
# ================================

failed_students = list(
    filter(
        lambda student: student["marks"] < 33,
        graded_students
    )
)

print("\nFailed Students")

for student in failed_students:
    print(student["name"])


# ================================
# Performance Check
# ================================

all_passed = all(
    student["marks"] >= 33
    for student in graded_students
)

any_topper = any(
    student["marks"] >= 90
    for student in graded_students
)

print("\nEveryone Passed :", all_passed)
print("Any Topper (90+) :", any_topper)


# ================================
# Class Statistics
# ================================

total_students = len(graded_students)

passed_count = len(passed_students)

failed_count = len(failed_students)

average_marks = (
    sum(student["marks"] for student in graded_students)
    / total_students
)

print("\n===============================")
print("Class Statistics")
print("===============================")

print("Total Students :", total_students)
print("Passed :", passed_count)
print("Failed :", failed_count)
print("Average Marks :", round(average_marks, 2))
```
