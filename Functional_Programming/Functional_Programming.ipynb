# 🚀 Functional Programming in Python

## Overview

Functional Programming is a programming style that focuses on applying functions to data instead of modifying data directly.

Python provides several built-in functions that make data processing concise and readable.

## Topics Covered

- `map()`
- `filter()`
- `reduce()`
- `zip()`
- `enumerate()`
- `any()`
- `all()`

## `map()`

`map()` applies a function to every element of an iterable and returns a map object.

### Syntax

```python
map(function, iterable)
```

````python
numbers = [3, 6, 9, 12, 15]

result = map(lambda number: number * 3, numbers)

print(list(result))
````

````python
cities = ["delhi", "indore", "jaipur", "mumbai"]

capitalized = map(lambda city: city.title(), cities)

print(list(capitalized))
````

## Important Notes

- `map()` does not modify the original iterable.
- It returns a **map object**.
- Convert it using `list()`, `tuple()`, or another iterable when needed.

````python
numbers = [5, 10, 15]

mapped = map(lambda value: value + 5, numbers)

print(mapped)
print(list(mapped))
````

## `map()` with a Normal Function

A normal function can also be passed to `map()`.

This approach is useful when the logic is reusable or contains multiple operations.

````python
def calculate_cube(number):

    return number ** 3


numbers = [2, 4, 6, 8]

result = map(calculate_cube, numbers)

print(list(result))
````

## `map()` with Multiple Iterables

`map()` can process multiple iterables simultaneously.

The function should accept the same number of parameters as the iterables provided.

````python
price = [250, 450, 800, 1200]

quantity = [2, 1, 3, 2]

total_amount = map(
    lambda p, q: p * q,
    price,
    quantity
)

print(list(total_amount))
````

````python
first_name = ["Deepika", "Aman", "Neha"]

last_name = ["Gautam", "Sharma", "Verma"]

full_name = map(
    lambda first, last: first + " " + last,
    first_name,
    last_name
)

print(list(full_name))
````

## `map()` vs Loop

Both approaches produce the same result.

Choose the one that improves readability for your use case.

````python
numbers = [12, 18, 24, 30]

# Using Loop
result = []

for number in numbers:

    result.append(number // 6)

print(result)
````

````python
numbers = [12, 18, 24, 30]

# Using map()
result = map(
    lambda number: number // 6,
    numbers
)

print(list(result))
````

## Best Practices

- Use `map()` for simple transformations.
- Use a normal function if the logic is complex.
- Keep lambda expressions short and readable.
- Use `list()` only when a list is required.

````python
temperatures = [0, 20, 35, 40]

fahrenheit = list(
    map(
        lambda celsius: (celsius * 9 / 5) + 32,
        temperatures
    )
)

print(fahrenheit)
````

## Key Points

- `map()` transforms every element.
- It never filters or removes elements.
- The output length is equal to the input length.
- Multiple iterables are supported.

## `filter()`

`filter()` returns only those elements that satisfy a given condition.

### Syntax

```python
filter(function, iterable)
```

````python
numbers = [5, 10, 15, 20, 25, 30]

result = filter(
    lambda number: number % 10 == 0,
    numbers
)

print(list(result))
````

## `filter()` with a Normal Function

A normal function can also be passed to `filter()`.

The function must return either `True` or `False`.

````python
def is_positive(number):

    return number > 0


numbers = [-8, 15, -2, 7, 0, 19]

result = filter(is_positive, numbers)

print(list(result))
````

## Filtering Strings

`filter()` can be used with strings or any iterable object.

````python
languages = [
    "Python",
    "C",
    "Java",
    "Go",
    "JavaScript"
]

result = filter(
    lambda language: len(language) > 4,
    languages
)

print(list(result))
````

## Filtering Dictionaries

`filter()` is commonly used to select records that satisfy a condition.

````python
employees = [

    {"name": "Aman", "salary": 45000},
    {"name": "Riya", "salary": 72000},
    {"name": "Neha", "salary": 58000},
    {"name": "Karan", "salary": 39000}

]

high_salary = filter(
    lambda employee: employee["salary"] >= 50000,
    employees
)

print(list(high_salary))
````

## `filter()` vs `map()`

| `map()` | `filter()` |
|----------|------------|
| Transforms every element | Selects elements based on a condition |
| Output size is usually the same as input | Output size may be smaller |
| Function returns a new value | Function returns `True` or `False` |

````python
numbers = [3, 6, 9, 12]

# map()
mapped = list(
    map(
        lambda number: number * 5,
        numbers
    )
)

# filter()
filtered = list(
    filter(
        lambda number: number > 6,
        numbers
    )
)

print("Map :", mapped)
print("Filter :", filtered)
````

## `reduce()`

`reduce()` repeatedly applies a function to the elements of an iterable and returns a **single final value**.

### Import

```python
from functools import reduce
```

### Syntax

```python
reduce(function, iterable)
```

````python
from functools import reduce

numbers = [5, 10, 15, 20]

result = reduce(
    lambda x, y: x + y,
    numbers
)

print(result)
````

## How `reduce()` Works

`reduce()` processes the iterable two elements at a time.

Example:

```text
[5, 10, 15, 20]

Step 1 → 5 + 10 = 15
Step 2 → 15 + 15 = 30
Step 3 → 30 + 20 = 50

Final Result → 50
```

````python
from functools import reduce

numbers = [2, 3, 4, 5]

product = reduce(
    lambda x, y: x * y,
    numbers
)

print(product)
````

````python
from functools import reduce

numbers = [12, 45, 8, 91, 27]

maximum = reduce(
    lambda x, y: x if x > y else y,
    numbers
)

print(maximum)
````

## `reduce()` with Initial Value

An initial value can be provided as the third argument.

### Syntax

```python
reduce(function, iterable, initial_value)
```

````python
from functools import reduce

numbers = [1, 2, 3, 4]

result = reduce(
    lambda x, y: x + y,
    numbers,
    100
)

print(result)
````

## `reduce()` vs Loop

Both approaches produce the same result.

Use the one that makes your code easier to understand.

````python
numbers = [4, 8, 12, 16]

total = 0

for number in numbers:

    total += number

print(total)
````

````python
from functools import reduce

numbers = [4, 8, 12, 16]

total = reduce(
    lambda x, y: x + y,
    numbers
)

print(total)
````

## `map()` vs `filter()` vs `reduce()`

| Function | Purpose | Output |
|----------|---------|--------|
| `map()` | Transform every element | Iterable of transformed elements |
| `filter()` | Select elements based on a condition | Iterable of filtered elements |
| `reduce()` | Combine elements into a single value | Single value |

## `zip()`

`zip()` combines elements from two or more iterables based on their positions.

### Syntax

```python
zip(iterable1, iterable2, ...)
```

````python
students = ["Aarav", "Diya", "Kabir"]

marks = [85, 92, 78]

result = zip(students, marks)

print(list(result))
````

## `zip()` with Multiple Iterables

`zip()` can combine more than two iterables.

````python
names = ["Aman", "Neha", "Riya"]

ages = [21, 23, 22]

cities = ["Delhi", "Indore", "Pune"]

result = zip(names, ages, cities)

print(list(result))
````

## Different Length Iterables

If iterables have different lengths, `zip()` stops when the shortest iterable ends.

````python
letters = ["A", "B", "C", "D"]

numbers = [1, 2]

result = zip(letters, numbers)

print(list(result))
````

````python
employees = ["Rahul", "Priya", "Karan"]

departments = ["HR", "IT", "Sales"]

employee_data = list(zip(employees, departments))

print(employee_data)
````

## Unzipping

Use the `*` operator to separate zipped data back into individual iterables.

````python
records = [

    ("Pen", 20),
    ("Notebook", 50),
    ("Marker", 35)

]

items, prices = zip(*records)

print(items)

print(prices)
````

## `enumerate()`

`enumerate()` adds an index to each element of an iterable.

It is commonly used when both the **index** and **value** are required during iteration.

### Syntax

```python
enumerate(iterable, start=0)
```

````python
fruits = ["Apple", "Mango", "Orange", "Banana"]

for index, fruit in enumerate(fruits):

    print(index, fruit)
````

## Custom Starting Index

Use the `start` parameter to begin indexing from a custom value.

````python
subjects = ["Python", "SQL", "Power BI", "Machine Learning"]

for index, subject in enumerate(subjects, start=1):

    print(index, subject)
````

## Converting to a List

`enumerate()` returns an enumerate object.

Convert it to a list when required.

````python
colors = ["Red", "Green", "Blue"]

result = list(enumerate(colors))

print(result)
````

````python
cities = ["Delhi", "Indore", "Jaipur"]

result = list(enumerate(cities, start=101))

print(result)
````

## `enumerate()` vs Manual Counter

| Manual Counter | `enumerate()` |
|----------------|---------------|
| More code | Less code |
| Counter managed manually | Counter handled automatically |
| More error-prone | Cleaner and more readable |

````python
# Manual Counter

languages = ["Python", "Java", "C++"]

counter = 1

for language in languages:

    print(counter, language)

    counter += 1
````

````python
# Using enumerate()

languages = ["Python", "Java", "C++"]

for index, language in enumerate(languages, start=1):

    print(index, language)
````

## `any()`

`any()` returns **True** if at least one element in the iterable is `True`.

### Syntax

```python
any(iterable)
```

````python
values = [False, False, True, False]

print(any(values))
````

````python
numbers = [0, 0, 15, 0]

print(any(numbers))
````

## `all()`

`all()` returns **True** only if every element in the iterable is `True`.

### Syntax

```python
all(iterable)
```

````python
values = [True, True, True]

print(all(values))
````

````python
numbers = [5, 8, 10, 12]

result = all(
    number > 0 for number in numbers
)

print(result)
````

## Using `any()` with Conditions

````python
marks = [42, 55, 68, 91]

has_distinction = any(
    mark >= 90 for mark in marks
)

print(has_distinction)
````

## Using `all()` with Conditions

````python
ages = [21, 25, 32, 19]

all_adults = all(
    age >= 18 for age in ages
)

print(all_adults)
````

## `any()` vs `all()`

| `any()` | `all()` |
|----------|----------|
| Returns `True` if at least one element is `True` | Returns `True` only if every element is `True` |
| Stops when it finds the first `True` | Stops when it finds the first `False` |

````python
values = [0, 5, 10]

print("any() :", any(values))
print("all() :", all(values))
````

## Functional Programming Summary

| Function | Purpose |
|----------|---------|
| `map()` | Transform each element |
| `filter()` | Select elements matching a condition |
| `reduce()` | Reduce iterable to a single value |
| `zip()` | Combine multiple iterables |
| `enumerate()` | Iterate with index |
| `any()` | Check if at least one element is `True` |
| `all()` | Check if all elements are `True` |
