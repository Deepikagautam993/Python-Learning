```python
# ================================
# 🦋 BUTTERFLY PATTERN
# ================================

n = 4

# upper part
for i in range(1, n+1):

    for j in range(i):
        print("*", end=" ")

    for j in range(2*(n-i)):
        print(" ", end=" ")

    for j in range(i):
        print("*", end=" ")

    print()

# lower part
for i in range(n-1, 0, -1):

    for j in range(i):
        print("*", end=" ")

    for j in range(2*(n-i)):
        print(" ", end=" ")

    for j in range(i):
        print("*", end=" ")

    print()
```

```python
# ================================
# 🔺 HOLLOW PYRAMID
# ================================

n = 5

for i in range(1, n+1):

    for j in range(n-i):
        print(" ", end=" ")

    for j in range(1, 2*i):

        if j == 1 or j == 2*i-1 or i == n:
            print("*", end=" ")
        else:
            print(" ", end=" ")

    print()
```

```python
# ================================
# 🔢 FLOYD'S TRIANGLE
# ================================

n = 5
num = 1

for i in range(1, n+1):

    for j in range(i):
        print(num, end=" ")
        num += 1

    print()
```

```python
# ================================
# 🔢 0-1 TRIANGLE
# ================================

n = 5

for i in range(n):

    for j in range(i+1):

        if (i + j) % 2 == 0:
            print(1, end=" ")
        else:
            print(0, end=" ")

    print()
```

```python
# ================================
# 🔤 CHARACTER PATTERN
# ================================

n = 5

for i in range(n):

    for j in range(i+1):
        print(chr(65 + j), end=" ")

    print()
```

```python
# ================================
# 💎 NUMBER DIAMOND
# ================================

n = 4

for i in range(1, n+1):

    for j in range(n-i):
        print(" ", end="")

    for j in range(1, i+1):
        print(j, end=" ")

    print()

for i in range(n-1, 0, -1):

    for j in range(n-i):
        print(" ", end="")

    for j in range(1, i+1):
        print(j, end=" ")

    print()
```
