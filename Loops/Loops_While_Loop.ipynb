# What is a Loop?

A loop is a programming structure that repeats a block of code again and again.

Example:

If we want to print Hello 5 times:

Without loop:
print("Hello")
print("Hello")
print("Hello")
print("Hello")
print("Hello")

Using loop:
Repeat print statement 5 times.

# Why do we use loops?

Loops help us:

- Reduce code repetition
- Write clean code
- Save time
- Handle large data efficiently

---

Types of loops in Python:

1. while loop
2. for loop

# while Loop

while loop executes a block of code as long as a condition is True.

Syntax:

while condition:
    statement

When condition becomes False, the loop stops.

```python
# Simple while loop

count = 1

while count <= 5:
    print(count)
    count += 1
```

# Working of while Loop

Example:

count = 1

Step 1:
Check condition

count <= 5

Step 2:
If True execute code

Step 3:
Update value

count = count + 1

Step 4:
Again check condition

```python
# Print numbers 1 to 10

number = 1

while number <= 10:
    print(number)
    number += 1
```

```python
# Reverse counting

number = 5

while number >= 1:
    print(number)
    number -= 1
```

# break Statement

break is used to stop the loop immediately.

After break, no further iterations execute.

```python
i = 1

while i <= 5:

    if i == 3:
        break

    print(i)

    i += 1
```

# continue Statement

continue skips the current iteration.

The loop continues with the next iteration.

```python
i = 1

while i <= 5:

    if i == 3:
        i += 1
        continue

    print(i)

    i += 1
```

# Difference

break:
Stops complete loop.

continue:
Only skips current iteration.

# while with else

else executes when loop completes normally.

If break occurs, else will not execute.

```python
i = 1

while i <= 3:
    print(i)
    i += 1

else:
    print("Loop Finished")
```

```python
i = 1

while i <= 5:

    if i == 3:
        break

    print(i)

    i += 1

else:
    print("Done")
```

# Assignment Questions

Q1. Print numbers 1 to 10

Q2. Print reverse counting 10 to 1

Q3. Print multiplication table

Q4. Find sum of 1 to n

Q5. Count digits in a number

Q6. Reverse a number

Q7. Check palindrome number

Q8. Find factorial

```python
# Q1 Print 1 to 10

i = 1

while i <= 10:
    print(i)
    i += 1



# Q2 Reverse counting

i = 10

while i >= 1:
    print(i)
    i -= 1



# Q3 Multiplication table

num = int(input("Enter number: "))

i = 1

while i <= 10:
    print(num, "x", i, "=", num*i)
    i += 1



# Q4 Sum of 1 to n

n = int(input("Enter n: "))

total = 0
i = 1

while i <= n:
    total += i
    i += 1

print(total)



# Q5 Count digits

num = int(input("Enter number: "))

count = 0

while num > 0:
    count += 1
    num //= 10

print(count)
```

```python
# Q6 Reverse Number

num = int(input("Enter number: "))

reverse = 0

while num > 0:

    digit = num % 10

    reverse = reverse * 10 + digit

    num //= 10

print(reverse)



# Q7 Palindrome

num = int(input("Enter number: "))

original = num
reverse = 0

while num > 0:

    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10


if original == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")



# Q8 Factorial

n = int(input("Enter number: "))

fact = 1
i = 1

while i <= n:
    fact *= i
    i += 1

print(fact)
```
