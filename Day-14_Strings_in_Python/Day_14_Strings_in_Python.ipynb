# DAY-14: STRINGS IN PYTHON

## Topic: Introduction to Strings

A string is a sequence of characters enclosed in quotes.

Strings are **immutable**, meaning they cannot be changed after creation.

Used in:
- Data processing
- Web development
- AI/ML preprocessing
- Text analysis

```python
# String creation

s = "Python Programming"

print(s)
print(type(s))
```

# STRING INDEXING

Each character in a string has an index.

Types:
- Positive indexing → starts from 0
- Negative indexing → starts from -1

```python
text = "Deepika"

print(text[0])   # D
print(text[1])   # e
print(text[-1])  # a
print(text[-2])  # k
```

# STRING SLICING

Used to extract part of a string.

Syntax:
string[start : end : step]

Rules:
- end is excluded
- step defines skipping

```python
text = "Deepika"

print(text[0:4])    # Deep
print(text[:3])     # Dee
print(text[3:])     # pika
print(text[0:7:2])  # Depk
```

# STRING REVERSAL

Used to reverse a string.

Very important interview question.

```python
text = "Deepika"

print(text[::-1])
```

# STRING IMMUTABILITY

Strings cannot be changed directly after creation.

```python
s = "hello"

# s[0] = "H"  ❌ Not allowed

# Correct way
s = "H" + s[1:]
print(s)
```

# STRING METHODS

Important methods:
- upper()
- lower()
- title()
- capitalize()

```python
s = "hello python"

print(s.upper())
print(s.lower())
print(s.title())
print(s.capitalize())
```

# SEARCH IN STRING

Used to find substring or check existence.

```python
s = "I love Python"

print(s.find("Python"))
print(s.count("o"))
print("Python" in s)
```

# REPLACE METHOD

Used to replace part of a string.

```python
s = "I love Python"

print(s.replace("Python", "Java"))
```

# SPLIT AND JOIN

Split → string to list  
Join → list to string

```python
s = "apple,banana,mango"

print(s.split(","))
```

```python
words = ["I", "love", "Python"]

print(" ".join(words))
```

# STRING FORMATTING

Used to insert variables inside strings.

Best method → f-string

```python
name = "Deepika"
age = 20

print(f"My name is {name} and age is {age}")
```

# PALINDROME CHECK

A string that reads same forward and backward.

```python
s = "madam"

if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

# VOWEL COUNT

Counts vowels in a string.

```python
s = "hello world"
vowels = "aeiou"

count = 0

for ch in s.lower():
    if ch in vowels:
        count += 1

print(count)
```

# CHARACTER FREQUENCY

```python
s = "aabbc"
freq = {}

for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

print(freq)
```

# FIRST NON REPEATING CHARACTER

```python
s = "aabbcde"

for ch in s:
    if s.count(ch) == 1:
        print(ch)
        break
```

```python
# Reverse a string without using slicing

s = "deepika"
rev = ""

for ch in s:
    rev = ch + rev

print(rev)
```

```python
# Count number of uppercase letters in a string

s = "DeepikaGautamPython"
count = 0

for ch in s:
    if ch.isupper():
        count += 1

print(count)
```

```python
# Remove all spaces from a string

s = "I love Python Programming"
result = ""

for ch in s:
    if ch != " ":
        result += ch

print(result)
```

```python
# Check if two strings are anagrams

s1 = "listen"
s2 = "silent"

if sorted(s1) == sorted(s2):
    print("Anagram")
else:
    print("Not Anagram")
```

```python
# Find the longest word in a string

s = "I love Python Programming very much"
words = s.split()

longest = ""

for w in words:
    if len(w) > len(longest):
        longest = w

print(longest)
```

```python
# Count digits in a string

s = "abc12345xyz"
count = 0

for ch in s:
    if ch.isdigit():
        count += 1

print(count)
```

```python
# Toggle uppercase to lowercase and lowercase to uppercase

s = "DeePiKa"
result = ""

for ch in s:
    if ch.isupper():
        result += ch.lower()
    else:
        result += ch.upper()

print(result)
```

# MINI PROJECT - TEXT ANALYZER

```python
text = input("Enter text: ")

print("Characters:", len(text))
print("Words:", len(text.split()))
print("Uppercase:", text.upper())
print("Lowercase:", text.lower())

vowels = "aeiou"
count = sum(1 for ch in text.lower() if ch in vowels)

print("Vowels:", count)
```
