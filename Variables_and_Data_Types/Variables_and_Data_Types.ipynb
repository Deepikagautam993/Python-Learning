# 🏷️ Variables in Python

A variable is a name given to a memory location where we store data.

A variable works like a container that stores values.

Example:

name = "Deepika"

Here:

name → Variable

"Deepika" → Value

Variables help us store and reuse data easily.

```python
# Creating a variable

name = "Deepika"

print(name)
```

# Why Do We Use Variables?

Without variables:

We have to write the same value again and again.

Example:

print("Deepika")
print("Deepika")

Using variables:

name = "Deepika"

print(name)
print(name)

Code becomes cleaner and easier to maintain.

```python
name = "Deepika"

print(name)
print(name)
print(name)
```

# Creating Variables

Syntax:

variable_name = value


Example:

age = 20

Here:

age → variable

20 → value


Python automatically understands the data type.

```python
age = 20

print(age)
```

# Assignment Operator (=)

The = operator is used to assign a value to a variable.

Example:

x = 10

Meaning:

Store 10 inside variable x.

```python
x = 10

print(x)
```

# Multiple Variables

Python allows us to create multiple variables.

Example:

name = "Deepika"
age = 20
course = "AIML"

```python
name = "Deepika"
age = 20
course = "AIML"

print(name)
print(age)
print(course)
```

# Variable Naming Rules

Rules:

1. Variable can contain letters, numbers and underscore.

Correct:

student_name = "Deepika"

age1 = 20


2. Variable cannot start with a number.

Wrong:

1name = "Python"


3. Spaces are not allowed.

Wrong:

student name = "A"


4. Keywords cannot be used as variable names.

Wrong:

class = "AIML"

# Dynamic Typing

Python is dynamically typed.

We don't need to declare datatype.

Example:

x = 10

Later:

x = "Python"


Python automatically changes the type.

```python
x = 10

print(x)

x = "Python"

print(x)
```

# Data Types in Python

Data type tells what type of value a variable stores.

Main data types:

1. Integer (int)
2. Float
3. String
4. Boolean

```python
age = 20
height = 5.6
name = "Deepika"
is_student = True

print(type(age))
print(type(height))
print(type(name))
print(type(is_student))
```

# Type Conversion (Type Casting)

Type conversion means changing one data type into another.

Functions:

int()

str()

float()

```python
num = "100"

num = int(num)

print(num + 50)
```

# input() and Type Conversion

input() always returns string.

Example:

age = input("Enter age:")

The entered value is stored as string.

For calculations:

age = int(input("Enter age:"))

```python
age = int(input("Enter your age: "))

print(age + 5)
```

# Multiple Assignment

We can assign values to multiple variables in one line.

Example:

a, b, c = 10, 20, 30

```python
a, b, c = 10, 20, 30

print(a)
print(b)
print(c)
```

# Variable Swapping

Python allows swapping without using third variable.

Example:

a, b = b, a

```python
a = 10
b = 20

a, b = b, a

print(a)
print(b)
```

# 🔍 Checking Data Type

We can check the datatype of a variable using:

type()

Example:

type(variable)

It returns the class/type of stored value.

```python
name = "Python"
age = 10
price = 99.5

print(type(name))
print(type(age))
print(type(price))
```

# ⭐ Important Points

- Variable stores data
- = is an assignment operator
- Python automatically detects datatype
- Variable values can be changed
- Python is dynamically typed
- input() returns string
- Type conversion changes one datatype into another
- type() is used to check datatype
