# 🚀 Pattern Problems in Python

This notebook contains all important pattern programs from basic to advanced level.

We focus on:
- loop logic
- nested loops
- pattern building thinking

## 📌 Pattern Logic Rule

Every pattern follows 3 steps:

1. Rows → outer loop  
2. Columns → inner loop  
3. Condition → print logic  

👉 Always think row-wise first

```python
# ============================
# ⭐ SQUARE PATTERN (n = 4)
# ============================

n = 4

for i in range(n):              # i = row (0 to 3)

    for j in range(n):          # j = column (0 to 3)

        print("*", end=" ")     # every cell = star

    print()                     # new line after each row
```

```python
# ============================
# ⭐ RIGHT TRIANGLE PATTERN
# ============================

n = 4

for i in range(1, n+1):          # i = row (1 to 4)

    for j in range(i):          # j depends on row

        print("*", end=" ")

    print()
```

```python
# ============================
# 🔺 PYRAMID PATTERN
# ============================

n = 4

for i in range(1, n+1):          # rows

    for j in range(n - i):       # spaces

        print(" ", end="")

    for j in range(i):           # stars

        print("* ", end="")

    print()
```

```python
# ============================
# 🔻 INVERTED PYRAMID
# ============================

n = 4

for i in range(n, 0, -1):        # rows reverse

    for j in range(n - i):       # spaces increase

        print(" ", end="")

    for j in range(i):           # stars decrease

        print("* ", end="")

    print()
```

```python
# ============================
# 🟦 HOLLOW SQUARE
# ============================

n = 4

for i in range(n):          # row

    for j in range(n):      # column

        if i == 0 or i == n-1 or j == 0 or j == n-1:

            print("*", end=" ")

        else:

            print(" ", end=" ")

    print()
```

```python
# ============================
# 💎 DIAMOND PATTERN
# ============================

n = 4

for i in range(1, n+1):

    for j in range(n - i):
        print(" ", end="")

    for j in range(i):
        print("* ", end="")

    print()

for i in range(n-1, 0, -1):

    for j in range(n - i):
        print(" ", end="")

    for j in range(i):
        print("* ", end="")

    print()
```

```python
# ============================
# 🔢 NUMBER INCREASING TRIANGLE
# ============================

n = 4

for i in range(1, n+1):

    for j in range(1, i+1):

        print(j, end=" ")

    print()
```

```python
# ============================
# 🔢 SAME NUMBER TRIANGLE
# ============================

n = 4

for i in range(1, n+1):

    for j in range(i):

        print(i, end=" ")

    print()
```

```python
# ============================
# 🔢 REVERSE NUMBER TRIANGLE
# ============================

n = 4

for i in range(n, 0, -1):

    for j in range(1, i+1):

        print(j, end=" ")

    print()
```

```python
# ============================
# 🔺 HOLLOW TRIANGLE
# ============================

n = 4

for i in range(1, n+1):

    for j in range(1, i+1):

        if j == 1 or j == i or i == n:
            print("*", end=" ")
        else:
            print(" ", end=" ")

    print()
```

```python
# ============================
# 🔢 NUMBER PYRAMID
# ============================

n = 4

for i in range(1, n+1):

    for j in range(n - i):
        print(" ", end="")

    for j in range(1, i+1):
        print(j, end=" ")

    print()
```

```python
# ============================
# 💎 NUMBER DIAMOND
# ============================

n = 4

for i in range(1, n+1):

    for j in range(n - i):
        print(" ", end="")

    for j in range(1, i+1):
        print(j, end=" ")

    print()

for i in range(n-1, 0, -1):

    for j in range(n - i):
        print(" ", end="")

    for j in range(1, i+1):
        print(j, end=" ")

    print()
```
