# Output in Python

Output means displaying the result of a program.

Python uses the print() function to show output on the screen.

Syntax:

print(value)

Example:

print("Hello")

```python
name = "Deepika"

print(name)
```

# Printing Multiple Values

We can print multiple values using a single print statement.

Example:

print(value1, value2)

By default, values are separated by space.

```python
name = "Deepika"
age = 20

print(name, age)
```

# sep Parameter

sep is used to change the separator between multiple values.

Default:

space

Example:

print("Python","Java",sep="-")

```python
print("Python", "Java", sep="-")
```

# end Parameter

By default print() moves to the next line.

Using end parameter we can change this behavior.

```python
print("Hello", end=" ")

print("Python")
```

# Input Function

Input is used to take data from user.

Syntax:

variable = input("message")

Important:

input() always returns string.

```python
name = input("Enter your name: ")

print(name)
```

# Type Conversion with Input

Since input() returns string,
we convert it when we need numbers.

Common conversions:

int()
float()
str()

```python
age = int(input("Enter age: "))

print(age + 5)
```

# Operators in Python

Operators are symbols used to perform operations on values.

Types:

1. Arithmetic Operators
2. Comparison Operators
3. Logical Operators
4. Assignment Operators

# Arithmetic Operators

Used for mathematical calculations.

+ Addition

- Subtraction

* Multiplication

/ Division

// Floor Division

% Modulus

** Power

```python
a = 10
b = 3

print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)
print(a%b)
print(a**b)
```

# Comparison Operators

Comparison operators compare two values.

They return:

True or False

Operators:

==
!=
>
<
>=
<=

```python
print(10 == 10)
print(10 != 5)
print(20 > 10)
print(5 < 2)
```

# Logical Operators

Logical operators combine conditions.

Operators:

and
or
not

```python
age = 20

print(age > 18 and age < 30)

print(age < 18 or age > 18)

print(not True)
```

# Assignment Operators

Used to assign and update values.

=

+=

-=

*=

/=

```python
x = 10

x += 5

print(x)

x -= 3

print(x)
```

```python
# =====================================
# Day 03 - Input, Output and Operators
# Assignment Solutions
# =====================================


# Q1. Create a login system using username and password

username = input("Enter username: ")
password = input("Enter password: ")

if username == "admin" and password == "123":
    print("Login Successful")
else:
    print("Invalid Login")



# Q2. Create a simple calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)
print("Division:", num1 / num2)



# Q3. Take age input and check Adult or Minor

age = int(input("Enter your age: "))

if age >= 18:
    print("Adult")
else:
    print("Minor")



# Q4. Take marks input and print grade

marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 40:
    print("Grade C")
else:
    print("Fail")



# Q5. Create a program using logical operators

age = int(input("Enter age: "))
has_id = True

if age >= 18 and has_id:
    print("You can enter")
else:
    print("Entry not allowed")
```

```python
# Q6. Check even or odd

number = int(input("Enter number: "))

if number % 2 == 0:
    print("Even")
else:
    print("Odd")



# Q7. Compare two numbers

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("First number is greater")
elif b > a:
    print("Second number is greater")
else:
    print("Both are equal")



# Q8. Student pass fail program

marks = int(input("Enter marks: "))

if marks >= 40:
    print("Pass")
else:
    print("Fail")
```
