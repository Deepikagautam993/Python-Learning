# 🚀 Object Oriented Programming (OOP) in Python

Object-Oriented Programming is a programming approach
where we organize code using classes and objects.

OOP helps us to:
- Write reusable code
- Manage large projects
- Represent real-world entities

Topics Covered:

- Class and Object
- Attributes and Methods
- Constructor
- Destructor
- Encapsulation
- Inheritance
- Polymorphism
- Abstraction
- Class Method
- Static Method
- Mini Projects

# 📌 Procedural Programming vs OOP

Procedural Programming:

- Program is divided into functions
- Focus is on steps

OOP:

- Program is divided into objects
- Focus is on data and behaviour

Example:

Student is an object.

Data:
- name
- age
- course

Behaviour:
- study()
- attend_class()

# 📌 Class

Class is a blueprint or template
used to create objects.

It contains:
- Attributes (variables)
- Methods (functions)

```python
# Creating a class

class Student:
    pass


# Creating object

student1 = Student()

print(student1)
```

# 📌 Object

Object is an instance of a class.

Class = Blueprint

Object = Real entity created from blueprint


Example:

Class:
Car

Objects:
BMW, Audi, Tesla

```python
class Student:

    name = "Deepika"
    course = "AIML"


# Creating object

student1 = Student()


print(student1.name)
print(student1.course)
```

# 📌 Attributes

Attributes are variables inside a class.

Types:

1. Instance Attributes
2. Class Attributes

```python
class Student:

    # Class attribute
    college = "BGIEM"


    def __init__(self, name, age):

        # Instance attributes
        self.name = name
        self.age = age



s1 = Student("Deepika",20)


print(s1.name)
print(s1.age)
print(s1.college)
```

# 📌 Methods

Functions inside a class are called methods.

Methods define object behaviour.

```python
class Student:


    def study(self):

        print("Student is studying")



s1 = Student()


s1.study()
```

# 📌 self Keyword

self represents the current object.

It allows us to access:
- Instance variables
- Methods

```python
class Student:


    def show(self):

        print("Hello Student")

s1 = Student()
s1.show()
```

# 📌 Constructor (__init__)

Constructor is a special method.

It automatically runs when object is created.

Used to initialize object data.

```python
class Student:


    def __init__(self, name, age):

        self.name = name
        self.age = age

s1 = Student("Deepika",20)

print(s1.name)
print(s1.age)
```

# 📌 Destructor (__del__)

Destructor is called when object is destroyed.

It is used for cleanup tasks.

```python
class Student:


    def __init__(self):

        print("Object created")


    def __del__(self):

        print("Object destroyed")

s1 = Student()

del s1
```

# 🚀 Encapsulation

Encapsulation means wrapping data and methods
together inside a class.

It also provides data protection by controlling access.

Example:

Bank Account

Data:
- balance
- account_number

Methods:
- deposit()
- withdraw()

# 📌 Access Modifiers

Python has three types of access control:

1. Public
2. Protected
3. Private

```python
# Public variable

class Student:

    def __init__(self):

        self.name = "Deepika"

s1 = Student()

print(s1.name)
```

# Protected Variable

Protected variables are created using single underscore (_).

They can be accessed inside class and child classes.

```python
class Student:


    def __init__(self):

        self._marks = 90
s1 = Student()

print(s1._marks)
```

# Private Variable

Private variables are created using double underscore (__).

They cannot be accessed directly outside class.

Used for data hiding.

```python
class Bank:


    def __init__(self):

        self.__balance = 5000

    def show_balance(self):

        print(self.__balance)

account = Bank()
account.show_balance()
```

# Getter and Setter

Getter:
Used to read private data.

Setter:
Used to update private data.

```python
class Student:

    def __init__(self,name):

        self.__name = name

    # Getter

    def get_name(self):

        return self.__name

    # Setter

    def set_name(self,name):

        self.__name = name

s1 = Student("Deepika")

print(s1.get_name())


s1.set_name("Rahul")

print(s1.get_name())
```

# 🚀 Inheritance

Inheritance allows one class to use
properties and methods of another class.

Parent Class:
Existing class

Child Class:
New class

```python
# Parent class

class Animal:

    def sound(self):
        print("Animal sound")

# Child class

class Dog(Animal):

    def bark(self):
        print("Dog barking")

d = Dog()
d.sound()
d.bark()
```

# Types of Inheritance

1. Single Inheritance
2. Multiple Inheritance
3. Multilevel Inheritance
4. Hierarchical Inheritance
5. Hybrid Inheritance

# 1. Single Inheritance

One parent class and one child class.

```python
class Parent:

    def show(self):
        print("Parent")

class Child(Parent):

    pass

obj = Child()
obj.show()
```

# 2. Multiple Inheritance

One child class with multiple parent classes.

```python
class A:
    def methodA(self):

        print("Class A")

class B:

    def methodB(self):

        print("Class B")

class C(A,B):
    pass

obj = C()
obj.methodA()
obj.methodB()
```

# 3. Multilevel Inheritance

Inheritance chain.

A → B → C

```python
class A:

    def showA(self):

        print("A")

class B(A):

    def showB(self):

        print("B")

class C(B):

    def showC(self):

        print("C")

obj = C()

obj.showA()

obj.showB()

obj.showC()
```

# 4. Hierarchical Inheritance

One parent class and multiple child classes.

```python
class Animal:
    def eat(self):

        print("Eating")

class Dog(Animal):

    pass

class Cat(Animal):

    pass

d = Dog()

c = Cat()

d.eat()

c.eat()
```

# super() Function

super() is used to call parent class methods
and constructors.

```python
class Parent:

    def __init__(self):

        print("Parent constructor")

class Child(Parent):
    def __init__(self):

        super().__init__()

        print("Child constructor")

obj = Child()
```

# 🚀 Polymorphism

Polymorphism means:

One name, many forms.

Same method can behave differently.

```python
class Dog:

    def sound(self):

        print("Bark")

class Cat:


    def sound(self):

        print("Meow")

animals = [Dog(), Cat()]

for animal in animals:

    animal.sound()
```

# Method Overriding

Child class changes parent class method.

```python
class Animal:

    def sound(self):

        print("Animal sound")

class Dog(Animal):
    def sound(self):

        print("Dog bark")

d = Dog()
d.sound()
```

# Method Overloading

Python does not support traditional
method overloading.

We achieve it using default arguments.

```python
class Calculator:

    def add(self,a,b,c=0):

        return a+b+c

obj = Calculator()

print(obj.add(2,3))

print(obj.add(2,3,4))
```

# 🚀 Abstraction

Abstraction means hiding implementation details
and showing only essential features.

Python uses abc module.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):

        pass

class Car(Vehicle):
    def start(self):

        print("Car starts")

c = Car()
c.start()
```

# Class Method

Class method works with class level data.

Uses:
@classmethod

```python
class Student:
    college = "BGIEM"

    @classmethod
    def show_college(cls):

        print(cls.college)

Student.show_college()
```

# Static Method

Static method does not use
class or object data.

Uses:
@staticmethod

```python
class Math:
    @staticmethod
    def add(a,b):

        return a+b

print(Math.add(5,10))
```

# 🚀 Mini Project 2

Bank Account System

Features:

- Deposit
- Withdraw
- Balance check

```python
class BankAccount:

    def __init__(self,balance):

        self.__balance = balance

    def deposit(self,amount):

        self.__balance += amount

    def withdraw(self,amount):

        if amount <= self.__balance:

            self.__balance -= amount

        else:

            print("Insufficient balance")

    def show_balance(self):

        print("Balance:",self.__balance)

account = BankAccount(5000)

account.deposit(1000)

account.withdraw(2000)

account.show_balance()
```
