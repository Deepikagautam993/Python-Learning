# ✅ Question 18

Modify the Bank Account class.

Do not allow:

- Negative Deposit
- Negative Withdraw

```python
class BankAccount:

    def __init__(self):
        self.__balance = 10000

    def deposit(self, amount):

        if amount > 0:
            self.__balance += amount

        else:
            print("Invalid Deposit Amount")

    def withdraw(self, amount):

        if amount <= 0:
            print("Invalid Withdrawal Amount")

        elif amount > self.__balance:
            print("Insufficient Balance")

        else:
            self.__balance -= amount

    def show_balance(self):
        print("Balance :", self.__balance)


account = BankAccount()

account.deposit(-500)

account.withdraw(-100)

account.show_balance()
```

# ✅ Question 46 ⭐⭐⭐⭐

Setter Validation

Marks should never exceed 100.

```python
# Solution 46

class Student:

    def __init__(self):
        self._marks = 0

    @property
    def marks(self):
        return self._marks

    @marks.setter
    def marks(self, value):

        if 0 <= value <= 100:
            self._marks = value

        else:
            print("Invalid Marks")


student = Student()

student.marks = 95

print(student.marks)

student.marks = 120
```

```python
# Solution 60

class File:

    def __init__(self, filename):
        self.filename = filename
        print(f"{self.filename} opened")

    def __del__(self):
        print(f"{self.filename} closed")


file = File("data.txt")

del file
```
