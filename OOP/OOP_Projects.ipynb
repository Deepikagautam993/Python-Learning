# 🏫 Project 1: Student Management System

## 🎯 Problem Statement

Develop a **Student Management System** using Object-Oriented Programming (OOP) in Python.

The system should manage student information efficiently by storing student details and providing methods to perform common academic operations.

---

## ✨ Features

- Add student details
- Display student information
- Calculate percentage
- Check Pass/Fail status
- Update student marks

---

## 🛠️ OOP Concepts Used

- Class
- Object

- Instance Variables
- Instance Methods
- Encapsulation (Basic)

---

## 🎯 Learning Objective

This project demonstrates how Object-Oriented Programming can model real-world entities. Instead of storing student data in separate variables, we create a `Student` class that combines both data (attributes) and behavior (methods), making the code more organized, reusable, and scalable.

---

```python
# ==========================================================
# Project 1 : Student Management System
# ==========================================================

class Student:
    """
    A class to represent a student.
    """

    def __init__(self, name, age, roll_no, marks):
        self.name = name
        self.age = age
        self.roll_no = roll_no
        self.marks = marks

    def display_details(self):
        print("\n" + "=" * 40)
        print("Student Details")
        print("=" * 40)
        print(f"Name      : {self.name}")
        print(f"Age       : {self.age}")
        print(f"Roll No   : {self.roll_no}")
        print(f"Marks     : {self.marks}/500")

    def calculate_percentage(self):
        return (self.marks / 500) * 100

    def check_result(self):
        if self.calculate_percentage() >= 40:
            return "Pass ✅"
        return "Fail ❌"

    def update_marks(self, new_marks):
        self.marks = new_marks
        print("\nMarks updated successfully!")


# -----------------------------
# Creating Student Objects
# -----------------------------

student1 = Student("Deepika", 21, 101, 430)
student2 = Student("Rahul", 22, 102, 350)

# -----------------------------
# Display Student Details
# -----------------------------

student1.display_details()
print(f"Percentage : {student1.calculate_percentage():.2f}%")
print(f"Result     : {student1.check_result()}")

student2.display_details()
print(f"Percentage : {student2.calculate_percentage():.2f}%")
print(f"Result     : {student2.check_result()}")

# -----------------------------
# Updating Marks
# -----------------------------

student2.update_marks(450)

student2.display_details()
print(f"Percentage : {student2.calculate_percentage():.2f}%")
print(f"Result     : {student2.check_result()}")
```

## 🎯 Learning Objective

This project demonstrates how OOP can be used to model a real-world banking system. It focuses on organizing account data and banking operations inside a single class while validating user actions such as deposits and withdrawals.

```python
# ==========================================================
# Project 2 : Bank Account Management System
# ==========================================================

class BankAccount:

    def __init__(self, account_holder, account_number, balance=0):
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = balance

    def display_details(self):
        print("\n" + "=" * 40)
        print("Account Details")
        print("=" * 40)
        print(f"Account Holder : {self.account_holder}")
        print(f"Account Number : {self.account_number}")
        print(f"Current Balance: ₹{self.balance}")

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"₹{amount} deposited successfully.")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount.")
        elif amount > self.balance:
            print("Insufficient Balance!")
        else:
            self.balance -= amount
            print(f"₹{amount} withdrawn successfully.")

    def check_balance(self):
        print(f"Available Balance : ₹{self.balance}")


# ----------------------------------------
# Creating Objects
# ----------------------------------------

account1 = BankAccount("Deepika Gautam", 100101, 5000)

account1.display_details()

account1.deposit(2000)

account1.withdraw(1500)

account1.withdraw(10000)

account1.check_balance()
```

## 🎯 Problem Statement

Develop a **Library Management System** using Object-Oriented Programming (OOP).

The system should help a librarian manage books in a library. Users should be able to issue books, return books, and view the available books.

---

## ✨ Features

- Display all available books
- Issue a book
- Return a book
- Prevent issuing unavailable books
- Add returned books back to the library

---

## 🛠️ OOP Concepts Used

- Conditional Statements

---

## 🎯 Learning Objective

This project demonstrates how a real-world library can be modeled using Object-Oriented Programming. It also shows how objects manage and modify shared data efficiently.

```python
# ==========================================
# Project 3 : Library Management System
# ==========================================

class Library:

    def __init__(self, books):
        self.books = books

    def show_books(self):
        print("\n📚 Available Books")

        if len(self.books) == 0:
            print("No books available.")
            return

        for index, book in enumerate(self.books, start=1):
            print(f"{index}. {book}")

    def issue_book(self, book_name):
        if book_name in self.books:
            self.books.remove(book_name)
            print(f'"{book_name}" has been issued successfully.')
        else:
            print(f'Sorry! "{book_name}" is not available.')

    def return_book(self, book_name):
        if book_name in self.books:
            print(f'"{book_name}" is already available in the library.')
        else:
            self.books.append(book_name)
            print(f'Thank you! "{book_name}" has been returned.')


# Initial Book Collection
book_list = [
    "Python Programming",
    "Data Structures",
    "Machine Learning",
    "Operating System",
    "Database Management"
]

# Create Library Object
library = Library(book_list)

# Display Books
library.show_books()

# Issue Books
library.issue_book("Python Programming")
library.issue_book("Java Programming")

# Display Updated List
library.show_books()

# Return Books
library.return_book("Python Programming")

# Display Final List
library.show_books()
```

## 🎯 Learning Objective

This project demonstrates how OOP can be used to manage employee records and automate salary calculations. It also shows how methods inside a class can perform different operations using object data.

```python
# ==========================================
# Project 4 : Employee Payroll System
# ==========================================

class Employee:

    def __init__(self, emp_id, name, department, basic_salary):
        self.emp_id = emp_id
        self.name = name
        self.department = department
        self.basic_salary = basic_salary

    def calculate_salary(self):
        bonus = self.basic_salary * 0.10
        deduction = self.basic_salary * 0.05
        net_salary = self.basic_salary + bonus - deduction
        return bonus, deduction, net_salary

    def display_salary_slip(self):
        bonus, deduction, net_salary = self.calculate_salary()

        print("\n" + "=" * 40)
        print("        Employee Salary Slip")
        print("=" * 40)
        print(f"Employee ID   : {self.emp_id}")
        print(f"Name          : {self.name}")
        print(f"Department    : {self.department}")
        print(f"Basic Salary  : ₹{self.basic_salary:.2f}")
        print(f"Bonus (10%)   : ₹{bonus:.2f}")
        print(f"Deduction(5%) : ₹{deduction:.2f}")
        print("-" * 40)
        print(f"Net Salary    : ₹{net_salary:.2f}")
        print("=" * 40)


# Creating Employee Objects

emp1 = Employee(101, "Deepika Gautam", "AI/ML", 50000)
emp2 = Employee(102, "Rahul Sharma", "Software", 42000)

emp1.display_salary_slip()
emp2.display_salary_slip()
```

# 📦 Project 5: Inventory Management System

## 🎯 Problem Statement

Develop an **Inventory Management System** using Object-Oriented Programming (OOP).

The system should help a shop owner manage products and their stock. Users should be able to add products, update stock, sell products, and view the inventory.

---

## ✨ Features

- Add new products
- Display all products
- Update product stock
- Sell products
- Prevent selling when stock is insufficient

---

## 🛠️ OOP Concepts Used

- Class
- Object
- Constructor (`__init__`)
- Object Composition
- Instance Variables
- Instance Methods
- List of Objects

---

## 🎯 Learning Objective

This project demonstrates how multiple objects can work together in a real-world application. Instead of storing only strings or numbers, the inventory stores complete Product objects, making the design more modular and scalable.

```python
# ==========================================
# Project 5 : Inventory Management System
# ==========================================

class Product:

    def __init__(self, product_id, name, price, stock):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.stock = stock


class Inventory:

    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)
        print(f"{product.name} added successfully.")

    def show_products(self):
        print("\nAvailable Products")
        print("-" * 60)

        for product in self.products:
            print(
                f"ID: {product.product_id} | "
                f"Name: {product.name} | "
                f"Price: ₹{product.price} | "
                f"Stock: {product.stock}"
            )

    def sell_product(self, product_id, quantity):

        for product in self.products:

            if product.product_id == product_id:

                if quantity <= product.stock:
                    product.stock -= quantity
                    total = quantity * product.price

                    print(f"\n{quantity} {product.name} sold successfully.")
                    print(f"Total Bill : ₹{total}")

                else:
                    print("\nNot enough stock available.")

                return

        print("\nProduct not found.")

    def update_stock(self, product_id, new_stock):

        for product in self.products:

            if product.product_id == product_id:
                product.stock = new_stock
                print(f"Stock updated for {product.name}.")
                return

        print("Product not found.")


inventory = Inventory()

product1 = Product(101, "Laptop", 65000, 10)
product2 = Product(102, "Keyboard", 1200, 25)
product3 = Product(103, "Mouse", 700, 30)

inventory.add_product(product1)
inventory.add_product(product2)
inventory.add_product(product3)

inventory.show_products()

inventory.sell_product(101, 2)

inventory.update_stock(102, 40)

inventory.show_products()
```

# 🛒 Project 6: Shopping Cart System

## 🎯 Problem Statement

Develop a **Shopping Cart System** using Object-Oriented Programming (OOP).

The system should allow customers to add products to their cart, remove products, view cart items, and calculate the final bill.

---

## ✨ Features

- Create products
- Add products to cart
- Remove products from cart
- Display cart items
- Calculate total bill

---

## 🛠️ OOP Concepts Used

- Class
- Object
- Constructor (`__init__`)
- Object Composition
- List of Objects
- Instance Methods

---

## 🎯 Learning Objective

This project demonstrates how multiple classes interact with each other in a real-world application. It introduces the concept of object composition, where a Shopping Cart contains multiple Product objects.

```python
# ==========================================
# Project 6 : Shopping Cart System
# ==========================================

class Product:

    def __init__(self, product_id, name, price):
        self.product_id = product_id
        self.name = name
        self.price = price


class ShoppingCart:

    def __init__(self):
        self.cart_items = []

    def add_product(self, product):
        self.cart_items.append(product)
        print(f"{product.name} added to cart.")

    def remove_product(self, product_id):

        for product in self.cart_items:

            if product.product_id == product_id:
                self.cart_items.remove(product)
                print(f"{product.name} removed from cart.")
                return

        print("Product not found in cart.")

    def view_cart(self):

        if not self.cart_items:
            print("Your cart is empty.")
            return

        print("\nYour Shopping Cart")
        print("-" * 45)

        total = 0

        for product in self.cart_items:
            print(f"{product.name} - ₹{product.price}")
            total += product.price

        print("-" * 45)
        print(f"Total Bill : ₹{total}")


# -------------------------------
# Creating Products
# -------------------------------

laptop = Product(101, "Laptop", 65000)
mouse = Product(102, "Wireless Mouse", 1200)
keyboard = Product(103, "Mechanical Keyboard", 3500)
headphone = Product(104, "Headphones", 2500)

# -------------------------------
# Creating Cart
# -------------------------------

cart = ShoppingCart()

cart.add_product(laptop)
cart.add_product(mouse)
cart.add_product(keyboard)

cart.view_cart()

cart.remove_product(102)

cart.add_product(headphone)

cart.view_cart()
```

# 🏥 Project 7: Hospital Management System

## 🎯 Problem Statement

Develop a **Hospital Management System** using Object-Oriented Programming (OOP).

The system should manage doctors and patients while displaying their information separately. Since both doctors and patients are people, we will avoid code duplication by using **Inheritance**.

---

## ✨ Features

## 🛠️ OOP Concepts Used

- Class

- Instance Methods

---

## 🎯 Learning Objective

This project demonstrates how **Inheritance** helps in reducing code duplication. A common `Person` class is created, and both `Doctor` and `Patient` inherit its properties while adding their own unique attributes and methods.

```python
# ==========================================
# Project 7 : Hospital Management System
# ==========================================

class Person:

    def __init__(self, person_id, name, age):
        self.person_id = person_id
        self.name = name
        self.age = age


class Doctor(Person):

    def __init__(self, person_id, name, age, specialization):
        super().__init__(person_id, name, age)
        self.specialization = specialization

    def display_details(self):
        print("\nDoctor Details")
        print("-" * 30)
        print(f"Doctor ID      : {self.person_id}")
        print(f"Name           : {self.name}")
        print(f"Age            : {self.age}")
        print(f"Specialization : {self.specialization}")


class Patient(Person):

    def __init__(self, person_id, name, age, disease):
        super().__init__(person_id, name, age)
        self.disease = disease

    def display_details(self):
        print("\nPatient Details")
        print("-" * 30)
        print(f"Patient ID : {self.person_id}")
        print(f"Name       : {self.name}")
        print(f"Age        : {self.age}")
        print(f"Disease    : {self.disease}")


doctor1 = Doctor(101, "Dr. Sharma", 45, "Cardiologist")
doctor2 = Doctor(102, "Dr. Verma", 38, "Neurologist")

patient1 = Patient(201, "Deepika", 21, "Fever")
patient2 = Patient(202, "Rahul", 24, "Migraine")

doctor1.display_details()
doctor2.display_details()

patient1.display_details()
patient2.display_details()
```

## 🎯 Problem Statement

Develop an **ATM Management System** using Object-Oriented Programming (OOP).

The system should simulate the basic functionality of an ATM. A user should be able to verify their PIN, check account balance, deposit money, and withdraw money securely.

---

## ✨ Features

- Verify ATM PIN
- Check account balance
- Deposit money
- Withdraw money
- Prevent invalid transactions
- Protect account balance using encapsulation

---

## 🛠️ OOP Concepts Used

- Class
- Object
- Constructor (`__init__`)
- Encapsulation
- Private Variables
- Instance Methods
- Conditional Statements

---

## 🎯 Learning Objective

This project demonstrates how **Encapsulation** improves data security by hiding sensitive information like account balance and PIN. Users can access or modify these values only through class methods.

```python
# ==========================================
# Project 9 : ATM Management System
# ==========================================

class ATM:

    def __init__(self, account_holder, pin, balance):
        self.account_holder = account_holder
        self.__pin = pin
        self.__balance = balance

    def verify_pin(self, entered_pin):
        return entered_pin == self.__pin

    def check_balance(self):
        print(f"Available Balance : ₹{self.__balance}")

    def deposit(self, amount):
        if amount <= 0:
            print("Invalid deposit amount.")
            return

        self.__balance += amount
        print(f"₹{amount} deposited successfully.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount.")
            return

        if amount > self.__balance:
            print("Insufficient balance.")
            return

        self.__balance -= amount
        print(f"₹{amount} withdrawn successfully.")


# -------------------------------
# Creating ATM Object
# -------------------------------

atm = ATM("Deepika Gautam", 1234, 10000)

entered_pin = 1234

if atm.verify_pin(entered_pin):

    print(f"Welcome, {atm.account_holder}")

    atm.check_balance()

    atm.deposit(5000)

    atm.withdraw(3000)

    atm.check_balance()

else:
    print("Incorrect PIN!")
```

# 🛍️ Project 10: Mini E-Commerce Management System

## 🎯 Problem Statement

Develop a **Mini E-Commerce Management System** using Object-Oriented Programming (OOP).

The system should allow customers to browse products, add products to a shopping cart, and place an order while calculating the total bill.

This project is a simplified version of platforms like Amazon and Flipkart.

---

## ✨ Features

- Create multiple products
- Add products to cart
- View cart items
- Calculate total bill
- Place an order
- Display order summary

---

## 🛠️ OOP Concepts Used

- Class
- Object
- Constructor (`__init__`)
- Object Composition
- Encapsulation (Basic)
- List of Objects
- Instance Methods

---

## 🎯 Learning Objective

This project combines multiple OOP concepts into a single real-world application. It demonstrates how different classes collaborate to solve a practical business problem while keeping the code clean, modular, and easy to maintain.

```python
# =====================================================
# Project 10 : Mini E-Commerce Management System
# =====================================================

class Product:

    def __init__(self, product_id, name, price):
        self.product_id = product_id
        self.name = name
        self.price = price


class ShoppingCart:

    def __init__(self):
        self.items = []

    def add_product(self, product):
        self.items.append(product)
        print(f"{product.name} added to cart.")

    def show_cart(self):

        if not self.items:
            print("Your cart is empty.")
            return

        print("\nShopping Cart")
        print("-" * 40)

        total = 0

        for product in self.items:
            print(f"{product.name} - ₹{product.price}")
            total += product.price

        print("-" * 40)
        print(f"Total Bill : ₹{total}")

    def calculate_total(self):
        total = 0

        for product in self.items:
            total += product.price

        return total


class Customer:

    def __init__(self, customer_id, name):
        self.customer_id = customer_id
        self.name = name
        self.cart = ShoppingCart()

    def place_order(self):

        print("\nOrder Summary")
        print("=" * 40)
        print(f"Customer : {self.name}")

        self.cart.show_cart()

        print("\nOrder Placed Successfully 🎉")


# ------------------------------------------
# Creating Products
# ------------------------------------------

laptop = Product(101, "Laptop", 65000)
mouse = Product(102, "Wireless Mouse", 1200)
keyboard = Product(103, "Mechanical Keyboard", 3500)
headphones = Product(104, "Headphones", 2500)

# ------------------------------------------
# Creating Customer
# ------------------------------------------

customer = Customer(1, "Deepika Gautam")

customer.cart.add_product(laptop)
customer.cart.add_product(mouse)
customer.cart.add_product(keyboard)

customer.place_order()
```
